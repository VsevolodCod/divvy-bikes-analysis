````python
def clean_data(df: pl.DataFrame) -> pl.DataFrame:
    """
    Основная функция очистки данных
    """
    # Удаляем строки с пропущенными координатами (их мало)
    df = df.filter(
        pl.col("end_lat").is_not_null() & 
        pl.col("end_lng").is_not_null()
    )
    
    # Помечаем станции как "unknown" вместо None
    station_cols = ["start_station_name", "start_station_id", 
                    "end_station_name", "end_station_id"]
    
    for col in station_cols:
        df = df.with_columns(
            pl.when(pl.col(col).is_null())
            .then(pl.lit("unknown"))
            .otherwise(pl.col(col))
            .alias(col)
        )
    
    # Удаляем явные дубликаты по ride_id (на всякий случай)
    df = df.unique(subset=["ride_id"], keep="first")
    
    #колонка с длительностью поездки
    df = df.with_columns(
        ((pl.col("ended_at") - pl.col("started_at")).dt.total_seconds() / 60)
        .alias("duration_minutes")
        .cast(pl.Float64)
    )
    
    #некорректные длительности
    df = df.filter(
        (pl.col("duration_minutes") > 1) &  # Мин 1 минута
        (pl.col("duration_minutes") < 24 * 60)  # Макс 24 часа
    )
    
    # Фильтруем поездки, которые закончились раньше, чем начались
    df = df.filter(pl.col("ended_at") > pl.col("started_at"))
    
    #доп временные колонки для анализа
    df = df.with_columns([
        pl.col("started_at").dt.hour().alias("start_hour"),
        pl.col("started_at").dt.weekday().alias("start_weekday"),
        pl.col("started_at").dt.month().alias("start_month"),
        pl.col("started_at").dt.date().alias("start_date"),
    ])
    
    #корректность координат (Чикаго примерные границы)
    df = df.filter(
        (pl.col("start_lat").is_between(41.6, 42.1)) &
        (pl.col("start_lng").is_between(-87.95, -87.5)) &
        (pl.col("end_lat").is_between(41.6, 42.1)) &
        (pl.col("end_lng").is_between(-87.95, -87.5))
    )
    
    return df

df_clean = clean_data(df_full)
````

````python
def analyze_suspicious_data(df: pl.DataFrame):
    """
    Поиск потенциальных аномалий и краж
    """
    print("АНАЛИЗ ПОДОЗРИТЕЛЬНЫХ ДАННЫХ")
    
    # Поездки с очень большой длительностью
    long_rides = df.filter(pl.col("duration_minutes") > 12 * 60)  # >12 часов
    print(f"Поездки длительностью >12 часов: {len(long_rides)}")
    
    # Поездки без станций (возможно, самокат уехал далеко)
    no_station_rides = df.filter(
        (pl.col("start_station_name") == "unknown") & 
        (pl.col("end_station_name") == "unknown")
    )
    print(f"Поездки без стартовой и конечной станций: {len(no_station_rides)}")
    
    # Ночные поездки (возможно, воровство)
    night_rides = df.filter(
        (pl.col("start_hour").is_between(0, 4)) &  # С 00:00 до 04:00
        (pl.col("duration_minutes") > 180)  # >3 часов
    )
    print(f"Длинные ночные поездки (0-4 утра, >3ч): {len(night_rides)}")
    
    # Проверка расстояния между стартом и финишем
    # (нужна функция расчета расстояния по координатам)
    
    # Анализ по типам пользователей
    print("\nРаспределение по типам пользователей:")
    print(df.group_by("member_casual").agg(pl.count()).sort("count", descending=True))
    
    # Анализ по типам транспорта
    print("\nРаспределение по типам транспорта:")
    print(df.group_by("rideable_type").agg(pl.count()).sort("count", descending=True))
    
    return long_rides, no_station_rides, night_rides

long_rides, no_station_rides, night_rides = analyze_suspicious_data(df_clean)
````

````python
def check_station_consistency(df: pl.DataFrame):
    """
    Проверка консистентности названий и ID станций
    """
    print("\nПРОВЕРКА КОНСИСТЕНТНОСТИ СТАНЦИЙ")
    
    # маппинг station_name -> station_id
    start_mapping = df.filter(
        pl.col("start_station_name") != "unknown"
    ).select([
        "start_station_name", 
        "start_station_id"
    ]).unique()
    
    #конфликты (один name -> несколько id)
    conflicts = start_mapping.group_by("start_station_name").agg(
        pl.col("start_station_id").n_unique().alias("unique_ids")
    ).filter(pl.col("unique_ids") > 1)
    
    if len(conflicts) > 0:
        print(f"Найдено конфликтов в маппинге станций: {len(conflicts)}")
        for conflict in conflicts.head(5).rows():
            print(f"  Станция '{conflict[0]}' имеет {conflict[1]} разных ID")
    else:
        print("Конфликтов в маппинге станций не найдено")
    
    return start_mapping

station_mapping = check_station_consistency(df_clean)
````

````python
import polars as pl
import numpy as np

def clean_data_extended(df: pl.DataFrame) -> pl.DataFrame:
    """
    Расширенная очистка данных с учетом найденных проблем
    """
    # конфликт станции 'Loomis St & 89th St'
    df = df.with_columns([
        pl.when(pl.col("start_station_name") == "Loomis St & 89th St")
        .then(pl.lit("674"))  # Приводим к одному ID
        .otherwise(pl.col("start_station_id"))
        .alias("start_station_id"),
        
        pl.when(pl.col("end_station_name") == "Loomis St & 89th St")
        .then(pl.lit("674"))
        .otherwise(pl.col("end_station_id"))
        .alias("end_station_id")
    ])
    
    # Вычисляем расстояние между станциями (гаверсинус)
    R = 6371  

    df = df.with_columns([
        (pl.col("start_lat") * np.pi / 180).alias("start_lat_rad"),
        (pl.col("start_lng") * np.pi / 180).alias("start_lng_rad"),
        (pl.col("end_lat") * np.pi / 180).alias("end_lat_rad"),
        (pl.col("end_lng") * np.pi / 180).alias("end_lng_rad")
    ])
    

    df = df.with_columns([
        ((pl.col("end_lat_rad") - pl.col("start_lat_rad")).sin().pow(2) +
         pl.col("start_lat_rad").cos() * pl.col("end_lat_rad").cos() *
         (pl.col("end_lng_rad") - pl.col("start_lng_rad")).sin().pow(2))
        .alias("a")
    ])
    
    df = df.with_columns([
        (2 * pl.col("a").sqrt().arcsin() * R)
        .alias("distance_km")
    ])
    
    # Удаляем промежуточные колонки
    df = df.drop(["start_lat_rad", "start_lng_rad", "end_lat_rad", "end_lng_rad", "a"])
    
    #скорость (км/ч)
    df = df.with_columns([
        (pl.col("distance_km") / (pl.col("duration_minutes") / 60))
        .alias("speed_kmh")
    ])
    
    # аномальные скорости (>30 км/ч для велосипеда)
    df = df.filter(
        (pl.col("speed_kmh") <= 30) | 
        (pl.col("speed_kmh").is_null())
    )
    
    # поездки без станций
    df = df.with_columns([
        ((pl.col("start_station_name") == "unknown") | 
         (pl.col("end_station_name") == "unknown"))
        .alias("has_missing_stations")
    ])
    
    #доп проверка на аномалии
    df = df.with_columns([
        # Поездки, которые начались и закончились в одном месте
        ((pl.col("distance_km") < 0.01) & (pl.col("duration_minutes") > 5))
        .alias("same_location_long_time"),
        
        # Поездки с нулевой скоростью, но ненулевой длительностью
        ((pl.col("speed_kmh") < 0.1) & (pl.col("duration_minutes") > 10))
        .alias("zero_speed_trip")
    ])
    
    # категории поездок по длительности
    df = df.with_columns([
        pl.when(pl.col("duration_minutes") <= 10)
        .then(pl.lit("short"))
        .when(pl.col("duration_minutes") <= 30)
        .then(pl.lit("medium"))
        .when(pl.col("duration_minutes") <= 60)
        .then(pl.lit("long"))
        .otherwise(pl.lit("very_long"))
        .alias("trip_duration_category")
    ])
    
    # категории поездок по расстоянию
    df = df.with_columns([
        pl.when(pl.col("distance_km") <= 1)
        .then(pl.lit("very_short"))
        .when(pl.col("distance_km") <= 3)
        .then(pl.lit("short"))
        .when(pl.col("distance_km") <= 7)
        .then(pl.lit("medium"))
        .when(pl.col("distance_km") <= 15)
        .then(pl.lit("long"))
        .otherwise(pl.lit("very_long"))
        .alias("trip_distance_category")
    ])
    
    return df

df_final = clean_data_extended(df_clean)

df_final.write_parquet(SCOOTER_2021_PATH)
print(f"\nДанные сохранены в {SCOOTER_2021_PATH}")
````

````python
def create_final_report(df: pl.DataFrame):
    """
    Создание финального отчета по данным 2021 года
    """
    print("=" * 80)
    print("ФИНАЛЬНЫЙ ОТЧЕТ ПО ДАННЫМ ВЕЛОПРОКАТА ЧИКАГО 2021 ГОДА")
    print("=" * 80)
    
    total_rides = len(df)
    print(f"\nБАЗОВАЯ СТАТИСТИКА:")
    print(f"Всего поездок: {total_rides:,}")
    print(f"Период данных: {df['start_date'].min()} - {df['start_date'].max()}")
    print(f"Дней в данных: {df['start_date'].n_unique()}")
    
    print(f"\nРАСПРЕДЕЛЕНИЕ ПОЛЬЗОВАТЕЛЕЙ:")
    user_stats = df.group_by("member_casual").agg([
        pl.len().alias("count"),
        (pl.len() / total_rides * 100).alias("percentage"),
        pl.col("duration_minutes").mean().alias("avg_duration_min"),
        pl.col("distance_km").mean().alias("avg_distance_km"),
        pl.col("speed_kmh").mean().alias("avg_speed_kmh")
    ])
    print(user_stats)
    
    print(f"\nРАСПРЕДЕЛЕНИЕ ПО ТИПАМ ВЕЛОСИПЕДОВ:")
    bike_stats = df.group_by("rideable_type").agg([
        pl.len().alias("count"),
        (pl.len() / total_rides * 100).alias("percentage"),
        pl.col("duration_minutes").mean().alias("avg_duration_min"),
        pl.col("distance_km").mean().alias("avg_distance_km")
    ])
    print(bike_stats)

    missing_stations = df.filter(pl.col("has_missing_stations") == True)
    print(f"\nПОЕЗДКИ С ПРОПУЩЕННЫМИ СТАНЦИЯМИ:")
    print(f"Количество: {len(missing_stations):,} ({len(missing_stations)/total_rides*100:.2f}%)")
    print(f"Средняя длительность: {missing_stations['duration_minutes'].mean():.1f} мин")
    print(f"Среднее расстояние: {missing_stations['distance_km'].mean():.2f} км")
    
    print(f"\nОБНАРУЖЕННЫЕ АНОМАЛИИ:")
    
    # Поездки >12 часов
    long_rides = df.filter(pl.col("duration_minutes") > 12 * 60)
    print(f"• Поездки >12 часов: {len(long_rides):,}")
    
    # Поездки с аном скоростью
    fast_rides = df.filter(pl.col("speed_kmh") > 25)
    print(f"• Поездки >25 км/ч: {len(fast_rides):,}")
    
    # Поездки на месте
    same_location = df.filter(pl.col("same_location_long_time") == True)
    print(f"• Долгие поездки на месте (>5 мин, <10м): {len(same_location):,}")



    
    print(f"\nВРЕМЕННОЙ АНАЛИЗ:")
    
    monthly = df.group_by("start_month").agg([
        pl.len().alias("rides"),
        (pl.len() / total_rides * 100).alias("percentage"),
        pl.col("duration_minutes").mean().alias("avg_duration"),
        (pl.col("member_casual") == "casual").mean().alias("casual_pct")
    ]).sort("start_month")
    
    print("АКТИВНОСТЬ ПО МЕСЯЦАМ:")
    for row in monthly.rows():
        month_name = ["Янв", "Фев", "Мар", "Апр", "Май", "Июн", 
                     "Июл", "Авг", "Сен", "Окт", "Ноя", "Дек"][int(row[0])-1]
        print(f"  {month_name}: {row[1]:,} поездок ({row[2]:.1f}%), "
              f"ср. время: {row[3]:.1f} мин, casual: {row[4]*100:.1f}%")
    
    print(f"\nАКТИВНОСТЬ ПО ЧАСАМ:")
    hourly = df.group_by("start_hour").agg([
        pl.len().alias("rides")
    ]).sort("start_hour")
    
    peak_hour = hourly.filter(pl.col("rides") == hourly["rides"].max())
    print(f"  Пиковый час: {peak_hour['start_hour'][0]}:00 - {peak_hour['rides'][0]:,} поездок")
    
    print(f"\nРАСПРЕДЕЛЕНИЕ ПО РАССТОЯНИЯМ:")
    distance_stats = df.select([
        pl.col("distance_km").min().alias("min"),
        pl.col("distance_km").max().alias("max"),
        pl.col("distance_km").mean().alias("mean"),
        pl.col("distance_km").median().alias("median"),
        pl.col("distance_km").std().alias("std")
    ])
    
    print(f"  Минимальное: {distance_stats['min'][0]:.3f} км")
    print(f"  Максимальное: {distance_stats['max'][0]:.2f} км")
    print(f"  Среднее: {distance_stats['mean'][0]:.2f} км")
    print(f"  Медиана: {distance_stats['median'][0]:.2f} км")
    
    # Топ станций
    print(f"\nТОП-5 САМЫХ ПОПУЛЯРНЫХ СТАНЦИЙ:")
    top_start_stations = df.filter(pl.col("start_station_name") != "unknown") \
                          .group_by("start_station_name") \
                          .agg(pl.len().alias("count")) \
                          .sort("count", descending=True) \
                          .head(5)
    
    for i, row in enumerate(top_start_stations.rows(), 1):
        print(f"  {i}. {row[0]}: {row[1]:,} поездок")
    
create_final_report(df_final)
````

## Проблемы в исходных данных и их решение

### Проблема 1: Пропущенные данные о станциях
**Проблема:** 14.83% поездок (711,842) не имели информации о станциях начала или окончания.

**Решение:** Создан флаг `has_missing_stations` для отслеживания таких поездок, пропущенные значения заменены на "unknown" для сохранения целостности данных.

```python
# Код обработки пропущенных станций
df = df.with_columns([
    pl.when(pl.col('start_station_name').is_null())
      .then(pl.lit('unknown'))
      .otherwise(pl.col('start_station_name'))
      .alias('start_station_name'),
    
    ((pl.col('start_station_name') == 'unknown') | 
     (pl.col('end_station_name') == 'unknown'))
    .alias('has_missing_stations')
])
```

**Результат:** Выявлено 711,842 поездок с пропущенными станциями для дальнейшего мониторинга.

### Проблема 2: Аномально долгие поездки
**Проблема:** 2,920 поездок длительностью более 12 часов.

**Решение:** Установлен лимит в 24 часа как максимально допустимое время аренды.

```python
# Код фильтрации аномально долгих поездок
df = df.filter(pl.col('duration_minutes') < 24 * 60)
```

**Результат:** Все поездки ограничены физически возможными временными рамками.

### Проблема 3: Аномальные скорости

### Проблема 4: Поездки на месте
