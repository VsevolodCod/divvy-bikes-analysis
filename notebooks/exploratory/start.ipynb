```python
import requests
import json

gbfs_discovery_url = "https://gbfs.lyft.com/gbfs/2.3/chi/gbfs.json"

response = requests.get(gbfs_discovery_url)
gbfs_data = response.json()

feeds = gbfs_data['data']['en']['feeds']
feeds_by_name = {feed['name']: feed['url'] for feed in feeds}
station_info_url = feeds_by_name['station_information']
station_info_response = requests.get(station_info_url)
station_info_data = station_info_response.json()

# Выведем названия первых 5 станций
stations = station_info_data['data']['stations']
for i, station in enumerate(stations[:5]):
    print(f"{i+1}. {station['name']}")

print("\n---\n")

station_status_url = feeds_by_name['station_status']
station_status_response = requests.get(station_status_url)
station_status_data = station_status_response.json()

status_by_id = {station['station_id']: station for station in station_status_data['data']['stations']}

# Выведем статус для первых 5 станций
for i, station in enumerate(stations[:5]):
    station_id = station['station_id']
    status = status_by_id.get(station_id)
    if status:
        print(f"{station['name']}: {status['num_bikes_available']} велосипедов доступно, {status['num_docks_available']} доков свободно.")
```

```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('202501-divvy-tripdata.csv')

df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
df['start_hour'] = df['started_at'].dt.hour
df['start_dayofweek'] = df['started_at'].dt.dayofweek
df['start_month'] = df['started_at'].dt.month
df['trip_duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds()

def fill_stations_simple(df):
    def fill_single_column(df, target_col, use_start_station=False):
        if df[target_col].isnull().sum() == 0:
            return df
        features = ['rideable_type', 'start_lat', 'start_lng', 
                   'start_hour', 'start_dayofweek', 'start_month',
                   'member_casual', 'trip_duration']
        if use_start_station:
            features.append('start_station_name')
        train_idx = df[target_col].notna()
        predict_idx = df[target_col].isna()
        
        X_train = df.loc[train_idx, features].copy()
        y_train = df.loc[train_idx, target_col]
        X_predict = df.loc[predict_idx, features].copy()

        for col in features:
            if X_train[col].isnull().any():
                if X_train[col].dtype == 'object':
                    fill_value = X_train[col].mode()[0]
                else:
                    fill_value = X_train[col].median()
                X_train[col] = X_train[col].fillna(fill_value)
                X_predict[col] = X_predict[col].fillna(fill_value)

        cat_features = [f for f in features if df[f].dtype == 'object']

        le = LabelEncoder()
        y_train_encoded = le.fit_transform(y_train)

        model = CatBoostClassifier(
            iterations=500,
            depth=6,
            learning_rate=0.05,
            cat_features=cat_features,
            verbose=False,
            random_state=42
        )

        train_pool = Pool(X_train, y_train_encoded, cat_features=cat_features)
        model.fit(train_pool)

        predict_pool = Pool(X_predict, cat_features=cat_features)
        predictions = le.inverse_transform(model.predict(predict_pool).reshape(-1))

        df.loc[predict_idx, target_col] = predictions
        return df
 
    print("Заполняем start_station_name...")
    df = fill_single_column(df, 'start_station_name', use_start_station=False)

    print("Заполняем end_station_name...")
    df = fill_single_column(df, 'end_station_name', use_start_station=True)
    return df

# Заполняем данные
df_filled = fill_stations_simple(df)

# Результаты
print("\nРезультаты:")
print(f"start_station_name пропусков: {df_filled['start_station_name'].isnull().sum()}")
print(f"end_station_name пропусков: {df_filled['end_station_name'].isnull().sum()}")

# Сохраняем
df_filled.to_csv('filled_stations.csv', index=False)
print("Данные сохранены в 'filled_stations.csv'")
```
