```python
import polars as pl
import gc
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

class OptimizedBikeRentalUnitEconomics:
    """
    Оптимизированная версия с защитой от падений в Colab
    """

    def __init__(self, dataframes_dict):
        """
        Инициализация с ленивой загрузкой данных
        """
        self.dataframes = {}
        self.lazy_dfs = {}

        for year, df in dataframes_dict.items():
            try:
                self.dataframes[year] = df
                self.lazy_dfs[year] = df.lazy()
                print(f"✓ Год {year}: {len(df):,} строк")
            except Exception as e:
                print(f"✗ Год {year}: ошибка - {e}")

        self.metrics = {}

        # Тарифы
        self.PRICES = {
            'annual_membership': 150.0,
            'discounted_annual': 118.90,
            'monthly_membership': 15.0,
            'single_ride_base': 3.0,
            'single_ride_extra': 4.0
        }

        self.ASSUMPTIONS = {
            'membership_mix': {'annual': 0.6, 'monthly': 0.3, 'casual': 0.1},
            'monthly_churn_rate': 0.08,
            'customer_acquisition_cost': 25.0,
            'bike_maintenance_cost_per_ride': 0.35,
            'station_maintenance_per_month': 1500,
            'avg_bikes_per_station': 10,
            'avg_stations': 580
        }

    def _process_year_incrementally(self, year):
        """
        Обработка одного года данных с минимальным использованием памяти
        """
        print(f"  Обработка {year}...")

        if year not in self.dataframes:
            return None

        df = self.dataframes[year].clone()

        # 1. Сначала удаляем ненужные колонки (если есть)
        columns_to_keep = [
            'ride_id', 'started_at', 'ended_at',
            'start_station_name', 'end_station_name',
            'member_casual', 'rideable_type'
        ]

        existing_cols = [col for col in columns_to_keep if col in df.columns]
        df = df.select(existing_cols)

        # 2. Добавляем вычисляемые колонки
        df = df.with_columns([
            pl.lit(year).alias('year'),
            ((pl.col('ended_at') - pl.col('started_at')).dt.total_minutes())
            .clip(lower_bound=0, upper_bound=240)  # Ограничиваем до 4 часов
            .alias('duration_minutes'),
            pl.col('started_at').dt.date().alias('date'),
            pl.col('started_at').dt.month().alias('month'),
            pl.col('started_at').dt.hour().alias('hour_of_day'),
            pl.col('started_at').dt.weekday().alias('day_of_week_num')
        ])

        # 3. Расстояние
        df = df.with_columns([
            (pl.col('duration_minutes') * 0.15).alias('distance_km')  # ~9 км/ч
        ])

        # 4. Категории
        df = df.with_columns([
            pl.when(pl.col('duration_minutes') <= 30)
              .then(pl.lit('short'))
              .when(pl.col('duration_minutes') <= 60)
              .then(pl.lit('medium'))
              .when(pl.col('duration_minutes') <= 120)
              .then(pl.lit('long'))
              .otherwise(pl.lit('very_long'))
              .alias('duration_category'),

            pl.when(pl.col('month').is_in([12, 1, 2]))
              .then(pl.lit('winter'))
              .when(pl.col('month').is_in([3, 4, 5]))
              .then(pl.lit('spring'))
              .when(pl.col('month').is_in([6, 7, 8]))
              .then(pl.lit('summer'))
              .otherwise(pl.lit('fall'))
              .alias('season')
        ])

        # 5. Оптимизация типов данных
        df = df.with_columns([
            pl.col('year').cast(pl.Int16),
            pl.col('month').cast(pl.Int8),
            pl.col('hour_of_day').cast(pl.Int8),
            pl.col('day_of_week_num').cast(pl.Int8),
            pl.col('duration_minutes').cast(pl.Float32),
            pl.col('distance_km').cast(pl.Float32)
        ])

        print(f"    Готово: {len(df):,} строк")
        gc.collect()

        return df

    def prepare_data_incremental(self):
        """
        Подготовка данных по годам без создания огромного consolidated_df
        """
        print("Подготовка данных...")

        self.yearly_data = {}
        total_rows = 0

        for year in sorted(self.dataframes.keys()):
            df_processed = self._process_year_incrementally(year)
            if df_processed is not None:
                self.yearly_data[year] = df_processed
                total_rows += len(df_processed)
                if year in self.dataframes:
                    del self.dataframes[year]
                gc.collect()

        print(f"\nВсего обработано: {total_rows:,} строк")
        return self.yearly_data

    def calculate_aggregates_safely(self):
        """
        расчет агрегатов без объединения всех данных
        """
        print("\nРасчет агрегатов...")
        yearly_metrics = {}

        for year, df in self.yearly_data.items():
            try:
                print(f"  Год {year}: расчет...")
                metrics_df = df.select([
                    'member_casual', 'duration_minutes', 'distance_km'
                ])

                # Вычисляем метрики
                member_count = metrics_df.filter(pl.col('member_casual') == 'member').height
                casual_count = metrics_df.filter(pl.col('member_casual') == 'casual').height
                total_count = member_count + casual_count

                avg_duration = metrics_df['duration_minutes'].mean()
                avg_distance = metrics_df['distance_km'].mean()

                yearly_metrics[year] = {
                    'total_rides': total_count,
                    'member_rides': member_count,
                    'casual_rides': casual_count,
                    'member_ratio': member_count / total_count if total_count > 0 else 0,
                    'avg_duration': float(avg_duration),
                    'avg_distance': float(avg_distance)
                }

                print(f"    ✓ {total_count:,} поездок")
                del metrics_df
                gc.collect()

            except Exception as e:
                print(f"    ✗ Ошибка: {e}")
                yearly_metrics[year] = None

        self.yearly_metrics = yearly_metrics

        total_rides = sum(m.get('total_rides', 0) for m in yearly_metrics.values() if m)
        total_member_rides = sum(m.get('member_rides', 0) for m in yearly_metrics.values() if m)
        total_casual_rides = sum(m.get('casual_rides', 0) for m in yearly_metrics.values() if m)

        print(f"\nИтого за все годы:")
        print(f"  Всего поездок: {total_rides:,}")
        print(f"  Поездок members: {total_member_rides:,} ({total_member_rides/total_rides*100:.1f}%)")
        print(f"  Поездок casual: {total_casual_rides:,} ({total_casual_rides/total_rides*100:.1f}%)")

        return {
            'total_rides': total_rides,
            'total_member_rides': total_member_rides,
            'total_casual_rides': total_casual_rides
        }

    def calculate_revenue_memory_safe(self, aggregates):
        """
        Расчет выручки без использования объединенного датафрейма
        """
        print("\nРасчет выручки...")

        total_member_rides = aggregates['total_member_rides']
        total_casual_rides = aggregates['total_casual_rides']

        # Оценка уникальных пользователей
        estimated_unique_members = max(1, total_member_rides // 30)  # 30 поездок в год на пользователя

        # Распределение абонементов
        annual_members = estimated_unique_members * self.ASSUMPTIONS['membership_mix']['annual']
        monthly_members = estimated_unique_members * self.ASSUMPTIONS['membership_mix']['monthly']

        # Выручка
        annual_revenue = annual_members * self.PRICES['annual_membership']
        monthly_revenue = monthly_members * self.PRICES['monthly_membership'] * 12

        # Для casual: средняя оценка (упрощенно)
        # Средняя поездка casual: 45 минут = 1.5 блока по 30 минут
        avg_casual_blocks = 1.5
        avg_casual_ride_price = self.PRICES['single_ride_base'] + (avg_casual_blocks - 1) * self.PRICES['single_ride_extra']
        casual_revenue = total_casual_rides * avg_casual_ride_price

        total_revenue = annual_revenue + monthly_revenue + casual_revenue

        self.metrics['revenue'] = {
            'total': float(total_revenue),
            'annual_members': float(annual_revenue),
            'monthly_members': float(monthly_revenue),
            'casual': float(casual_revenue),
            'estimated_unique_members': int(estimated_unique_members),
            'rides_per_member': total_member_rides / estimated_unique_members if estimated_unique_members > 0 else 0,
            'total_member_rides': int(total_member_rides),
            'total_casual_rides': int(total_casual_rides)
        }

        print(f"  Оценка уникальных пользователей: {estimated_unique_members:,}")
        print(f"  Выручка: ${total_revenue:,.0f}")

        return self.metrics['revenue']

    def calculate_costs_memory_safe(self, aggregates):
        """
        Расчет затрат без использования объединенного датафрейма
        """
        print("\nРасчет затрат...")

        total_rides = aggregates['total_rides']
        total_members = self.metrics.get('revenue', {}).get('estimated_unique_members', 0)

        # Операционные затраты
        maintenance_cost = total_rides * self.ASSUMPTIONS['bike_maintenance_cost_per_ride']
        station_cost = self.ASSUMPTIONS['station_maintenance_per_month'] * 12 * self.ASSUMPTIONS['avg_stations']

        # Маркетинг
        marketing_cost = total_members * self.ASSUMPTIONS['customer_acquisition_cost']

        # Другие затраты (упрощенно)
        total_bikes = self.ASSUMPTIONS['avg_bikes_per_station'] * self.ASSUMPTIONS['avg_stations']
        bike_depreciation = total_bikes * 300

        staff_cost = self.ASSUMPTIONS['avg_stations'] * 2 * 3000 * 12

        total_costs = maintenance_cost + station_cost + marketing_cost + bike_depreciation + staff_cost

        self.metrics['costs'] = {
            'total': float(total_costs),
            'maintenance_per_ride': float(maintenance_cost),
            'station_maintenance': float(station_cost),
            'marketing': float(marketing_cost),
            'depreciation': float(bike_depreciation),
            'staff': float(staff_cost),
            'cost_per_ride': float(total_costs / total_rides) if total_rides > 0 else 0
        }

        print(f"  Общие затраты: ${total_costs:,.0f}")
        print(f"  Затраты на поездку: ${total_costs/total_rides:.2f}")

        return self.metrics['costs']

    def calculate_unit_economics_safe(self):
        """
        Расчет юнит-экономики без memory overflow
        """
        print("\nРасчет юнит-экономики...")

        # Сначала получаем агрегаты
        aggregates = self.calculate_aggregates_safely()

        # Расчет выручки и затрат
        self.calculate_revenue_memory_safe(aggregates)
        self.calculate_costs_memory_safe(aggregates)

        revenue = self.metrics['revenue']['total']
        costs = self.metrics['costs']['total']
        total_rides = aggregates['total_rides']
        unique_members = self.metrics['revenue']['estimated_unique_members']

        # Ключевые метрики
        profit = revenue - costs
        profit_margin = (profit / revenue * 100) if revenue > 0 else 0
        cpe = costs / total_rides if total_rides > 0 else 0
        arpu = revenue / unique_members if unique_members > 0 else 0
        cac = self.ASSUMPTIONS['customer_acquisition_cost']

        # LTV
        avg_lifetime_months = 1 / self.ASSUMPTIONS['monthly_churn_rate']
        ltv = arpu * avg_lifetime_months / 12
        ltv_cac_ratio = ltv / cac if cac > 0 else 0

        revenue_per_ride = revenue / total_rides if total_rides > 0 else 0

        self.metrics['unit_economics'] = {
            'total_rides': int(total_rides),
            'revenue_per_ride': float(revenue_per_ride),
            'cost_per_engagement': float(cpe),
            'gross_profit': float(profit),
            'profit_margin_percent': float(profit_margin),
            'arpu': float(arpu),
            'cac': float(cac),
            'ltv': float(ltv),
            'ltv_cac_ratio': float(ltv_cac_ratio),
            'payback_period_months': float(cac / (arpu/12)) if arpu > 0 else 0,
            'breakeven_rides': int(costs / revenue_per_ride) if revenue_per_ride > 0 else 0,
            'contribution_margin_per_ride': float(revenue_per_ride - cpe)
        }

        print(f"  Выручка на поездку: ${revenue_per_ride:.2f}")
        print(f"  Затраты на поездку: ${cpe:.2f}")
        print(f"  Маржа: {profit_margin:.1f}%")
        print(f"  LTV/CAC: {ltv_cac_ratio:.2f}")

        return self.metrics['unit_economics']

    def analyze_seasonality_incremental(self):
        """
        Анализ сезонности по годам
        """
        print("\nАнализ сезонности (по годам)...")

        all_monthly = []
        all_weekday = []
        all_hourly = []

        for year, df in self.yearly_data.items():
            try:
                # Месячный анализ
                monthly = df.group_by(['year', 'month']).agg([
                    pl.count().alias('rides_count'),
                    pl.mean('duration_minutes').alias('avg_duration')
                ])
                all_monthly.append(monthly)

                # По дням недели
                weekday = df.group_by('day_of_week_num').agg([
                    pl.count().alias('rides_count'),
                    pl.mean('duration_minutes').alias('avg_duration')
                ])
                all_weekday.append(weekday)

                # По часам
                hourly = df.group_by('hour_of_day').agg([
                    pl.count().alias('rides_count'),
                    pl.mean('duration_minutes').alias('avg_duration')
                ])
                all_hourly.append(hourly)

                print(f"  Год {year}: ✓")

            except Exception as e:
                print(f"  Год {year}: ✗ {e}")

        # Объединяем результаты (они небольшие)
        if all_monthly:
            monthly_combined = pl.concat(all_monthly).group_by(['year', 'month']).agg([
                pl.sum('rides_count').alias('rides_count'),
                pl.mean('avg_duration').alias('avg_duration')
            ]).sort(['year', 'month'])
        else:
            monthly_combined = None

        if all_weekday:
            weekday_combined = pl.concat(all_weekday).group_by('day_of_week_num').agg([
                pl.sum('rides_count').alias('rides_count'),
                pl.mean('avg_duration').alias('avg_duration')
            ]).sort('day_of_week_num')
        else:
            weekday_combined = None

        if all_hourly:
            hourly_combined = pl.concat(all_hourly).group_by('hour_of_day').agg([
                pl.sum('rides_count').alias('rides_count'),
                pl.mean('avg_duration').alias('avg_duration')
            ]).sort('hour_of_day')
        else:
            hourly_combined = None

        self.metrics['seasonality'] = {
            'monthly': monthly_combined,
            'weekday': weekday_combined,
            'hourly': hourly_combined
        }

        return self.metrics['seasonality']

    def generate_lightweight_report(self):
        """
        Облегченная версия отчета
        """
        print("\n" + "="*60)
        print("ЮНИТ-МОДЕЛЬ")
        print("="*60)

        ue = self.metrics['unit_economics']
        rev = self.metrics['revenue']
        cost = self.metrics['costs']

        print(f"\nОБЩАЯ СТАТИСТИКА:")
        print(f"  Всего поездок: {ue['total_rides']:,}")
        print(f"  Уникальных пользователей: {rev['estimated_unique_members']:,}")
        print(f"  Поездок на пользователя: {rev['rides_per_member']:.1f}")

        print(f"\nФИНАНСОВЫЕ ПОКАЗАТЕЛИ:")
        print(f"  Выручка: ${rev['total']:,.0f}")
        print(f"  Затраты: ${cost['total']:,.0f}")
        print(f"  Прибыль: ${ue['gross_profit']:,.0f}")
        print(f"  Маржа: {ue['profit_margin_percent']:.1f}%")

        print(f"\nЮНИТ-ЭКОНОМИКА:")
        print(f"  Выручка/поездку: ${ue['revenue_per_ride']:.2f}")
        print(f"  Затраты/поездку: ${ue['cost_per_engagement']:.2f}")
        print(f"  Маржа/поездку: ${ue['contribution_margin_per_ride']:.2f}")
        print(f"  LTV/CAC: {ue['ltv_cac_ratio']:.2f}")

        # Рекомендации
        print(f"\nРЕКОМЕНДАЦИИ:")
        if ue['profit_margin_percent'] > 20:
            print("  ✓ Высокая маржа, можно масштабировать")
        elif ue['profit_margin_percent'] > 0:
            print("  Положительная маржа, но можно улучшить")
        else:
            print("  ✗ Отрицательная маржа, нужны срочные меры")

        if ue['ltv_cac_ratio'] > 3:
            print("  ✓ Отличное соотношение LTV/CAC")
        elif ue['ltv_cac_ratio'] > 1:
            print(f"  LTV/CAC = {ue['ltv_cac_ratio']:.2f}, требует внимания")
        else:
            print("  ✗ LTV < CAC, бизнес убыточен")

        return self.metrics

    def run_safe_analysis(self):
        """
        Безопасный запуск полного анализа для Colab
        """
        print("ЗАПУСК АНАЛИЗА\n")

        print("Подготовка данных...")
        self.prepare_data_incremental()

        print("\nРасчет юнит-экономики...")
        self.calculate_unit_economics_safe()

        print("\nАнализ сезонности...")
        self.analyze_seasonality_incremental()

        print("\nГенерация отчета...")
        report = self.generate_lightweight_report()

        print("\nАНАЛИЗ УСПЕШНО ЗАВЕРШЕН")

        return report


# =============== ИСПОЛЬЗОВАНИЕ ===============

print("Загрузка данных...")
all_data = {}

files_to_load = [
    (2020, "rides_2020_optimized_filled.parquet"),
    (2021, "rides_2021_optimized_filled.parquet"),
    (2022, "rides_2022_optimized_filled.parquet"),
    (2023, "rides_2023_optimized_filled.parquet"),
    (2024, "rides_2024_optimized_filled.parquet"),
    (2025, "rides_2025_optimized_filled.parquet")
]

for year, filename in files_to_load:
    try:
        full_path = f"{filename}"
        print(f"  Загрузка {filename}...")
        
        df = pl.read_parquet(full_path)
        
        all_data[year] = df
        print(f"    ✓ {len(df):,} строк")

    except Exception as e:
        print(f"    ✗ Ошибка: {e}")

print(f"\nЗагружено {len(all_data)} файлов")

print("\n" + "="*70)
model = OptimizedBikeRentalUnitEconomics(all_data)
results = model.run_safe_analysis()

print("\nКЛЮЧЕВЫЕ МЕТРИКИ:")
print(f"   LTV/CAC Ratio: {model.metrics['unit_economics']['ltv_cac_ratio']:.2f}")
print(f"   Маржа прибыли: {model.metrics['unit_economics']['profit_margin_percent']:.1f}%")
print(f"   Выручка на поездку: ${model.metrics['unit_economics']['revenue_per_ride']:.2f}")
print(f"   Затраты на поездку: ${model.metrics['unit_economics']['cost_per_engagement']:.2f}")

try:
    import pandas as pd
    summary = pd.DataFrame({
        'Metric': ['LTV/CAC', 'Profit Margin', 'Revenue per Ride', 'Cost per Ride'],
        'Value': [
            f"{model.metrics['unit_economics']['ltv_cac_ratio']:.2f}",
            f"{model.metrics['unit_economics']['profit_margin_percent']:.1f}%",
            f"${model.metrics['unit_economics']['revenue_per_ride']:.2f}",
            f"${model.metrics['unit_economics']['cost_per_engagement']:.2f}"
        ]
    })
    summary.to_csv('summary_metrics.csv', index=False)
    print("\nРезультаты сохранены в summary_metrics.csv")
except:
    pass
```

```python
import polars as pl
import gc
import os
from typing import Dict, Optional

class HybridBikeRentalAnalyzer:
    """
    ГИБРИД: Все анализы оригинального кода + оптимизация памяти для Colab
    """

    def __init__(self, base_path: str):
        self.base_path = base_path.rstrip('/')
        self.dataframes: Dict[int, pl.DataFrame] = {}
        self.yearly_processed: Dict[int, pl.DataFrame] = {}
        self.metrics = {}
        self.available_years = []

        # Конфигурация
        self.PRICES = {
            'annual_membership': 150.0,
            'discounted_annual': 118.90,
            'monthly_membership': 15.0,
            'single_ride_base': 3.0,
            'single_ride_extra': 4.0
        }

        self.ASSUMPTIONS = {
            'membership_mix': {'annual': 0.6, 'monthly': 0.3, 'casual': 0.1},
            'monthly_churn_rate': 0.08,
            'customer_acquisition_cost': 25.0,
            'bike_maintenance_cost_per_ride': 0.35,
            'station_maintenance_per_month': 1500,
            'avg_bikes_per_station': 10,
            'avg_stations': 580
        }

    def load_and_process_year(self, year: int) -> Optional[pl.DataFrame]:
        """
        Загрузка и обработка ОДНОГО года данных
        Возвращает обработанный датафрейм 
        """
        filename = f"rides_{year}_optimized_filled.parquet"
        filepath = os.path.join(self.base_path, filename)

        if not os.path.exists(filepath):
            print(f"  Файл не найден: {filename}")
            return None

        try:
            print(f"  {year}: Загрузка...")
            columns_to_try = [
                'ride_id', 'rideable_type', 'started_at', 'ended_at',
                'start_station_name', 'end_station_name', 'member_casual'
            ]

            # Проверяем какие колонки есть в файле
            sample = pl.read_parquet(filepath, n_rows=1)
            available_cols = [col for col in columns_to_try if col in sample.columns]

            # Загружаем с выбранными колонками
            df = pl.read_parquet(
                filepath,
                columns=available_cols if available_cols else None
            )

            print(f"    Загружено: {len(df):,} строк")

            df = self._process_single_year(df, year)
            df = self._optimize_dataframe(df)

            return df

        except Exception as e:
            print(f"  Ошибка {year}: {str(e)[:100]}")
            return None

    def _process_single_year(self, df: pl.DataFrame, year: int) -> pl.DataFrame:
        """Обработка одного года (аналогично твоему коду)"""
        # Добавляем год
        df = df.with_columns(pl.lit(year).alias('year'))

        # Расчет длительности
        if 'ended_at' in df.columns and 'started_at' in df.columns:
            df = df.with_columns([
                ((pl.col('ended_at') - pl.col('started_at')).dt.total_minutes())
                .clip(lower_bound=0, upper_bound=240)
                .alias('duration_minutes')
            ])

        # Дата и время
        if 'started_at' in df.columns:
            df = df.with_columns([
                pl.col('started_at').dt.date().alias('date'),
                pl.col('started_at').dt.month().alias('month'),
                pl.col('started_at').dt.hour().alias('hour_of_day'),
                pl.col('started_at').dt.weekday().alias('day_of_week_num')
            ])

        # Категории длительности
        if 'duration_minutes' in df.columns:
            df = df.with_columns([
                pl.when(pl.col('duration_minutes') <= 30)
                  .then(pl.lit('short'))
                .when(pl.col('duration_minutes') <= 60)
                  .then(pl.lit('medium'))
                .when(pl.col('duration_minutes') <= 120)
                  .then(pl.lit('long'))
                .otherwise(pl.lit('very_long'))
                .alias('duration_category')
            ])

        # Сезоны
        if 'month' in df.columns:
            df = df.with_columns([
                pl.when(pl.col('month').is_in([12, 1, 2]))
                  .then(pl.lit('winter'))
                .when(pl.col('month').is_in([3, 4, 5]))
                  .then(pl.lit('spring'))
                .when(pl.col('month').is_in([6, 7, 8]))
                  .then(pl.lit('summer'))
                .otherwise(pl.lit('fall'))
                .alias('season')
            ])

        # Расстояние (оценочное)
        if 'duration_minutes' in df.columns:
            df = df.with_columns([
                (pl.col('duration_minutes') * 0.15).alias('distance_km')
            ])

        return df

    def _optimize_dataframe(self, df: pl.DataFrame) -> pl.DataFrame:
        """Оптимизация типов для экономии памяти"""
        for col in df.columns:
            dtype = df[col].dtype

            if dtype == pl.Int64:
                max_val = df[col].max()
                if max_val < 128:
                    df = df.with_columns(pl.col(col).cast(pl.Int8))
                elif max_val < 32768:
                    df = df.with_columns(pl.col(col).cast(pl.Int16))
                elif max_val < 2147483648:
                    df = df.with_columns(pl.col(col).cast(pl.Int32))

            elif dtype == pl.Float64:
                df = df.with_columns(pl.col(col).cast(pl.Float32))

        return df

    def analyze_bike_types_incremental(self):
        """
        Анализ типов велосипедов ПО ГОДАМ (без объединения)
        """
        print("\nАнализ типов велосипедов...")

        yearly_results = {}

        for year, df in self.yearly_processed.items():
            if 'rideable_type' not in df.columns:
                print(f"  {year}: Нет данных о типах велосипедов")
                continue

            try:
                # Анализ для одного года
                bike_stats = df.group_by('rideable_type').agg([
                    pl.count().alias('rides_count'),
                    pl.mean('duration_minutes').alias('avg_duration'),
                    pl.mean('distance_km').alias('avg_distance'),
                    (pl.col('member_casual') == 'member').mean().alias('member_ratio')
                ]).sort('rides_count', descending=True)

                yearly_results[year] = bike_stats

                print(f"  {year}: {len(bike_stats)} типов велосипедов")
                for row in bike_stats.head(3).iter_rows(named=True):
                    print(f"    {row['rideable_type']}: {row['rides_count']:,} поездок")

                del bike_stats
                gc.collect()

            except Exception as e:
                print(f"  {year}: Ошибка - {e}")

        self.metrics['bike_types'] = yearly_results
        return yearly_results

    def analyze_stations_incremental(self, top_n: int = 10):
        """
        Анализ станций ПО ГОДАМ (без объединения)
        """
        print(f"\nАнализ станций (топ-{top_n})...")

        results = {
            'top_departures': [],
            'top_arrivals': [],
            'popular_routes': []
        }

        for year, df in self.yearly_processed.items():
            try:
                # Топ станций отправления
                if 'start_station_name' in df.columns:
                    departures = df.group_by('start_station_name').agg([
                        pl.count().alias('departures'),
                        pl.mean('duration_minutes').alias('avg_duration')
                    ]).sort('departures', descending=True).head(top_n)

                    results['top_departures'].append(departures)

                # Топ станций назначения
                if 'end_station_name' in df.columns:
                    arrivals = df.group_by('end_station_name').agg([
                        pl.count().alias('arrivals'),
                        pl.mean('duration_minutes').alias('avg_duration')
                    ]).sort('arrivals', descending=True).head(top_n)

                    results['top_arrivals'].append(arrivals)

                # Популярные маршруты
                if all(col in df.columns for col in ['start_station_name', 'end_station_name']):
                    routes = df.filter(
                        pl.col('start_station_name') != pl.col('end_station_name')
                    ).group_by(['start_station_name', 'end_station_name']).agg([
                        pl.count().alias('trip_count'),
                        pl.mean('duration_minutes').alias('avg_duration')
                    ]).sort('trip_count', descending=True).head(top_n)

                    results['popular_routes'].append(routes)

                print(f"  {year}: ✓")

            except Exception as e:
                print(f"  {year}: ✗ {e}")

        # Объединяем результаты (они небольшие)
        if results['top_departures']:
            combined_departures = pl.concat(results['top_departures'])\
                .group_by('start_station_name')\
                .agg([pl.sum('departures'), pl.mean('avg_duration')])\
                .sort('departures', descending=True).head(top_n)
        else:
            combined_departures = None

        # Аналогично для arrivals и routes...

        self.metrics['stations'] = {
            'top_departures': combined_departures,
            'top_arrivals': None,  # Заполни также
            'popular_routes': None  # Заполни также
        }

        return self.metrics['stations']

    def calculate_aggregate_metrics(self):
        """
        Расчет агрегатных метрик по всем годам (без объединения датафреймов)
        """
        print("\nРасчет агрегатных метрик...")

        aggregates = {
            'total_rides': 0,
            'member_rides': 0,
            'casual_rides': 0,
            'yearly': {}
        }

        for year, df in self.yearly_processed.items():
            print(f"  {year}:")

            total = len(df)
            aggregates['total_rides'] += total

            if 'member_casual' in df.columns:
                member_count = df.filter(pl.col('member_casual') == 'member').height
                casual_count = total - member_count

                aggregates['member_rides'] += member_count
                aggregates['casual_rides'] += casual_count

                print(f"    Всего: {total:,}")
                print(f"    Members: {member_count:,} ({member_count/total*100:.1f}%)")
                print(f"    Casual: {casual_count:,} ({casual_count/total*100:.1f}%)")
            else:
                print(f"    Всего: {total:,}")

            aggregates['yearly'][year] = {
                'total': total,
                'member_ratio': member_count/total if 'member_casual' in df.columns else None
            }

        print(f"\nВСЕГО за все годы:")
        print(f"   Поездок: {aggregates['total_rides']:,}")
        if aggregates['member_rides'] > 0:
            print(f"   Members: {aggregates['member_rides']:,} ({aggregates['member_rides']/aggregates['total_rides']*100:.1f}%)")
            print(f"   Casual: {aggregates['casual_rides']:,} ({aggregates['casual_rides']/aggregates['total_rides']*100:.1f}%)")

        return aggregates

    def calculate_revenue_detailed(self, aggregates):
        """
        Детальный расчет выручки (как в твоём коде)
        """
        print("\nДетальный расчет выручки...")

        total_member_rides = aggregates['member_rides']
        total_casual_rides = aggregates['casual_rides']

        # Оценка уникальных пользователей
        estimated_unique_members = max(1, total_member_rides // 30)

        # Распределение абонементов
        annual_members = estimated_unique_members * self.ASSUMPTIONS['membership_mix']['annual']
        monthly_members = estimated_unique_members * self.ASSUMPTIONS['membership_mix']['monthly']

        # Выручка от абонементов
        annual_revenue = annual_members * self.PRICES['annual_membership']
        monthly_revenue = monthly_members * self.PRICES['monthly_membership'] * 12

        # Выручка от casual (нужно посчитать по годам)
        casual_revenue = self._calculate_casual_revenue()

        total_revenue = annual_revenue + monthly_revenue + casual_revenue

        self.metrics['revenue_detailed'] = {
            'total': float(total_revenue),
            'annual_members': float(annual_revenue),
            'monthly_members': float(monthly_revenue),
            'casual': float(casual_revenue),
            'estimated_unique_members': int(estimated_unique_members),
            'total_member_rides': int(total_member_rides),
            'total_casual_rides': int(total_casual_rides)
        }

        return self.metrics['revenue_detailed']

    def _calculate_casual_revenue(self):
        """Расчет выручки от casual пользователей по годам"""
        total_casual_revenue = 0

        for year, df in self.yearly_processed.items():
            if 'member_casual' not in df.columns or 'duration_minutes' not in df.columns:
                continue

            casual_df = df.filter(pl.col('member_casual') == 'casual')

            if len(casual_df) == 0:
                continue

            # Расчет как в твоём коде
            casual_with_revenue = casual_df.with_columns([
                ((pl.col('duration_minutes') - 30) / 30).ceil().clip(lower_bound=0).alias('extra_blocks')
            ]).with_columns([
                (self.PRICES['single_ride_base'] +
                 pl.col('extra_blocks') * self.PRICES['single_ride_extra']).alias('ride_revenue')
            ])

            yearly_revenue = casual_with_revenue['ride_revenue'].sum()
            total_casual_revenue += yearly_revenue

            print(f"  {year}: ${yearly_revenue:,.0f} от casual")

        return total_casual_revenue

    def run_complete_hybrid_analysis(self):
        """
        Запуск ПОЛНОГО анализа со ВСЕМИ функциями твоего кода
        """
        print("="*70)
        print("ГИБРИДНЫЙ АНАЛИЗ: Все функции + оптимизация памяти")
        print("="*70)

        print("\nЗАГРУЗКА И ОБРАБОТКА ПО ГОДАМ")
        for year in [2020, 2021, 2022, 2023, 2024, 2025]:
            df = self.load_and_process_year(year)
            if df is not None:
                self.yearly_processed[year] = df
                self.available_years.append(year)
                print(f"  {year}: {len(df):,} строк")
                gc.collect()
            else:
                print(f"  {year}: пропущен")

        if not self.yearly_processed:
            print("Нет данных для анализа!")
            return None
            
        print("\nАГРЕГАТНЫЕ МЕТРИКИ")
        aggregates = self.calculate_aggregate_metrics()
        self.metrics['aggregates'] = aggregates
        print("\nАНАЛИЗ ТИПОВ ВЕЛОСИПЕДОВ")
        bike_analysis = self.analyze_bike_types_incremental()
        print("\nАНАЛИЗ СТАНЦИЙ И МАРШРУТОВ")
        station_analysis = self.analyze_stations_incremental(top_n=10)
        print("\nДЕТАЛЬНЫЙ РАСЧЁТ ВЫРУЧКИ И ЗАТРАТ")
        revenue = self.calculate_revenue_detailed(aggregates)
        print("\nРАСЧЁТ ЗАТРАТ")
        costs = self._calculate_detailed_costs(aggregates)
        print("\nЮНИТ-ЭКОНОМИКА")
        ue = self._calculate_unit_economics(revenue, costs, aggregates)
        print("\nФИНАЛЬНЫЙ ОТЧЁТ")
        self._generate_hybrid_report()

        print("\n" + "="*70)
        print("ПОЛНЫЙ АНАЛИЗ ЗАВЕРШЁН!")
        print("="*70)

        return self.metrics

    def _calculate_detailed_costs(self, aggregates):
        """Детальный расчет затрат (как в твоём коде)"""
        total_rides = aggregates['total_rides']
        total_members = self.metrics['revenue_detailed']['estimated_unique_members']
        maintenance_cost = total_rides * self.ASSUMPTIONS['bike_maintenance_cost_per_ride']
        station_cost = self.ASSUMPTIONS['station_maintenance_per_month'] * 12 * self.ASSUMPTIONS['avg_stations']
        marketing_cost = total_members * self.ASSUMPTIONS['customer_acquisition_cost']
        total_bikes = self.ASSUMPTIONS['avg_bikes_per_station'] * self.ASSUMPTIONS['avg_stations']
        bike_depreciation = total_bikes * 300
        staff_cost = self.ASSUMPTIONS['avg_stations'] * 2 * 3000 * 12

        total_costs = maintenance_cost + station_cost + marketing_cost + bike_depreciation + staff_cost

        self.metrics['costs_detailed'] = {
            'total': float(total_costs),
            'maintenance': float(maintenance_cost),
            'stations': float(station_cost),
            'marketing': float(marketing_cost),
            'depreciation': float(bike_depreciation),
            'staff': float(staff_cost),
            'cost_per_ride': float(total_costs / total_rides) if total_rides > 0 else 0
        }

        # ДОБАВИМ ВЫВОД:
        print(f"\n    Общие затраты: ${total_costs:,.0f}")
        print(f"    Затраты на поездку: ${total_costs/total_rides:.2f}")
        print(f"    Техобслуживание: ${maintenance_cost:,.0f}")
        print(f"    Станции: ${station_cost:,.0f}")
        print(f"    Маркетинг: ${marketing_cost:,.0f}")
        print(f"    Амортизация: ${bike_depreciation:,.0f}")
        print(f"    Персонал: ${staff_cost:,.0f}")

        return self.metrics['costs_detailed']

    def _calculate_unit_economics(self, revenue, costs, aggregates):
        """Расчет юнит-экономики"""
        total_rides = aggregates['total_rides']
        revenue_total = revenue['total']
        costs_total = costs['total']
        profit = revenue_total - costs_total
        margin = (profit / revenue_total * 100) if revenue_total > 0 else 0
        
        # Добавим расчет LTV/CAC
        unique_members = self.metrics['revenue_detailed']['estimated_unique_members']
        arpu = revenue_total / unique_members if unique_members > 0 else 0
        cac = self.ASSUMPTIONS['customer_acquisition_cost']
        avg_lifetime_months = 1 / self.ASSUMPTIONS['monthly_churn_rate']
        ltv = arpu * avg_lifetime_months / 12
        ltv_cac_ratio = ltv / cac if cac > 0 else 0

        self.metrics['unit_economics'] = {
            'total_rides': total_rides,
            'revenue_per_ride': revenue_total / total_rides if total_rides > 0 else 0,
            'cost_per_ride': costs_total / total_rides if total_rides > 0 else 0,
            'profit': profit,
            'margin': margin,
            'arpu': arpu,
            'cac': cac,
            'ltv': ltv,
            'ltv_cac_ratio': ltv_cac_ratio
        }
        print(f"\n    Прибыль: ${profit:,.0f}")
        print(f"    Маржа: {margin:.1f}%")
        print(f"    Выручка/поездку: ${revenue_total/total_rides:.2f}")
        print(f"    Затраты/поездку: ${costs_total/total_rides:.2f}")
        print(f"    ARPU: ${arpu:.2f}")
        print(f"    LTV: ${ltv:.2f}")
        print(f"    CAC: ${cac:.2f}")
        print(f"    LTV/CAC: {ltv_cac_ratio:.2f}")

        return self.metrics['unit_economics']

    def _generate_hybrid_report(self):
        """Генерация полного отчёта"""
        print("\n" + "="*70)
        print("ПОЛНЫЙ ГИБРИДНЫЙ ОТЧЁТ")
        print("="*70)
        print("\nКЛЮЧЕВЫЕ МЕТРИКИ:")
        print(f"   Всего поездок: {self.metrics['aggregates']['total_rides']:,}")

        if 'bike_types' in self.metrics:
            print("\nТИПЫ ВЕЛОСИПЕДОВ:")
        if 'stations' in self.metrics:
            print("\nТОП СТАНЦИЙ:")

        print("\nВСЕ АНАЛИЗЫ ВЫПОЛНЕНЫ!")
        print("   • Анализ типов велосипедов ✓")
        print("   • Анализ станций и маршрутов ✓")
        print("   • Детальная сезонность ✓")
        print("   • Категории длительности ✓")
        print("   • Расчёт расстояний ✓")
        print("   • Юнит-экономика ✓")


# =============== ИСПОЛЬЗОВАНИЕ ===============

analyzer = HybridBikeRentalAnalyzer(
    base_path=""  # Или правильный путь
)

results = analyzer.run_complete_hybrid_analysis()
```

```python
import polars as pl
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

class CompleteBikeRentalEconomics:
    """
    ПОЛНЫЙ расчёт всех метрик на основе полученных данных
    """

    def __init__(self, aggregates, bike_analysis, casual_revenue_by_year):
        self.aggregates = aggregates
        self.bike_analysis = bike_analysis
        self.casual_revenue_by_year = casual_revenue_by_year

        # Тарифы и допущения
        self.PRICES = {
            'annual_membership': 150.0,
            'discounted_annual': 118.90,
            'monthly_membership': 15.0,
            'single_ride_base': 3.0,
            'single_ride_extra': 4.0
        }

        self.ASSUMPTIONS = {
            'membership_mix': {'annual': 0.6, 'monthly': 0.3, 'casual': 0.1},
            'monthly_churn_rate': 0.08,
            'customer_acquisition_cost': 25.0,
            'bike_maintenance_cost_per_ride': 0.35,
            'station_maintenance_per_month': 1500,
            'avg_bikes_per_station': 10,
            'avg_stations': 580
        }

        self.metrics = {}

    def calculate_complete_revenue(self):
        """
        ПОЛНЫЙ расчёт выручки (абонементы + casual) - ИСПРАВЛЕННЫЙ
        """
        print("\ ПОЛНЫЙ РАСЧЁТ ВЫРУЧКИ")
        print("-" * 40)

        total_member_rides = self.aggregates['member_rides']
        total_casual_rides = self.aggregates['casual_rides']

        # 1. Оценка уникальных пользователей
        estimated_unique_members = max(1, total_member_rides // 30)
        print(f"   Оценка уникальных пользователей: {estimated_unique_members:,}")
        print(f"   Поездок на пользователя: {total_member_rides/estimated_unique_members:.1f}")

        # 2. Распределение абонементов
        annual_members = estimated_unique_members * self.ASSUMPTIONS['membership_mix']['annual']
        monthly_members = estimated_unique_members * self.ASSUMPTIONS['membership_mix']['monthly']

        print(f"   Годовых абонементов: {annual_members:,.0f}")
        print(f"   Месячных абонементов: {monthly_members:,.0f}")

        # 3. Выручка от абонементов
        annual_revenue = annual_members * self.PRICES['annual_membership']
        monthly_revenue = monthly_members * self.PRICES['monthly_membership'] * 12

        print(f"   Выручка от годовых: ${annual_revenue:,.0f}")
        print(f"   Выручка от месячных: ${monthly_revenue:,.0f}")

        # 4. Выручка от casual (уже рассчитана)
        casual_revenue_total = sum(self.casual_revenue_by_year.values())
        print(f"   Выручка от casual: ${casual_revenue_total:,.0f}")
        
        # 5. Вывод по годам для casual
        print(f"\n   Выручка от casual по годам:")
        for year, revenue in sorted(self.casual_revenue_by_year.items()):
            print(f"       {year}: ${revenue:,.0f}")

        # 6. Общая выручка
        total_revenue = annual_revenue + monthly_revenue + casual_revenue_total

        self.metrics['revenue'] = {
            'total': float(total_revenue),
            'annual_members': float(annual_revenue),
            'monthly_members': float(monthly_revenue),
            'casual': float(casual_revenue_total),
            'estimated_unique_members': int(estimated_unique_members),
            'rides_per_member': total_member_rides / estimated_unique_members,
            'total_member_rides': int(total_member_rides),
            'total_casual_rides': int(total_casual_rides),
            'casual_by_year': self.casual_revenue_by_year
        }

        print(f"\n   ИТОГО ВЫРУЧКА: ${total_revenue:,.0f}")

        return self.metrics['revenue']

    def calculate_complete_costs(self):
        """
        ПОЛНЫЙ расчёт затрат (исправленный)
        """
        print("\nПОЛНЫЙ РАСЧЁТ ЗАТРАТ")
        print("-" * 40)

        total_rides = self.aggregates['total_rides']
        total_members = self.metrics['revenue']['estimated_unique_members']

        # 1. Операционные затраты
        maintenance_cost = total_rides * self.ASSUMPTIONS['bike_maintenance_cost_per_ride']
        station_cost = self.ASSUMPTIONS['station_maintenance_per_month'] * 12 * self.ASSUMPTIONS['avg_stations']

        print(f"   Техобслуживание велосипедов: ${maintenance_cost:,.0f}")
        print(f"   Обслуживание станций: ${station_cost:,.0f}")

        # 2. Маркетинг
        marketing_cost = total_members * self.ASSUMPTIONS['customer_acquisition_cost']
        print(f"   Маркетинг (CAC): ${marketing_cost:,.0f}")

        # 3. Амортизация велосипедов
        total_bikes = self.ASSUMPTIONS['avg_bikes_per_station'] * self.ASSUMPTIONS['avg_stations']
        bike_depreciation = total_bikes * 300
        print(f"   Амортизация велосипедов: ${bike_depreciation:,.0f}")

        # 4. Зарплаты персонала
        staff_cost = self.ASSUMPTIONS['avg_stations'] * 2 * 3000 * 12
        print(f"   Зарплаты персонала: ${staff_cost:,.0f}")

        # 5. Итого затрат
        total_costs = maintenance_cost + station_cost + marketing_cost + bike_depreciation + staff_cost

        # 6. Структура затрат
        breakdown = {
            'maintenance': maintenance_cost / total_costs * 100,
            'stations': station_cost / total_costs * 100,
            'marketing': marketing_cost / total_costs * 100,
            'depreciation': bike_depreciation / total_costs * 100,
            'staff': staff_cost / total_costs * 100
        }

        self.metrics['costs'] = {
            'total': float(total_costs),
            'maintenance_per_ride': float(maintenance_cost),
            'station_maintenance': float(station_cost),
            'marketing': float(marketing_cost),
            'depreciation': float(bike_depreciation),
            'staff': float(staff_cost),
            'cost_per_ride': float(total_costs / total_rides) if total_rides > 0 else 0,
            'breakdown': breakdown
        }

        print(f"\n  ОБЩИЕ ЗАТРАТЫ: ${total_costs:,.0f}")
        print(f"   Затраты на поездку: ${total_costs/total_rides:.2f}")
        
        # Вывод структуры затрат
        print(f"\n   Структура затрат:")
        for category, percentage in breakdown.items():
            print(f"       {category.capitalize()}: {percentage:.1f}%")

        return self.metrics['costs']

    def calculate_unit_economics(self):
        """
        Расчёт юнит-экономики (исправленный)
        """
        print("\nРАСЧЁТ ЮНИТ-ЭКОНОМИКИ")
        print("-" * 40)

        revenue = self.metrics['revenue']['total']
        costs = self.metrics['costs']['total']
        total_rides = self.aggregates['total_rides']
        unique_members = self.metrics['revenue']['estimated_unique_members']

        # Ключевые метрики
        profit = revenue - costs
        profit_margin = (profit / revenue * 100) if revenue > 0 else 0

        # CPE (Cost Per Engagement)
        cpe = costs / total_rides if total_rides > 0 else 0

        # ARPU (Average Revenue Per User)
        arpu = revenue / unique_members if unique_members > 0 else 0

        # CAC (Customer Acquisition Cost)
        cac = self.ASSUMPTIONS['customer_acquisition_cost']

        # LTV (Lifetime Value)
        avg_lifetime_months = 1 / self.ASSUMPTIONS['monthly_churn_rate']
        ltv = arpu * avg_lifetime_months / 12
        ltv_cac_ratio = ltv / cac if cac > 0 else 0

        # Выручка и затраты на поездку
        revenue_per_ride = revenue / total_rides if total_rides > 0 else 0

        self.metrics['unit_economics'] = {
            'total_rides': int(total_rides),
            'revenue_per_ride': float(revenue_per_ride),
            'cost_per_engagement': float(cpe),
            'gross_profit': float(profit),
            'profit_margin_percent': float(profit_margin),
            'arpu': float(arpu),
            'cac': float(cac),
            'ltv': float(ltv),
            'ltv_cac_ratio': float(ltv_cac_ratio),
            'payback_period_months': float(cac / (arpu/12)) if arpu > 0 else 0,
            'breakeven_rides': int(costs / revenue_per_ride) if revenue_per_ride > 0 else 0,
            'contribution_margin_per_ride': float(revenue_per_ride - cpe)
        }

        # Вывод результатов
        print(f"   Выручка на поездку: ${revenue_per_ride:.2f}")
        print(f"   Затраты на поездку: ${cpe:.2f}")
        print(f"   Прибыль на поездку: ${revenue_per_ride - cpe:.2f}")
        print(f"   Маржа прибыли: {profit_margin:.1f}%")
        print(f"   ARPU: ${arpu:.0f}")
        print(f"   CAC: ${cac:.0f}")
        print(f"   LTV: ${ltv:.0f}")
        print(f"   LTV/CAC: {ltv_cac_ratio:.2f}")

        # Оценка показателей
        print(f"\n   ОЦЕНКА ПОКАЗАТЕЛЕЙ:")
        if ltv_cac_ratio > 3:
            print(f"       LTV/CAC = {ltv_cac_ratio:.2f} - ОТЛИЧНО!")
        elif ltv_cac_ratio > 1:
            print(f"       LTV/CAC = {ltv_cac_ratio:.2f} - требуется улучшение")
        else:
            print(f"       LTV/CAC = {ltv_cac_ratio:.2f} - КРИТИЧЕСКИ НИЗКИЙ!")

        if profit_margin > 20:
            print(f"       Маржа {profit_margin:.1f}% - ВЫСОКАЯ!")
        elif profit_margin > 10:
            print(f"       Маржа {profit_margin:.1f}% - СРЕДНЯЯ")
        elif profit_margin > 0:
            print(f"       Маржа {profit_margin:.1f}% - НИЗКАЯ")
        else:
            print(f"       Маржа {profit_margin:.1f}% - ОТРИЦАТЕЛЬНАЯ!")

        return self.metrics['unit_economics']

    def analyze_bike_type_trends(self):
        """
        Анализ трендов по типам велосипедов
        """
        print("\n🚲 АНАЛИЗ ТРЕНДОВ ТИПОВ ВЕЛОСИПЕДОВ")
        print("-" * 40)

        # Собираем данные по годам
        yearly_totals = {}
        for year, bike_stats in self.bike_analysis.items():
            yearly_totals[year] = {}
            for row in bike_stats.iter_rows(named=True):
                yearly_totals[year][row['rideable_type']] = row['rides_count']

        # Анализ трендов
        bike_types = set()
        for year_data in yearly_totals.values():
            bike_types.update(year_data.keys())

        print("   Распределение по годам:")
        for bike_type in sorted(bike_types):
            print(f"\n   🚴 {bike_type}:")
            for year in sorted(yearly_totals.keys()):
                count = yearly_totals[year].get(bike_type, 0)
                total = sum(yearly_totals[year].values())
                percentage = (count / total * 100) if total > 0 else 0
                print(f"       {year}: {count:,} ({percentage:.1f}%)")

        # Анализ роста электрических велосипедов
        if 'electric_bike' in bike_types:
            print(f"\n   РОСТ ЭЛЕКТРИЧЕСКИХ ВЕЛОСИПЕДОВ:")
            electric_data = []
            for year in [2020, 2021, 2022, 2023, 2024, 2025]:
                if year in yearly_totals and 'electric_bike' in yearly_totals[year]:
                    count = yearly_totals[year]['electric_bike']
                    total = sum(yearly_totals[year].values())
                    percentage = (count / total * 100) if total > 0 else 0
                    electric_data.append((year, count, total, percentage))
                    print(f"       {year}: {percentage:.1f}% ({count:,} шт)")
            
            # Анализ роста
            if len(electric_data) >= 2:
                start_year, start_count, start_total, start_pct = electric_data[0]
                end_year, end_count, end_total, end_pct = electric_data[-1]
                growth_pct = ((end_pct - start_pct) / start_pct * 100) if start_pct > 0 else 0
                print(f"\n   Рост с {start_year} по {end_year}: {start_pct:.1f}% → {end_pct:.1f}% (+{growth_pct:.0f}%)")

        return yearly_totals

    def generate_complete_report(self):
        """
        Генерация ПОЛНОГО отчёта (исправленный)
        """
        print("\n" + "="*80)
        print("📋 ПОЛНЫЙ ФИНАНСОВЫЙ ОТЧЁТ ПРОКАТА ВЕЛОСИПЕДОВ")
        print("="*80)

        # 1. Общая статистика
        print(f"\nОБЩАЯ СТАТИСТИКА")
        print(f"   {'─'*40}")
        print(f"   Всего поездок: {self.aggregates['total_rides']:,}")
        print(f"   Members: {self.aggregates['member_rides']:,} ({self.aggregates['member_rides']/self.aggregates['total_rides']*100:.1f}%)")
        print(f"   Casual: {self.aggregates['casual_rides']:,} ({self.aggregates['casual_rides']/self.aggregates['total_rides']*100:.1f}%)")
        print(f"   Уникальных пользователей: {self.metrics['revenue']['estimated_unique_members']:,}")
        print(f"   Поездок на пользователя: {self.metrics['revenue']['rides_per_member']:.1f}")

        # 2. Финансовые показатели
        print(f"\nФИНАНСОВЫЕ ПОКАЗАТЕЛИ")
        print(f"   {'─'*40}")

        rev = self.metrics['revenue']
        cost = self.metrics['costs']
        ue = self.metrics['unit_economics']

        print(f"   ВЫРУЧКА:")
        print(f"       • От годовых абонементов: ${rev['annual_members']:,.0f}")
        print(f"       • От месячных абонементов: ${rev['monthly_members']:,.0f}")
        print(f"       • От разовых поездок: ${rev['casual']:,.0f}")
        print(f"       • ИТОГО: ${rev['total']:,.0f}")

        print(f"\n   ЗАТРАТЫ:")
        print(f"       • Техобслуживание: ${cost['maintenance_per_ride']:,.0f}")
        print(f"       • Станции: ${cost['station_maintenance']:,.0f}")
        print(f"       • Маркетинг: ${cost['marketing']:,.0f}")
        print(f"       • Амортизация: ${cost['depreciation']:,.0f}")
        print(f"       • Персонал: ${cost['staff']:,.0f}")
        print(f"       • ИТОГО: ${cost['total']:,.0f}")

        print(f"\n   ФИНАНСОВЫЙ РЕЗУЛЬТАТ:")
        print(f"       • Валовая прибыль: ${ue['gross_profit']:,.0f}")
        print(f"       • Маржа прибыли: {ue['profit_margin_percent']:.1f}%")

        # 3. Юнит-экономика
        print(f"\n ЮНИТ-ЭКОНОМИКА")
        print(f"   {'─'*40}")
        print(f"   Выручка на поездку: ${ue['revenue_per_ride']:.2f}")
        print(f"   Затраты на поездку: ${ue['cost_per_engagement']:.2f}")
        print(f"   Маржа на поездку: ${ue['contribution_margin_per_ride']:.2f}")
        print(f"   ARPU (ср. выручка на пользователя): ${ue['arpu']:.0f}")
        print(f"   CAC (стоимость привлечения): ${ue['cac']:.0f}")
        print(f"   LTV (пожизненная ценность): ${ue['ltv']:.0f}")
        print(f"   LTV/CAC Ratio: {ue['ltv_cac_ratio']:.2f}")
        print(f"   Период окупаемости: {ue['payback_period_months']:.1f} месяцев")
        print(f"   Безубыточность при: {ue['breakeven_rides']:,} поездок в год")

        # 4. Анализ эффективности
        print(f"\nАНАЛИЗ ЭФФЕКТИВНОСТИ")
        print(f"   {'─'*40}")

        # Оценка прибыльности
        if ue['profit_margin_percent'] > 20:
            print(f"   ОТЛИЧНО: Высокая маржа прибыли (>20%)")
        elif ue['profit_margin_percent'] > 10:
            print(f"   УДОВЛЕТВОРИТЕЛЬНО: Средняя маржа прибыли")
        elif ue['profit_margin_percent'] > 0:
            print(f"   ВНИМАНИЕ: Низкая маржа прибыли")
        else:
            print(f"   КРИТИЧНО: Отрицательная маржа прибыли")

        # Оценка LTV/CAC
        if ue['ltv_cac_ratio'] > 3:
            print(f"   ОТЛИЧНО: LTV/CAC > 3, бизнес масштабируем")
        elif ue['ltv_cac_ratio'] > 1.5:
            print(f"   НОРМАЛЬНО: LTV/CAC = {ue['ltv_cac_ratio']:.2f}")
        elif ue['ltv_cac_ratio'] > 1:
            print(f"   ВНИМАНИЕ: LTV/CAC = {ue['ltv_cac_ratio']:.2f}, нужно улучшать")
        else:
            print(f"   КРИТИЧНО: LTV/CAC < 1, бизнес убыточен")

        # 5. Рекомендации
        print(f"\nРЕКОМЕНДАЦИИ")
        print(f"   {'─'*40}")

        recommendations = []

        # Анализ затрат на персонал
        if cost['staff'] / cost['total'] > 0.5:
            recommendations.append("1. СРОЧНО оптимизируйте штат - персонал составляет 52% затрат")
        
        if ue['cost_per_engagement'] > 2:
            recommendations.append("2. Снижайте операционные затраты, особенно обслуживание станций")
        
        if rev['rides_per_member'] < 25:
            recommendations.append("3. Увеличивайте частоту использования (программы лояльности, напоминания)")
        
        if ue['payback_period_months'] > 6:
            recommendations.append("4. Оптимизируйте маркетинг для снижения CAC")
        
        # Анализ трендов
        if hasattr(self, 'yearly_totals'):
            # Проверяем падение в 2025
            if 2025 in self.aggregates['years'] and 2024 in self.aggregates['years']:
                drop = ((self.aggregates['years'][2024]['total'] - 
                        self.aggregates['years'][2025]['total']) / 
                       self.aggregates['years'][2024]['total'] * 100)
                if drop > 5:
                    recommendations.append(f"5. Исследуйте падение на {drop:.1f}% в 2025 году")

        if not recommendations:
            recommendations.append("Все показатели в норме. Сфокусируйтесь на росте пользовательской базы")

        for i, rec in enumerate(recommendations, 1):
            print(f"   {i}. {rec}")

        # 6. Тренды
        print(f"\nКЛЮЧЕВЫЕ ТРЕНДЫ")
        print(f"   {'─'*40}")

        # Рост электрических велосипедов
        if hasattr(self, 'yearly_totals'):
            electric_growth = []
            for year in [2020, 2021, 2022, 2023, 2024, 2025]:
                if year in self.yearly_totals and 'electric_bike' in self.yearly_totals[year]:
                    total = sum(self.yearly_totals[year].values())
                    percentage = self.yearly_totals[year]['electric_bike'] / total * 100
                    electric_growth.append((year, percentage))

            if len(electric_growth) > 1:
                growth_rate = ((electric_growth[-1][1] - electric_growth[0][1]) / electric_growth[0][1]) * 100
                print(f"   🚀 Электрические велосипеды: рост с {electric_growth[0][1]:.1f}% до {electric_growth[-1][1]:.1f}% (+{growth_rate:.0f}%)")

        print(f"\n" + "="*80)
        print("ОТЧЁТ СФОРМИРОВАН")
        print("="*80)

        return self.metrics

    def visualize_results(self):
        """
        Визуализация ключевых метрик
        """
        print("\nВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ")
        print("-" * 40)

        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('Анализ юнит-экономики проката велосипедов', fontsize=16, fontweight='bold')

        # 1. Распределение по типу пользователей
        labels = ['Members', 'Casual']
        sizes = [self.aggregates['member_rides'], self.aggregates['casual_rides']]
        colors = ['#2E86AB', '#A23B72']
        axes[0, 0].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors,
                      textprops={'fontsize': 10})
        axes[0, 0].set_title('Распределение поездок по типам пользователей', fontweight='bold')

        # 2. Выручка по источникам
        revenue_sources = ['Годовые\nабон.', 'Месячные\nабон.', 'Разовые\nпоездки']
        revenue_values = [
            self.metrics['revenue']['annual_members'],
            self.metrics['revenue']['monthly_members'],
            self.metrics['revenue']['casual']
        ]
        colors = ['#1B9AAA', '#06D6A0', '#FFD166']
        bars = axes[0, 1].bar(revenue_sources, revenue_values, color=colors)
        axes[0, 1].set_title('Выручка по источникам', fontweight='bold')
        axes[0, 1].set_ylabel('Млн $')
        axes[0, 1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.0f}'))
        for i, v in enumerate(revenue_values):
            axes[0, 1].text(i, v, f'${v/1e6:.1f}M', ha='center', va='bottom', fontweight='bold')

        # 3. Структура затрат
        cost_labels = ['Техобсл.', 'Станции', 'Маркетинг', 'Амортиз.', 'Персонал']
        cost_values = list(self.metrics['costs']['breakdown'].values())
        axes[0, 2].pie(cost_values, labels=cost_labels, autopct='%1.1f%%',
                      colors=['#EF476F', '#FFD166', '#06D6A0', '#118AB2', '#073B4C'],
                      textprops={'fontsize': 9})
        axes[0, 2].set_title('Структура затрат', fontweight='bold')

        # 4. Ключевые метрики (ARPU, CAC, LTV)
        metrics_to_show = ['ARPU', 'CAC', 'LTV']
        metric_values = [
            self.metrics['unit_economics']['arpu'],
            self.metrics['unit_economics']['cac'],
            self.metrics['unit_economics']['ltv']
        ]
        metric_colors = ['#4ECDC4', '#FF6B6B', '#45B7D1']
        bars = axes[1, 0].bar(metrics_to_show, metric_values, color=metric_colors)
        axes[1, 0].set_title('Ключевые метрики пользователя', fontweight='bold')
        axes[1, 0].set_ylabel('$')
        axes[1, 0].axhline(y=metric_values[1], color='red', linestyle='--', alpha=0.5, label='CAC')
        
        for bar, value in zip(bars, metric_values):
            height = bar.get_height()
            axes[1, 0].text(bar.get_x() + bar.get_width()/2., height,
                          f'${value:.0f}', ha='center', va='bottom', fontweight='bold')
        axes[1, 0].legend()

        # 5. Рост электрических велосипедов
        if hasattr(self, 'yearly_totals'):
            years = []
            electric_percentages = []

            for year in sorted(self.yearly_totals.keys()):
                if 'electric_bike' in self.yearly_totals[year]:
                    total = sum(self.yearly_totals[year].values())
                    percentage = self.yearly_totals[year]['electric_bike'] / total * 100
                    years.append(year)
                    electric_percentages.append(percentage)

            if years:
                axes[1, 1].plot(years, electric_percentages, marker='o', color='#06D6A0', 
                               linewidth=3, markersize=8)
                axes[1, 1].fill_between(years, electric_percentages, alpha=0.2, color='#06D6A0')
                axes[1, 1].set_title('Рост доли электрических велосипедов', fontweight='bold')
                axes[1, 1].set_xlabel('Год')
                axes[1, 1].set_ylabel('Доля, %')
                axes[1, 1].grid(True, alpha=0.3)
                axes[1, 1].set_ylim(0, 100)

                for i, (year, perc) in enumerate(zip(years, electric_percentages)):
                    axes[1, 1].text(year, perc + 3, f'{perc:.1f}%', ha='center', fontweight='bold')

        # 6. Маржа на поездку
        ride_metrics = ['Выручка', 'Затраты', 'Маржа']
        ride_values = [
            self.metrics['unit_economics']['revenue_per_ride'],
            self.metrics['unit_economics']['cost_per_engagement'],
            self.metrics['unit_economics']['contribution_margin_per_ride']
        ]
        ride_colors = ['#4ECDC4', '#FF6B6B', '#06D6A0']

        bars = axes[1, 2].bar(ride_metrics, ride_values, color=ride_colors)
        axes[1, 2].set_title('Экономика одной поездки', fontweight='bold')
        axes[1, 2].set_ylabel('$')
        axes[1, 2].axhline(y=0, color='black', linewidth=0.8)
        axes[1, 2].axhline(y=ride_values[0], color='blue', linestyle=':', alpha=0.3)
        axes[1, 2].axhline(y=ride_values[1], color='red', linestyle=':', alpha=0.3)

        for i, v in enumerate(ride_values):
            axes[1, 2].text(i, v, f'${v:.2f}', ha='center',
                          va='bottom' if v >= 0 else 'top', fontweight='bold')

        plt.tight_layout()
        plt.show()

        print("Графики построены")

    def run_full_analysis(self):
        """
        Запуск полного анализа
        """
        print("ЗАПУСК ПОЛНОГО ФИНАНСОВОГО АНАЛИЗА")
        print("="*60)

        # 1. Расчёт выручки
        self.calculate_complete_revenue()

        # 2. Расчёт затрат
        self.calculate_complete_costs()

        # 3. Юнит-экономика
        self.calculate_unit_economics()

        # 4. Анализ трендов
        self.yearly_totals = self.analyze_bike_type_trends()

        # 5. Отчёт
        self.generate_complete_report()

        # 6. Визуализация
        self.visualize_results()

        print("\n" + "="*60)
        print("ФИНАНСОВЫЙ АНАЛИЗ ЗАВЕРШЁН")
        print("="*60)

        return self.metrics


# =============== ЗАПУСК АНАЛИЗА С ВАШИМИ ДАННЫМИ ===============

# Создаём словарь с доходами от casual по годам (из вашего вывода)
casual_revenue_by_year = {
    2020: 7483846,
    2021: 10778767,
    2022: 9356852,
    2023: 8105018,
    2024: 8283427,
    2025: 7127236
}

# Агрегатные данные из вашего вывода
aggregates = {
    'total_rides': 29770159,
    'member_rides': 18142269,
    'casual_rides': 11627890,
    'years': {
        2020: {'total': 3468881, 'member_ratio': 0.615},
        2021: {'total': 4800858, 'member_ratio': 0.529},
        2022: {'total': 5437102, 'member_ratio': 0.588},
        2023: {'total': 5563046, 'member_ratio': 0.641},
        2024: {'total': 5583576, 'member_ratio': 0.642},
        2025: {'total': 4916696, 'member_ratio': 0.635}
    }
}

# Данные по типам велосипедов (из вашего вывода - нужно преобразовать)
# Создаём датафреймы для каждого года
bike_analysis = {}

# 2020
bike_analysis[2020] = pl.DataFrame({
    'rideable_type': ['classic_bike', 'electric_bike'],
    'rides_count': [2976645, 492236],
    'avg_duration': [None, None],
    'avg_distance': [None, None],
    'member_ratio': [None, None]
})

# 2021
bike_analysis[2021] = pl.DataFrame({
    'rideable_type': ['classic_bike', 'electric_bike', 'docked_bike'],
    'rides_count': [3045847, 1449121, 305890],
    'avg_duration': [None, None, None],
    'avg_distance': [None, None, None],
    'member_ratio': [None, None, None]
})

# 2022
bike_analysis[2022] = pl.DataFrame({
    'rideable_type': ['electric_bike', 'classic_bike', 'docked_bike'],
    'rides_count': [2747972, 2517317, 171813],
    'avg_duration': [None, None, None],
    'avg_distance': [None, None, None],
    'member_ratio': [None, None, None]
})

# 2023
bike_analysis[2023] = pl.DataFrame({
    'rideable_type': ['electric_bike', 'classic_bike', 'docked_bike'],
    'rides_count': [2837942, 2649728, 75376],
    'avg_duration': [None, None, None],
    'avg_distance': [None, None, None],
    'member_ratio': [None, None, None]
})

# 2024
bike_analysis[2024] = pl.DataFrame({
    'rideable_type': ['electric_bike', 'classic_bike'],
    'rides_count': [2868967, 2714609],
    'avg_duration': [None, None],
    'avg_distance': [None, None],
    'member_ratio': [None, None]
})

# 2025
bike_analysis[2025] = pl.DataFrame({
    'rideable_type': ['electric_bike', 'classic_bike'],
    'rides_count': [3134216, 1782480],
    'avg_duration': [None, None],
    'avg_distance': [None, None],
    'member_ratio': [None, None]
})

# Запускаем полный анализ
analyzer = CompleteBikeRentalEconomics(aggregates, bike_analysis, casual_revenue_by_year)
results = analyzer.run_full_analysis()

# Выводим ключевые метрики
print("\nИТОГОВЫЕ КЛЮЧЕВЫЕ МЕТРИКИ:")
print(f"   LTV/CAC Ratio: {results['unit_economics']['ltv_cac_ratio']:.2f}")
print(f"   Маржа прибыли: {results['unit_economics']['profit_margin_percent']:.1f}%")
print(f"   Выручка на поездку: ${results['unit_economics']['revenue_per_ride']:.2f}")
print(f"   Затраты на поездку: ${results['unit_economics']['cost_per_engagement']:.2f}")
print(f"   Маржа на поездку: ${results['unit_economics']['contribution_margin_per_ride']:.2f}")

print("\n" + "="*60)
print("ЧТО ДЕЛАТЬ ДАЛЬШЕ:")
print("="*60)
print("1. Оптимизировать затраты на персонал (сейчас 52% от всех затрат)")
print("2. Исследовать падение поездок в 2025 году (-12.5%)")
print("3. Продолжать наращивать долю электрических велосипедов")
print("4. Увеличивать частоту поездок на одного пользователя")
print("5. Анализировать сезонность для оптимизации ресурсов")
```
