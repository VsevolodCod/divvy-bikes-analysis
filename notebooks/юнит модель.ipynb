```python
class BikeRentalUnitEconomics:
    def __init__(self, df):
        self.df = df
        
    def calculate_metrics(self):
        # 1. Расчет выручки на основе тарифов из статьи
        # Годовой абонемент: $150 (со скидкой $118.90)
        # Месячный: $15
        # Разовая поездка: $3 за 30 минут + $4 за каждые доп. 30 минут
        
        # Разделяем пользователей
        members = self.df.filter(pl.col('member_casual') == 'member')
        casuals = self.df.filter(pl.col('member_casual') == 'casual')
        
        return {
            'total_rides': len(self.df),
            'member_rides': len(members),
            'casual_rides': len(casuals),
            'avg_duration_minutes': self.df['duration_minutes'].mean(),
            'avg_distance_km': self.df['distance_km'].mean() if 'distance_km' in self.df.columns else None
        }
```

```python
def calculate_unit_economics(df_all):
    """Расчет юнит-экономики"""
    
    # Основные допущения (на основе статьи)
    PRICES = {
        'annual_membership': 150.0,
        'discounted_annual': 118.90,
        'monthly_membership': 15.0,
        'single_ride_base': 3.0,  # первые 30 минут
        'single_ride_extra': 4.0  # каждые доп. 30 минут
    }
    
    # Анализ поведения пользователей
    analysis = df_all.group_by(['member_casual', 'year']).agg([
        pl.count().alias('rides_count'),
        pl.mean('duration_minutes').alias('avg_duration'),
        (pl.col('duration_minutes') > 30).sum().alias('long_rides')
    ])
    
    # Расчет доходов
    # Для members: предполагаем годовые абонементы
    member_revenue = analysis.filter(
        pl.col('member_casual') == 'member'
    )['rides_count'].sum() * PRICES['annual_membership']
    
    # Для casual: рассчитываем по поездкам
    casual_df = df_all.filter(pl.col('member_casual') == 'casual')
    casual_revenue = casual_df.with_columns([
        ((pl.col('duration_minutes') - 30) / 30).ceil().alias('extra_blocks')
    ]).with_columns([
        (PRICES['single_ride_base'] + 
         pl.when(pl.col('extra_blocks') > 0)
           .then(pl.col('extra_blocks') * PRICES['single_ride_extra'])
           .otherwise(0)).alias('ride_revenue')
    ])['ride_revenue'].sum()
    
    # Расчет средних показателей
    total_rides = len(df_all)
    total_revenue = member_revenue + casual_revenue
    
    return {
        'total_rides': total_rides,
        'total_revenue': total_revenue,
        'revenue_per_ride': total_revenue / total_rides,
        'member_contribution': member_revenue / total_revenue * 100,
        'casual_contribution': casual_revenue / total_revenue * 100,
        'avg_ride_duration': df_all['duration_minutes'].mean(),
        'peak_hours': df_all['start_hour'].mode()[0] if 'start_hour' in df_all.columns else None
    }
```

```python
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import numpy as np

class BikeRentalUnitEconomics:
    """
    Комплексная юнит-модель для сервиса проката велосипедов
    """
    
    def __init__(self, dataframes_dict):
        """
        Инициализация с данными за все годы
        
        Parameters:
        -----------
        dataframes_dict : dict
            Словарь с датафреймами по годам {2020: df_2020, 2021: df_2021, ...}
        """
        self.dataframes = dataframes_dict
        self.consolidated_df = None
        self.metrics = {}
        
        # Тарифы на основе статьи
        self.PRICES = {
            'annual_membership': 150.0,  # Годовой абонемент
            'discounted_annual': 118.90,  # Со скидкой
            'monthly_membership': 15.0,   # Месячный абонемент
            'single_ride_base': 3.0,      # Базовые 30 минут
            'single_ride_extra': 4.0      # Дополнительные 30 минут
        }
        
        # Предположения для модели (можно менять)
        self.ASSUMPTIONS = {
            'membership_mix': {'annual': 0.6, 'monthly': 0.3, 'casual': 0.1},
            'monthly_churn_rate': 0.08,  # 8% отток в месяц
            'customer_acquisition_cost': 25.0,
            'bike_maintenance_cost_per_ride': 0.35,
            'station_maintenance_per_month': 1500,
            'avg_bikes_per_station': 10,
            'avg_stations': 580
        }
        
    def _get_column_name(self, df, possible_names):
        """Находит существующее название колонки из списка возможных"""
        for name in possible_names:
            if name in df.columns:
                return name
        return None
    
    def _standardize_column(self, df, target_name, source_names):
        """Стандартизирует колонку, используя разные возможные имена"""
        source_name = self._get_column_name(df, source_names)
        if source_name and source_name != target_name:
            return df.rename({source_name: target_name})
        return df
    
    def prepare_consolidated_data(self):
        """Подготовка единого датасета из всех годов"""
        processed_dfs = []
        
        for year, df in self.dataframes.items():
            print(f"  Обработка данных за {year} год...")
            df_processed = df.clone()
            
            # Добавляем год
            df_processed = df_processed.with_columns([
                pl.lit(year).alias('year')
            ])
            
            # Стандартизируем названия колонок
            df_processed = self._standardize_column(df_processed, 'day_of_week', 
                                                   ['day_of_week', 'start_day_of_week', 'start_weekday'])
            df_processed = self._standardize_column(df_processed, 'hour_of_day', 
                                                   ['hour_of_day', 'start_hour'])
            df_processed = self._standardize_column(df_processed, 'month', 
                                                   ['month', 'start_month'])
            df_processed = self._standardize_column(df_processed, 'date', 
                                                   ['date', 'start_date'])
            
            # Проверяем и добавляем длительность поездки
            if 'duration_minutes' not in df_processed.columns:
                # Вычисляем из started_at и ended_at
                df_processed = df_processed.with_columns([
                    ((pl.col('ended_at') - pl.col('started_at')).dt.total_minutes())
                    .clip(lower_bound=0)  # Избегаем отрицательных значений
                    .alias('duration_minutes')
                ])
            
            # Преобразуем день недели из строки в число (для 2020 года)
            if 'day_of_week' in df_processed.columns:
                if df_processed['day_of_week'].dtype == pl.String:
                    days_mapping = {
                        'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4,
                        'Friday': 5, 'Saturday': 6, 'Sunday': 7
                    }
                    df_processed = df_processed.with_columns([
                        pl.col('day_of_week')
                        .replace_strict(days_mapping, default=0)
                        .cast(pl.Int8)
                        .alias('day_of_week_num')
                    ])
                else:
                    df_processed = df_processed.with_columns([
                        pl.col('day_of_week').cast(pl.Int8).alias('day_of_week_num')
                    ])
            else:
                # Вычисляем из даты начала
                df_processed = df_processed.with_columns([
                    pl.col('started_at').dt.weekday().alias('day_of_week_num')
                ])
            
            # Добавляем час дня, если нет
            if 'hour_of_day' not in df_processed.columns:
                df_processed = df_processed.with_columns([
                    pl.col('started_at').dt.hour().alias('hour_of_day')
                ])
            
            # Добавляем месяц, если нет
            if 'month' not in df_processed.columns:
                df_processed = df_processed.with_columns([
                    pl.col('started_at').dt.month().alias('month')
                ])
            
            # Добавляем дату, если нет
            if 'date' not in df_processed.columns:
                df_processed = df_processed.with_columns([
                    pl.col('started_at').dt.date().alias('date')
                ])
            
            # Добавляем расстояние, если нет
            if 'distance_km' not in df_processed.columns:
                if 'estimated_distance_km' in df_processed.columns:
                    df_processed = df_processed.rename({'estimated_distance_km': 'distance_km'})
                else:
                    # Оцениваем: средняя скорость 15 км/ч = 0.25 км/мин
                    df_processed = df_processed.with_columns([
                        (pl.col('duration_minutes') * 0.25).alias('distance_km')
                    ])
            
            # Добавляем категорию длительности
            df_processed = df_processed.with_columns([
                pl.when(pl.col('duration_minutes') <= 30)
                  .then(pl.lit('short'))
                  .when(pl.col('duration_minutes') <= 60)
                  .then(pl.lit('medium'))
                  .when(pl.col('duration_minutes') <= 120)
                  .then(pl.lit('long'))
                  .otherwise(pl.lit('very_long'))
                  .alias('duration_category'),
                
                # Сезон (для анализа)
                pl.when(pl.col('month').is_in([12, 1, 2]))
                  .then(pl.lit('winter'))
                  .when(pl.col('month').is_in([3, 4, 5]))
                  .then(pl.lit('spring'))
                  .when(pl.col('month').is_in([6, 7, 8]))
                  .then(pl.lit('summer'))
                  .otherwise(pl.lit('fall'))
                  .alias('season')
            ])
            
            # Приводим числовые колонки к единому типу
            # Это нужно для избежания ошибок при объединении
            for col in df_processed.columns:
                dtype = df_processed[col].dtype
                if dtype in [pl.Int8, pl.Int16, pl.Int32, pl.Int64, pl.UInt8, pl.UInt16, pl.UInt32, pl.UInt64]:
                    # Целочисленные типы приводим к Int64
                    df_processed = df_processed.with_columns([
                        pl.col(col).cast(pl.Int64).alias(col)
                    ])
                elif dtype in [pl.Float32, pl.Float64]:
                    # Вещественные типы приводим к Float64
                    df_processed = df_processed.with_columns([
                        pl.col(col).cast(pl.Float64).alias(col)
                    ])
            
            # Оставляем только нужные колонки
            required_columns = [
                'ride_id', 'rideable_type', 'started_at', 'ended_at', 
                'start_station_name', 'end_station_name', 'member_casual',
                'duration_minutes', 'duration_category', 'year', 
                'day_of_week_num', 'hour_of_day', 'month', 'date',
                'distance_km', 'season'
            ]
            
            # Отбираем только существующие колонки
            existing_cols = [col for col in required_columns if col in df_processed.columns]
            df_processed = df_processed.select(existing_cols)
            
            # Убедимся, что все числовые колонки имеют правильный тип
            numeric_columns = ['duration_minutes', 'distance_km', 'day_of_week_num', 
                              'hour_of_day', 'month', 'year']
            for col in numeric_columns:
                if col in df_processed.columns:
                    if col in ['duration_minutes', 'distance_km']:
                        df_processed = df_processed.with_columns([
                            pl.col(col).cast(pl.Float64).alias(col)
                        ])
                    else:
                        df_processed = df_processed.with_columns([
                            pl.col(col).cast(pl.Int64).alias(col)
                        ])
            
            processed_dfs.append(df_processed)
            print(f"    Готово: {len(df_processed):,} строк, {len(df_processed.columns)} колонок")
        
        # Объединяем все годы с помощью vertical_relaxed, который разрешает разницу в типах
        try:
            self.consolidated_df = pl.concat(processed_dfs, how='vertical_relaxed')
        except:
            # Если vertical_relaxed не поддерживается, пробуем обычный concat
            print("  Попытка объединения с приведением типов...")
            # Приведем все датафреймы к одинаковой схеме
            for i in range(len(processed_dfs)):
                for col in processed_dfs[0].columns:
                    if col in processed_dfs[i].columns:
                        # Приводим к типу первого датафрейма
                        target_type = processed_dfs[0][col].dtype
                        processed_dfs[i] = processed_dfs[i].with_columns([
                            pl.col(col).cast(target_type).alias(col)
                        ])
            self.consolidated_df = pl.concat(processed_dfs, how='vertical')
        
        print(f"\nОбъединенный датасет: {len(self.consolidated_df):,} строк, {len(self.consolidated_df.columns)} колонок")
        
        # Проверяем данные
        print("\nСтатистика объединенных данных:")
        print(f"   Годы: {self.consolidated_df['year'].unique().sort().to_list()}")
        print(f"   Типы пользователей: {self.consolidated_df['member_casual'].unique().to_list()}")
        print(f"   Средняя длительность: {self.consolidated_df['duration_minutes'].mean():.1f} минут")
        print(f"   Среднее расстояние: {self.consolidated_df['distance_km'].mean():.2f} км")
        
        return self.consolidated_df

    
    def calculate_revenue(self):
        """Расчет выручки по разным сегментам"""
        
        print("  Расчет выручки...")
        
        # Разделяем пользователей
        members = self.consolidated_df.filter(pl.col('member_casual') == 'member')
        casuals = self.consolidated_df.filter(pl.col('member_casual') == 'casual')
        
        # Для members (абонементы)
        # Оцениваем количество уникальных пользователей (грубо по количеству поездок)
        # В реальности нужно считать уникальные ID, но их нет в данных
        total_member_rides = len(members)
        
        # Предполагаем, что каждый member делает в среднем 30 поездок в год
        # Это приблизительная оценка - можно уточнить при наличии данных
        estimated_unique_members = max(1, total_member_rides // 30)
        
        # Предполагаем распределение годовых/месячных
        annual_members = estimated_unique_members * self.ASSUMPTIONS['membership_mix']['annual']
        monthly_members = estimated_unique_members * self.ASSUMPTIONS['membership_mix']['monthly']
        
        # Выручка от members
        annual_revenue = annual_members * self.PRICES['annual_membership']
        monthly_revenue = monthly_members * self.PRICES['monthly_membership'] * 12  # Годовая
        
        # Для casual (разовые поездки)
        casual_revenue_df = casuals.with_columns([
            # Количество дополнительных блоков по 30 минут
            ((pl.col('duration_minutes') - 30) / 30).ceil().clip(lower_bound=0).alias('extra_blocks')
        ]).with_columns([
            # Стоимость поездки
            (self.PRICES['single_ride_base'] + 
             pl.col('extra_blocks') * self.PRICES['single_ride_extra']).alias('ride_revenue')
        ])
        
        casual_revenue = casual_revenue_df['ride_revenue'].sum()
        
        total_revenue = annual_revenue + monthly_revenue + casual_revenue
        
        self.metrics['revenue'] = {
            'total': float(total_revenue),
            'annual_members': float(annual_revenue),
            'monthly_members': float(monthly_revenue),
            'casual': float(casual_revenue),
            'estimated_unique_members': int(estimated_unique_members),
            'rides_per_member': total_member_rides / estimated_unique_members if estimated_unique_members > 0 else 0,
            'total_member_rides': int(total_member_rides),
            'total_casual_rides': len(casuals)
        }
        
        print(f"    Оценка уникальных пользователей: {estimated_unique_members:,}")
        print(f"    Выручка: ${total_revenue:,.0f}")
        
        return self.metrics['revenue']
    
    def calculate_costs(self):
        """Расчет всех затрат"""
        
        print("  Расчет затрат...")
        
        total_rides = len(self.consolidated_df)
        
        # Операционные затраты
        maintenance_cost = total_rides * self.ASSUMPTIONS['bike_maintenance_cost_per_ride']
        station_cost = self.ASSUMPTIONS['station_maintenance_per_month'] * 12 * self.ASSUMPTIONS['avg_stations']
        
        # Маркетинговые затраты (оценка)
        total_members = self.metrics.get('revenue', {}).get('estimated_unique_members', 0)
        marketing_cost = total_members * self.ASSUMPTIONS['customer_acquisition_cost']
        
        # Амортизация велосипедов
        total_bikes = self.ASSUMPTIONS['avg_bikes_per_station'] * self.ASSUMPTIONS['avg_stations']
        bike_depreciation = total_bikes * 300  # $300 в год на велосипед
        
        # Зарплаты персонала (оценка)
        staff_cost = self.ASSUMPTIONS['avg_stations'] * 2 * 3000 * 12  # 2 человека на станцию, $3000/мес
        
        total_costs = maintenance_cost + station_cost + marketing_cost + bike_depreciation + staff_cost
        
        self.metrics['costs'] = {
            'total': float(total_costs),
            'maintenance_per_ride': float(maintenance_cost),
            'station_maintenance': float(station_cost),
            'marketing': float(marketing_cost),
            'depreciation': float(bike_depreciation),
            'staff': float(staff_cost),
            'cost_per_ride': float(total_costs / total_rides) if total_rides > 0 else 0,
            'breakdown': {
                'maintenance': float(maintenance_cost / total_costs * 100) if total_costs > 0 else 0,
                'stations': float(station_cost / total_costs * 100) if total_costs > 0 else 0,
                'marketing': float(marketing_cost / total_costs * 100) if total_costs > 0 else 0,
                'depreciation': float(bike_depreciation / total_costs * 100) if total_costs > 0 else 0,
                'staff': float(staff_cost / total_costs * 100) if total_costs > 0 else 0
            }
        }
        
        print(f"    Общие затраты: ${total_costs:,.0f}")
        print(f"    Затраты на поездку: ${total_costs/total_rides:.2f}")
        
        return self.metrics['costs']
    
    def calculate_unit_economics(self):
        """Расчет ключевых метрик юнит-экономики"""
        
        print("  Расчет юнит-экономики...")
        
        # Сначала рассчитываем выручку и затраты
        if 'revenue' not in self.metrics:
            self.calculate_revenue()
        if 'costs' not in self.metrics:
            self.calculate_costs()
        
        revenue = self.metrics['revenue']['total']
        costs = self.metrics['costs']['total']
        total_rides = len(self.consolidated_df)
        unique_members = self.metrics['revenue']['estimated_unique_members']
        
        # Ключевые метрики
        profit = revenue - costs
        profit_margin = (profit / revenue * 100) if revenue > 0 else 0
        
        # CPE (Cost Per Engagement)
        cpe = costs / total_rides if total_rides > 0 else 0
        
        # ARPU (Average Revenue Per User)
        arpu = revenue / unique_members if unique_members > 0 else 0
        
        # CAC (Customer Acquisition Cost)
        cac = self.ASSUMPTIONS['customer_acquisition_cost']
        
        # LTV (Lifetime Value)
        # Предполагаем, что пользователь остается 1/churn_rate месяцев
        avg_lifetime_months = 1 / self.ASSUMPTIONS['monthly_churn_rate']
        ltv = arpu * avg_lifetime_months / 12  # Переводим в годы
        
        # Показатель эффективности
        ltv_cac_ratio = ltv / cac if cac > 0 else 0
        
        # Выручка и затраты на поездку
        revenue_per_ride = revenue / total_rides if total_rides > 0 else 0
        
        self.metrics['unit_economics'] = {
            'total_rides': int(total_rides),
            'revenue_per_ride': float(revenue_per_ride),
            'cost_per_engagement': float(cpe),
            'gross_profit': float(profit),
            'profit_margin_percent': float(profit_margin),
            'arpu': float(arpu),
            'cac': float(cac),
            'ltv': float(ltv),
            'ltv_cac_ratio': float(ltv_cac_ratio),
            'payback_period_months': float(cac / (arpu/12)) if arpu > 0 else 0,
            'breakeven_rides': int(costs / revenue_per_ride) if revenue_per_ride > 0 else 0,
            'contribution_margin_per_ride': float(revenue_per_ride - cpe)
        }
        
        print(f"    Выручка на поездку: ${revenue_per_ride:.2f}")
        print(f"    Затраты на поездку: ${cpe:.2f}")
        print(f"    Маржа: {profit_margin:.1f}%")
        print(f"    LTV/CAC: {ltv_cac_ratio:.2f}")
        
        return self.metrics['unit_economics']
    
    def analyze_seasonality(self):
        """Анализ сезонности и пиковых периодов"""
        
        print("  Анализ сезонности...")
        
        # Анализ по месяцам
        monthly_analysis = self.consolidated_df.group_by(['year', 'month']).agg([
            pl.count().alias('rides_count'),
            pl.mean('duration_minutes').alias('avg_duration'),
            pl.mean('distance_km').alias('avg_distance'),
            (pl.col('member_casual') == 'member').mean().alias('member_ratio')
        ]).sort(['year', 'month'])
        
        # Анализ по дням недели
        weekday_analysis = self.consolidated_df.group_by('day_of_week_num').agg([
            pl.count().alias('rides_count'),
            pl.mean('duration_minutes').alias('avg_duration'),
            pl.mean('distance_km').alias('avg_distance'),
            (pl.col('member_casual') == 'member').mean().alias('member_ratio')
        ]).sort('day_of_week_num')
        
        # Анализ по часам
        hourly_analysis = self.consolidated_df.group_by('hour_of_day').agg([
            pl.count().alias('rides_count'),
            pl.mean('duration_minutes').alias('avg_duration'),
            pl.mean('distance_km').alias('avg_distance')
        ]).sort('hour_of_day')
        
        # Анализ по сезонам
        seasonal_analysis = self.consolidated_df.group_by('season').agg([
            pl.count().alias('rides_count'),
            pl.mean('duration_minutes').alias('avg_duration'),
            pl.mean('distance_km').alias('avg_distance'),
            (pl.col('member_casual') == 'member').mean().alias('member_ratio')
        ]).sort('rides_count', descending=True)
        
        self.metrics['seasonality'] = {
            'monthly': monthly_analysis,
            'weekday': weekday_analysis,
            'hourly': hourly_analysis,
            'seasonal': seasonal_analysis
        }
        
        # Находим пиковые значения
        if len(monthly_analysis) > 0:
            peak_month = monthly_analysis.sort('rides_count', descending=True).row(0)
            print(f"    Пиковый месяц: {int(peak_month[1])} (год {int(peak_month[0])}) - {int(peak_month[2]):,} поездок")
        
        if len(hourly_analysis) > 0:
            peak_hour = hourly_analysis.sort('rides_count', descending=True).row(0)
            print(f"    Пиковый час: {int(peak_hour[0])}:00 - {int(peak_hour[1]):,} поездок")
        
        return self.metrics['seasonality']
    
    def analyze_bike_type_performance(self):
        """Анализ эффективности разных типов велосипедов"""
        
        print("  Анализ типов велосипедов...")
        
        if 'rideable_type' not in self.consolidated_df.columns:
            print("    Внимание: колонка 'rideable_type' не найдена в данных")
            return None
        
        bike_analysis = self.consolidated_df.group_by('rideable_type').agg([
            pl.count().alias('rides_count'),
            pl.mean('duration_minutes').alias('avg_duration'),
            pl.mean('distance_km').alias('avg_distance'),
            (pl.col('member_casual') == 'member').mean().alias('member_ratio'),
            pl.col('duration_minutes').quantile(0.5).alias('median_duration')
        ]).sort('rides_count', descending=True)
        
        self.metrics['bike_types'] = bike_analysis
        
        print("    Распределение по типам велосипедов:")
        for row in bike_analysis.iter_rows(named=True):
            print(f"      {row['rideable_type']}: {row['rides_count']:,} поездок "
                  f"({row['rides_count']/len(self.consolidated_df)*100:.1f}%)")
        
        return bike_analysis
    
    def analyze_station_performance(self):
        """Анализ эффективности станций"""
        
        print("  Анализ станций...")
        
        # Топ станций отправления
        top_start_stations = self.consolidated_df.group_by('start_station_name').agg([
            pl.count().alias('departures'),
            pl.mean('duration_minutes').alias('avg_duration'),
            (pl.col('member_casual') == 'member').mean().alias('member_ratio'),
            pl.col('distance_km').mean().alias('avg_distance')
        ]).sort('departures', descending=True).head(10)
        
        # Топ станций назначения
        top_end_stations = self.consolidated_df.group_by('end_station_name').agg([
            pl.count().alias('arrivals'),
            pl.mean('duration_minutes').alias('avg_duration'),
            (pl.col('member_casual') == 'member').mean().alias('member_ratio'),
            pl.col('distance_km').mean().alias('avg_distance')
        ]).sort('arrivals', descending=True).head(10)
        
        # Самые популярные маршруты
        popular_routes = self.consolidated_df.group_by(['start_station_name', 'end_station_name']).agg([
            pl.count().alias('trip_count'),
            pl.mean('duration_minutes').alias('avg_duration'),
            pl.mean('distance_km').alias('avg_distance')
        ]).filter(pl.col('start_station_name') != pl.col('end_station_name'))\
          .sort('trip_count', descending=True).head(10)
        
        self.metrics['stations'] = {
            'top_departures': top_start_stations,
            'top_arrivals': top_end_stations,
            'popular_routes': popular_routes
        }
        
        print("    Топ-3 станции отправления:")
        for i, row in enumerate(top_start_stations.head(3).iter_rows(named=True)):
            print(f"      {i+1}. {row['start_station_name']}: {row['departures']:,} отправлений")
        
        return self.metrics['stations']
    
    def generate_report(self):
        """Генерация полного отчета"""
        
        print("\n" + "="*70)
        print("ЮНИТ-МОДЕЛЬ ПРОКАТА ВЕЛОСИПЕДОВ - ПОЛНЫЙ ОТЧЕТ")
        print("="*70)
        
        ue = self.metrics['unit_economics']
        rev = self.metrics['revenue']
        cost = self.metrics['costs']
        
        print(f"\nОБЩАЯ СТАТИСТИКА:")
        print(f"   Всего поездок: {ue['total_rides']:,}")
        print(f"   Оценка уникальных пользователей: {rev['estimated_unique_members']:,}")
        print(f"   Поездок на пользователя: {rev['rides_per_member']:.1f}")
        print(f"   Поездки members: {rev['total_member_rides']:,} ({rev['total_member_rides']/ue['total_rides']*100:.1f}%)")
        print(f"   Поездки casual: {rev['total_casual_rides']:,} ({rev['total_casual_rides']/ue['total_rides']*100:.1f}%)")
        
        print(f"\nФИНАНСОВЫЕ ПОКАЗАТЕЛИ:")
        print(f"   Общая выручка: ${rev['total']:,.0f}")
        print(f"   Общие затраты: ${cost['total']:,.0f}")
        print(f"   Валовая прибыль: ${ue['gross_profit']:,.0f}")
        print(f"   Маржа прибыли: {ue['profit_margin_percent']:.1f}%")
        
        print(f"\nЮНИТ-ЭКОНОМИКА:")
        print(f"   Выручка на поездку: ${ue['revenue_per_ride']:.2f}")
        print(f"   Затраты на поездку (CPE): ${ue['cost_per_engagement']:.2f}")
        print(f"   Маржа с поездки: ${ue['contribution_margin_per_ride']:.2f}")
        print(f"   ARPU (ср. выручка на пользователя): ${ue['arpu']:.2f}")
        print(f"   CAC (стоимость привлечения): ${ue['cac']:.2f}")
        print(f"   LTV (пожизненная ценность): ${ue['ltv']:.2f}")
        print(f"   LTV/CAC Ratio: {ue['ltv_cac_ratio']:.2f}")
        
        print(f"\nКЛЮЧЕВЫЕ ПОКАЗАТЕЛИ:")
        print(f"   Период окупаемости: {ue['payback_period_months']:.1f} месяцев")
        print(f"   Безубыточность при: {ue['breakeven_rides']:,} поездок в год")
        
        # Анализ структуры затрат
        print(f"\nСТРУКТУРА ЗАТРАТ:")
        for category, percentage in cost['breakdown'].items():
            print(f"   {category.capitalize()}: {percentage:.1f}%")
        
        # Оценка эффективности
        print(f"\nОЦЕНКА ЭФФЕКТИВНОСТИ:")
        if ue['profit_margin_percent'] > 20:
            print("   ОТЛИЧНО: Высокая маржа прибыли (>20%)")
        elif ue['profit_margin_percent'] > 0:
            print("   УДОВЛЕТВОРИТЕЛЬНО: Положительная маржа")
        else:
            print("   КРИТИЧНО: Отрицательная маржа прибыли")
        
        if ue['ltv_cac_ratio'] > 3:
            print("   ОТЛИЧНО: LTV/CAC > 3, бизнес масштабируем")
        elif ue['ltv_cac_ratio'] > 1:
            print(f"   ВНИМАНИЕ: LTV/CAC = {ue['ltv_cac_ratio']:.2f}, нужно улучшать")
        else:
            print("   КРИТИЧНО: LTV/CAC < 1, бизнес убыточен")
        
        # Рекомендации
        print(f"\nРЕКОМЕНДАЦИИ:")
        recommendations = []
        
        if ue['cost_per_engagement'] > ue['revenue_per_ride']:
            recommendations.append("1. СРОЧНО снижайте операционные затраты")
        
        if rev['rides_per_member'] < 20:
            recommendations.append("2. Увеличивайте частоту использования (программы лояльности)")
        
        if ue['payback_period_months'] > 12:
            recommendations.append("3. Оптимизируйте маркетинг для снижения CAC")
        
        if ue['contribution_margin_per_ride'] < 1:
            recommendations.append("4. Повышайте тарифы или сокращайте длительность бесплатного периода")
        
        if not recommendations:
            recommendations.append("Все показатели в норме. Сфокусируйтесь на росте пользовательской базы")
        
        for rec in recommendations:
            print(f"   {rec}")
        
        # Сезонные рекомендации
        seasonal_data = self.metrics.get('seasonality', {}).get('seasonal')
        if seasonal_data is not None:
            print(f"\nСЕЗОННЫЕ РЕКОМЕНДАЦИИ:")
            for row in seasonal_data.iter_rows(named=True):
                season = row['season']
                rides = row['rides_count']
                percentage = rides / ue['total_rides'] * 100
                print(f"   {season.capitalize()}: {rides:,} поездок ({percentage:.1f}%)")
        
        return self.metrics
    
    def visualize_metrics(self, save_path=None):
        """Визуализация ключевых метрик"""
        
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('Анализ юнит-экономики проката велосипедов', fontsize=16, fontweight='bold')
        
        # 1. Распределение по типу пользователей
        member_casual_counts = self.consolidated_df.group_by('member_casual').count()
        axes[0, 0].pie(member_casual_counts['count'], 
                      labels=member_casual_counts['member_casual'],
                      autopct='%1.1f%%',
                      colors=['#2E86AB', '#A23B72'])
        axes[0, 0].set_title('Распределение по типу пользователей', fontweight='bold')
        
        # 2. Сезонность по месяцам (среднее за все годы)
        if 'seasonality' in self.metrics and 'monthly' in self.metrics['seasonality']:
            monthly = self.metrics['seasonality']['monthly']
            monthly_avg = monthly.group_by('month').agg(pl.mean('rides_count').alias('avg_rides')).sort('month')
            axes[0, 1].bar(monthly_avg['month'], monthly_avg['avg_rides'], color='#4ECDC4')
            axes[0, 1].set_title('Среднее количество поездок по месяцам', fontweight='bold')
            axes[0, 1].set_xlabel('Месяц')
            axes[0, 1].set_ylabel('Поездок')
            axes[0, 1].set_xticks(range(1, 13))
        
        # 3. Распределение по дням недели
        if 'seasonality' in self.metrics and 'weekday' in self.metrics['seasonality']:
            weekdays = self.metrics['seasonality']['weekday']
            axes[0, 2].bar(weekdays['day_of_week_num'], weekdays['rides_count'], color='#FF6B6B')
            axes[0, 2].set_title('Поездки по дням недели', fontweight='bold')
            axes[0, 2].set_xlabel('День недели (1-пн, 7-вс)')
            axes[0, 2].set_ylabel('Поездок')
            axes[0, 2].set_xticks(range(1, 8))
        
        # 4. Выручка по источникам
        revenue_sources = ['Годовые\nабон.', 'Месячные\nабон.', 'Разовые\nпоездки']
        revenue_values = [
            self.metrics['revenue']['annual_members'],
            self.metrics['revenue']['monthly_members'],
            self.metrics['revenue']['casual']
        ]
        colors = ['#1B9AAA', '#06D6A0', '#FFD166']
        axes[1, 0].bar(revenue_sources, revenue_values, color=colors)
        axes[1, 0].set_title('Выручка по источникам', fontweight='bold')
        axes[1, 0].set_ylabel('Долларов')
        for i, v in enumerate(revenue_values):
            axes[1, 0].text(i, v, f'${v:,.0f}', ha='center', va='bottom')
        
        # 5. Структура затрат
        if 'costs' in self.metrics and 'breakdown' in self.metrics['costs']:
            cost_data = self.metrics['costs']['breakdown']
            cost_labels = [k.capitalize() for k in cost_data.keys()]
            cost_values = list(cost_data.values())
            axes[1, 1].pie(cost_values, labels=cost_labels, autopct='%1.1f%%',
                          colors=['#EF476F', '#FFD166', '#06D6A0', '#118AB2', '#073B4C'])
            axes[1, 1].set_title('Структура затрат', fontweight='bold')
        
        # 6. Ключевые метрики (ARPU, CAC, LTV)
        metrics_to_show = ['ARPU', 'CAC', 'LTV']
        metric_values = [
            self.metrics['unit_economics']['arpu'],
            self.metrics['unit_economics']['cac'],
            self.metrics['unit_economics']['ltv']
        ]
        metric_colors = ['#4ECDC4', '#FF6B6B', '#45B7D1']
        bars = axes[1, 2].bar(metrics_to_show, metric_values, color=metric_colors)
        axes[1, 2].set_title('Ключевые метрики (ARPU, CAC, LTV)', fontweight='bold')
        axes[1, 2].set_ylabel('Долларов')
        
        # Добавляем значения на столбцы
        for bar, value in zip(bars, metric_values):
            height = bar.get_height()
            axes[1, 2].text(bar.get_x() + bar.get_width()/2., height,
                          f'${value:.0f}', ha='center', va='bottom')
        
        # Добавляем линию для LTV/CAC > 3
        if self.metrics['unit_economics']['ltv_cac_ratio'] > 3:
            axes[1, 2].axhline(y=self.metrics['unit_economics']['cac'] * 3, 
                              color='green', linestyle='--', alpha=0.5, label='LTV/CAC=3')
            axes[1, 2].legend()
        
        plt.tight_layout()
        
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"\nГрафики сохранены в {save_path}")
        
        plt.show()
    
    def export_to_excel(self, filename="unit_economics_report.xlsx"):
        """Экспорт результатов в Excel"""
        import pandas as pd
        
        with pd.ExcelWriter(filename, engine='openpyxl') as writer:
            # Основные метрики
            summary_data = {
                'Метрика': [
                    'Всего поездок', 'Уникальные пользователи', 'ARPU', 'CAC', 'LTV',
                    'LTV/CAC Ratio', 'Выручка', 'Затраты', 'Прибыль', 'Маржа прибыли',
                    'Выручка на поездку', 'Затраты на поездку', 'Маржа с поездки',
                    'Период окупаемости (мес)', 'Безубыточные поездки'
                ],
                'Значение': [
                    self.metrics['unit_economics']['total_rides'],
                    self.metrics['revenue']['estimated_unique_members'],
                    f"${self.metrics['unit_economics']['arpu']:.2f}",
                    f"${self.metrics['unit_economics']['cac']:.2f}",
                    f"${self.metrics['unit_economics']['ltv']:.2f}",
                    f"{self.metrics['unit_economics']['ltv_cac_ratio']:.2f}",
                    f"${self.metrics['revenue']['total']:,.0f}",
                    f"${self.metrics['costs']['total']:,.0f}",
                    f"${self.metrics['unit_economics']['gross_profit']:,.0f}",
                    f"{self.metrics['unit_economics']['profit_margin_percent']:.1f}%",
                    f"${self.metrics['unit_economics']['revenue_per_ride']:.2f}",
                    f"${self.metrics['unit_economics']['cost_per_engagement']:.2f}",
                    f"${self.metrics['unit_economics']['contribution_margin_per_ride']:.2f}",
                    f"{self.metrics['unit_economics']['payback_period_months']:.1f}",
                    f"{self.metrics['unit_economics']['breakeven_rides']:,}"
                ]
            }
            pd.DataFrame(summary_data).to_excel(writer, sheet_name='Сводка', index=False)
            
            # Выручка по источникам
            revenue_data = {
                'Источник': ['Годовые абонементы', 'Месячные абонементы', 'Разовые поездки', 'Итого'],
                'Выручка ($)': [
                    self.metrics['revenue']['annual_members'],
                    self.metrics['revenue']['monthly_members'],
                    self.metrics['revenue']['casual'],
                    self.metrics['revenue']['total']
                ],
                'Доля (%)': [
                    self.metrics['revenue']['annual_members'] / self.metrics['revenue']['total'] * 100,
                    self.metrics['revenue']['monthly_members'] / self.metrics['revenue']['total'] * 100,
                    self.metrics['revenue']['casual'] / self.metrics['revenue']['total'] * 100,
                    100
                ]
            }
            pd.DataFrame(revenue_data).to_excel(writer, sheet_name='Выручка', index=False)
            
            # Затраты
            costs_data = {
                'Категория': ['Техобслуживание', 'Станции', 'Маркетинг', 'Амортизация', 'Персонал', 'Итого'],
                'Сумма ($)': [
                    self.metrics['costs']['maintenance_per_ride'],
                    self.metrics['costs']['station_maintenance'],
                    self.metrics['costs']['marketing'],
                    self.metrics['costs']['depreciation'],
                    self.metrics['costs']['staff'],
                    self.metrics['costs']['total']
                ],
                'Доля (%)': list(self.metrics['costs']['breakdown'].values()) + [100]
            }
            pd.DataFrame(costs_data).to_excel(writer, sheet_name='Затраты', index=False)
            
            # Сезонность
            if 'seasonality' in self.metrics:
                self.metrics['seasonality']['monthly'].to_pandas().to_excel(
                    writer, sheet_name='Сезонность_месяцы', index=False)
                self.metrics['seasonality']['weekday'].to_pandas().to_excel(
                    writer, sheet_name='Сезонность_дни', index=False)
                self.metrics['seasonality']['hourly'].to_pandas().to_excel(
                    writer, sheet_name='Сезонность_часы', index=False)
            
            # Станции
            if 'stations' in self.metrics:
                self.metrics['stations']['top_departures'].to_pandas().to_excel(
                    writer, sheet_name='Топ_станции_отправ', index=False)
                self.metrics['stations']['popular_routes'].to_pandas().to_excel(
                    writer, sheet_name='Популярные_маршруты', index=False)
        
        print(f"\nОтчет экспортирован в {filename}")
    
    def run_complete_analysis(self):
        """Запуск полного анализа"""
        print("Начинаем анализ юнит-экономики...\n")
        
        # 1. Подготовка данных
        print("Шаг 1/5: Подготовка и объединение данных...")
        self.prepare_consolidated_data()
        
        # 2. Расчет метрик
        print("\nШаг 2/5: Расчет финансовых показателей...")
        self.calculate_revenue()
        self.calculate_costs()
        
        # 3. Юнит-экономика
        print("\nШаг 3/5: Расчет юнит-экономики...")
        self.calculate_unit_economics()
        
        # 4. Анализ
        print("\nШаг 4/5: Анализ данных...")
        self.analyze_seasonality()
        self.analyze_bike_type_performance()
        self.analyze_station_performance()
        
        # 5. Отчет
        print("\nШаг 5/5: Генерация отчета...")
        report = self.generate_report()
        
        return report
```
