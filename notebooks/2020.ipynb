````python
import polars as pl
from pathlib import Path
from math import radians, sin, cos, sqrt, atan2

# Базовые пути для сохранения/чтения
PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

PROCESSED_2020_DIR = PROJECT_ROOT / "data" / "processed" / "2020"
PROCESSED_2020_DIR.mkdir(parents=True, exist_ok=True)

CLEAN_PATH = PROCESSED_2020_DIR / "cleaned_2020_data.parquet"
SAMPLE_PATH = PROCESSED_2020_DIR / "cleaned_2020_data_sample.csv"
FILTERED_PATH = PROCESSED_2020_DIR / "cleaned_2020_data_filtered.parquet"

###УНИФИКАЦИЯ ТИПОВ ДАННЫХ ###
def unify_data_types(df, month_num):
    """Приводит все типы данных к единому формату для Polars"""
    
    # Приводим ID станций к строковому типу
    # Проверяем, есть ли эти колонки в датафрейме
    cols_to_cast = []
    if 'start_station_id' in df.columns:
        cols_to_cast.append(pl.col('start_station_id').cast(pl.Utf8))
    if 'end_station_id' in df.columns:
        cols_to_cast.append(pl.col('end_station_id').cast(pl.Utf8))
    
    if cols_to_cast:
        df = df.with_columns(cols_to_cast)
    
    # Унификация названий типов самокатов
    if 'rideable_type' in df.columns:
        df = df.with_columns(
            pl.when(pl.col('rideable_type') == 'docked_bike')
            .then(pl.lit('classic_bike'))
            .otherwise(pl.col('rideable_type'))
            .alias('rideable_type')
        )
    
    return df

### ОБРАБОТКА ПРОПУСКОВ ###
def handle_missing_data(df, month_num):
    """Умная обработка пропусков"""
    
    # очищаем названия станций
    station_cols = []
    for col in ['start_station_name', 'end_station_name']:
        if col in df.columns:
            station_cols.append(
                pl.when(pl.col(col).is_in(['None', 'none', 'NULL', 'null', '']))
                .then(None)
                .otherwise(pl.col(col))
                .alias(col)
            )
    
    if station_cols:
        df = df.with_columns(station_cols)
    
    if month_num <= 6:
        # Для месяцев с малым количеством пропусков - удаляем строки без координат
        df = df.filter(
            pl.col('end_lat').is_not_null() & 
            pl.col('end_lng').is_not_null()
        )
    else:
        # Для месяцев с большим количеством пропусков
        # Удаляем только строки без координат окончания
        mask_no_coords = pl.col('end_lat').is_null() & pl.col('end_lng').is_null()
        count_to_remove = df.filter(mask_no_coords).height
        print(f"  Удалено поездок без координат окончания: {count_to_remove}")
        
        df = df.filter(~mask_no_coords)
        
        # Заполняем пропущенные названия станций
        fill_cols = []
        for col in ['start_station_name', 'end_station_name']:
            if col in df.columns:
                fill_cols.append(pl.col(col).fill_null('Unknown Station'))
        
        for col in ['start_station_id', 'end_station_id']:
            if col in df.columns:
                fill_cols.append(pl.col(col).fill_null('unknown'))
        
        if fill_cols:
            df = df.with_columns(fill_cols)
    
    return df

###ОБНАРУЖЕНИЕ И УДАЛЕНИЕ АНОМАЛИЙ ###
def detect_and_remove_anomalies(df):
    """Обнаружение и удаление аномальных записей"""
    
    #длительность поездки в минутах
    df = df.with_columns(
        ((pl.col('ended_at') - pl.col('started_at')).dt.total_seconds() / 60)
        .alias('duration_minutes')
    )
    
    #Исправляем отрицательную длительность (переход через полночь)
    df = df.with_columns(
        pl.when(pl.col('duration_minutes') < 0)
        .then(pl.col('duration_minutes') + 1440)  # Добавляем 24 часа в минутах
        .otherwise(pl.col('duration_minutes'))
        .alias('duration_minutes')
    )
    
    #Удаляем аномально долгие/короткие поездки
    before = df.height
    
    df = df.filter(
        (pl.col('duration_minutes') >= 1) & 
        (pl.col('duration_minutes') <= 12 * 60)  # 12 часов максимум
    )
    
    after = df.height
    removed_by_duration = before - after
    if removed_by_duration > 0:
        print(f"  Удалено аномальных поездок по длительности: {removed_by_duration}")
    
    #поездки с одинаковыми координатами (помечаем, но не удаляем)
    same_location_expr = (pl.col('start_lat') == pl.col('end_lat')) & (pl.col('start_lng') == pl.col('end_lng'))
    same_location_count = df.filter(same_location_expr).height
    
    if same_location_count > 0:
        print(f"  Поездки с одинаковыми начальной и конечной точкой: {same_location_count}")
        # Добавляем метку
        df = df.with_columns(
            same_location_expr.alias('same_start_end')
        )
    else:
        # Все равно добавляем колонку со значением False
        df = df.with_columns(pl.lit(False).alias('same_start_end'))
    
    return df

###ДОБАВЛЕНИЕ АНАЛИТИЧЕСКИХ ПОЛЕЙ ###
def haversine_distance_udf(lat1: float, lon1: float, lat2: float, lon2: float) -> float:
    """UDF для расчета расстояния по гаверсинусу"""
    if lat2 is None or lon2 is None:
        return None
    
    R = 6371

    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    
    a = sin(dlat/2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    return R * c

def add_analytical_fields(df, month_num):
    """Добавляет поля для последующего анализа"""
    
    # Временные поля
    df = df.with_columns([
        pl.col('started_at').dt.date().alias('start_date'),
        pl.col('started_at').dt.hour().alias('start_hour'),
        pl.col('started_at').dt.strftime('%A').alias('start_day_of_week'),
        pl.col('started_at').dt.month().alias('start_month'),
        pl.col('started_at').dt.week().alias('start_week')
    ])
    
    # Расстояние между станциями (гаверсинус)
    # Используем map_elements для применения UDF
    df = df.with_columns(
        pl.struct(['start_lat', 'start_lng', 'end_lat', 'end_lng'])
        .map_elements(
            lambda x: haversine_distance_udf(x['start_lat'], x['start_lng'], x['end_lat'], x['end_lng']),
            return_dtype=pl.Float64
        )
        .alias('distance_km')
    )
    
    # Скорость поездки (км/ч)
    df = df.with_columns(
        (pl.col('distance_km') / (pl.col('duration_minutes') / 60))
        .alias('speed_kmh')
    )
    
    # Маркер для поездок с информацией о станциях
    station_info_expr = pl.lit(True)
    if 'start_station_name' in df.columns and 'end_station_name' in df.columns:
        station_info_expr = (
            pl.col('start_station_name').is_not_null() & 
            pl.col('end_station_name').is_not_null()
        )
    
    df = df.with_columns(
        station_info_expr.alias('has_station_info')
    )
    
    return df

### ОБЪЕДИНЕНИЕ ВСЕХ МЕСЯЦЕВ ###
def clean_and_combine_all_months(dataframes_dict):
    """Очищает и объединяет все месячные данные"""
    
    cleaned_dfs = []
    
    for month_name, df in dataframes_dict.items():
        print(f"\nОбработка: {month_name}")
        
        month_num = {
            'JFM': 3, 'April': 4, 'May': 5, 'June': 6,
            'July': 7, 'August': 8, 'September': 9,
            'October': 10, 'November': 11, 'December': 12
        }.get(month_name, 12)
        
        df_clean = df.clone()
        print(f"  Исходный размер: {df_clean.height:,} строк, {len(df_clean.columns)} колонок")
        
        #все шаги очистки
        df_clean = unify_data_types(df_clean, month_num)
        df_clean = handle_missing_data(df_clean, month_num)
        df_clean = detect_and_remove_anomalies(df_clean)
        df_clean = add_analytical_fields(df_clean, month_num)
        
        #идентификатор месяца
        df_clean = df_clean.with_columns([
            pl.lit(month_name).alias('data_month'),
            pl.lit(month_num).alias('data_month_num')
        ])
        
        cleaned_dfs.append(df_clean)
        print(f"  Размер после очистки: {df_clean.height:,} строк")
    
    # Объединяем все данные
    combined_df = pl.concat(cleaned_dfs)
    
    #проверка
    print(f"\n{'='*60}")
    print("ИТОГОВАЯ СТАТИСТИКА:")
    print(f"{'='*60}")
    print(f"Всего поездок: {combined_df.height:,}")
    print(f"Колонок: {len(combined_df.columns)}")
    print(f"Диапазон дат: {combined_df['started_at'].min()} до {combined_df['started_at'].max()}")
    
    # Проверяем типы самокатов
    if 'rideable_type' in combined_df.columns:
        ride_types = combined_df['rideable_type'].unique().to_list()
        print(f"Типы самокатов: {ride_types}")
    
    # Проверяем распределение пользователей
    if 'member_casual' in combined_df.columns:
        casual_pct = (combined_df['member_casual'] == 'casual').mean() * 100
        print(f"Доля casual пользователей: {casual_pct:.1f}%")
    
    # Проверяем пропуски
    if 'end_lat' in combined_df.columns:
        missing_coords = combined_df['end_lat'].is_null().sum()
        print(f"Пропуски в координатах окончания: {missing_coords}")
    
    return combined_df


all_data = {
    'JFM': df_JFM,
    'April': df_April,
    'May': df_May,
    'June': df_June,
    'July': df_July,
    'August': df_August,
    'September': df_September,
    'October': df_October,
    'November': df_November,
    'December': df_December
}


print("Начинаем очистку данных за 2020 год...")
df_2020_clean = clean_and_combine_all_months(all_data)

print("\nПервые 5 строк очищенных данных:")
print(df_2020_clean.head())

df_2020_clean.write_parquet(CLEAN_PATH)
print(f"\nДанные сохранены в '{CLEAN_PATH}'")

df_2020_clean.head(10000).write_csv(SAMPLE_PATH)
print(f"Пример данных сохранен в '{SAMPLE_PATH}'")

print("\nПроверка качества очистки:")
print(f"Минимальная длительность: {df_2020_clean['duration_minutes'].min():.1f} мин")
print(f"Максимальная длительность: {df_2020_clean['duration_minutes'].max():.1f} мин")
print(f"Средняя длительность: {df_2020_clean['duration_minutes'].mean():.1f} мин")
print(f"Медианная длительность: {df_2020_clean['duration_minutes'].median():.1f} мин")

print("\nРаспределение по месяцам:")
month_counts = df_2020_clean.group_by('data_month_num').agg(
    pl.count().alias('count'),
    (pl.col('member_casual') == 'casual').mean().alias('casual_pct')
).sort('data_month_num')

print(month_counts)
````

````python
import polars as pl
from pathlib import Path

# Если запустили ячейку отдельно, восстановим пути
if "CLEAN_PATH" not in globals():
    PROJECT_ROOT = Path.cwd()
    if PROJECT_ROOT.name == "notebooks":
        PROJECT_ROOT = PROJECT_ROOT.parent
    PROCESSED_2020_DIR = PROJECT_ROOT / "data" / "processed" / "2020"
    CLEAN_PATH = PROCESSED_2020_DIR / "cleaned_2020_data.parquet"
    SAMPLE_PATH = PROCESSED_2020_DIR / "cleaned_2020_data_sample.csv"
    FILTERED_PATH = PROCESSED_2020_DIR / "cleaned_2020_data_filtered.parquet"

df_2020_full = pl.read_parquet(CLEAN_PATH)
print(f"Размер полных данных: {df_2020_full.height:,} строк")

df_2020_sample = pl.read_csv(SAMPLE_PATH)
print(f"Размер сэмпла: {df_2020_sample.height:,} строк")

print("\n" + "="*50)
print("ПРОВЕРКА ОЧИЩЕННЫХ ДАННЫХ")
print("="*50)
print(f"Размер: {df_2020_full.height:,} строк, {len(df_2020_full.columns)} колонок")

print("\nТипы данных:")
for col in df_2020_full.columns:
    print(f"  {col:25} : {df_2020_full[col].dtype}")

print("\nПропуски данных:")
has_nulls = False
for col in df_2020_full.columns:
    null_count = df_2020_full[col].is_null().sum()
    if null_count > 0:
        has_nulls = True
        print(f"  {col:25} : {null_count:,} ({null_count/df_2020_full.height*100:.2f}%)")
if not has_nulls:
    print("  Нет пропусков!")

print("\nСтатистика длительности поездок:")
duration_stats = df_2020_full.select([
    pl.col('duration_minutes').min().alias('min'),
    pl.col('duration_minutes').max().alias('max'),
    pl.col('duration_minutes').mean().alias('mean'),
    pl.col('duration_minutes').median().alias('median'),
    pl.col('duration_minutes').std().alias('std')
])
print(duration_stats)

print("\nСтатистика скорости (км/ч):")
speed_stats = df_2020_full.select([
    pl.col('speed_kmh').min().alias('min'),
    pl.col('speed_kmh').max().alias('max'),
    pl.col('speed_kmh').mean().alias('mean'),
    pl.col('speed_kmh').median().alias('median')
])
print(speed_stats)

print("\nРаспределение длительности поездок (минуты):")
# Создаем временный датафрейм с bins
duration_bins_expr = pl.when(pl.col('duration_minutes') < 5).then(pl.lit('<5'))
duration_bins_expr = duration_bins_expr.when(pl.col('duration_minutes') < 10).then(pl.lit('5-10'))
duration_bins_expr = duration_bins_expr.when(pl.col('duration_minutes') < 15).then(pl.lit('10-15'))
duration_bins_expr = duration_bins_expr.when(pl.col('duration_minutes') < 30).then(pl.lit('15-30'))
duration_bins_expr = duration_bins_expr.when(pl.col('duration_minutes') < 60).then(pl.lit('30-60'))
duration_bins_expr = duration_bins_expr.otherwise(pl.lit('60+'))

duration_bins = df_2020_full.select(
    duration_bins_expr.alias('duration_bin')
).group_by('duration_bin').agg(pl.len().alias('count')).sort('duration_bin')
print(duration_bins)

# поездки с одинаковыми стартом и финишем
same_location = df_2020_full.filter(pl.col('same_start_end') == True)
print(f"\nПоездки с одинаковыми стартом и финишем: {same_location.height:,} ({same_location.height/df_2020_full.height*100:.2f}%)")
print("Топ-5 станций для таких поездок:")
top_same_location = same_location.group_by('start_station_name').agg(
    pl.len().alias('count')
).sort('count', descending=True).head(5)
print(top_same_location)

# поездки без информации о станциях
no_station_info = df_2020_full.filter(pl.col('has_station_info') == False)
print(f"\nПоездки без информации о станциях: {no_station_info.height:,}")
print(f"Процент от общего числа: {no_station_info.height/df_2020_full.height*100:.2f}%")
print("Распределение по месяцам:")
no_station_by_month = no_station_info.group_by('data_month').agg(
    pl.len().alias('count')
).sort('count', descending=True)
print(no_station_by_month)

# распределение по типам самокатов
print("\nРаспределение по типам самокатов:")
bike_type_stats = df_2020_full.group_by('rideable_type').agg(
    pl.len().alias('count'),
    pl.col('duration_minutes').mean().alias('avg_duration'),
    (pl.col('member_casual') == 'casual').mean().alias('casual_pct')
).sort('count', descending=True)
print(bike_type_stats)

# ============= АНАЛИЗ АНОМАЛЬНЫХ СКОРОСТЕЙ =============
print("\n" + "="*50)
print("АНАЛИЗ АНОМАЛЬНЫХ СКОРОСТЕЙ")
print("="*50)

# Нереально высокие скорости (> 30 км/ч для велосипеда)
high_speed = df_2020_full.filter(pl.col('speed_kmh') > 30)
print(f"Поездки со скоростью > 30 км/ч: {high_speed.height:,}")
if high_speed.height > 0:
    print(f"Максимальная скорость в этом наборе: {high_speed['speed_kmh'].max():.1f} км/ч")
    print(f"Минимальная скорость в этом наборе: {high_speed['speed_kmh'].min():.1f} км/ч")
    print(f"Средняя скорость в этом наборе: {high_speed['speed_kmh'].mean():.1f} км/ч")

# Нулевые скорости (вероятно, ошибка в координатах)
zero_speed = df_2020_full.filter(pl.col('speed_kmh') == 0)
print(f"\nПоездки с нулевой скоростью: {zero_speed.height:,}")

# Причина нулевой скорости - одинаковые координаты
zero_speed_same_location = zero_speed.filter(pl.col('same_start_end') == True)
print(f"Из них с одинаковыми координатами: {zero_speed_same_location.height:,}")

# Дополнительная фильтрация аномальных скоростей
print("\nФильтрация аномальных скоростей:")
print("Удаляем поездки со скоростью > 30 км/ч (аномалия)")
print("и поездки с нулевой скоростью (кроме тех, где действительно старт=финиш)")

# Фильтруем аномальные скорости
df_2020_filtered = df_2020_full.filter(
    (pl.col('speed_kmh') <= 30) |  # Разумный предел для велосипеда
    (pl.col('same_start_end') == True)  # Или если старт=финиш (скорость 0 нормальна)
)

print(f"\nРазмер после фильтрации скорости: {df_2020_filtered.height:,}")
print(f"Удалено: {df_2020_full.height - df_2020_filtered.height:,} поездок")
print(f"Процент удаленных: {(df_2020_full.height - df_2020_filtered.height)/df_2020_full.height*100:.2f}%")

# Сохраним дополнительно очищенные данные
df_2020_filtered.write_parquet(FILTERED_PATH)
print(f"\nДополнительно очищенные данные сохранены в '{FILTERED_PATH}'")

# ============= СЕЗОННЫЙ АНАЛИЗ =============
print("\n" + "="*50)
print("СЕЗОННЫЙ АНАЛИЗ")
print("="*50)

# Анализ по месяцам
monthly_stats = df_2020_filtered.group_by('data_month_num').agg([
    pl.len().alias('total_rides'),
    (pl.col('member_casual') == 'casual').mean().alias('casual_pct'),
    pl.col('duration_minutes').mean().alias('avg_duration'),
    pl.col('distance_km').mean().alias('avg_distance'),
    (pl.col('rideable_type') == 'electric_bike').mean().alias('electric_pct')
]).sort('data_month_num')

print("\nСтатистика по месяцам:")
print(monthly_stats)

# Переведем в проценты для читаемости
monthly_stats_formatted = monthly_stats.with_columns([
    (pl.col('casual_pct') * 100).round(1).alias('casual_%'),
    (pl.col('electric_pct') * 100).round(1).alias('electric_%'),
    pl.col('avg_duration').round(1),
    pl.col('avg_distance').round(2)
])

print("\nСезонность использования (в % и округленных значениях):")
print(monthly_stats_formatted.select(['data_month_num', 'total_rides', 'casual_%', 'electric_%', 'avg_duration', 'avg_distance']))

# ============= АНАЛИЗ ПОЛЬЗОВАТЕЛЕЙ (MEMBER vs CASUAL) =============
print("\n" + "="*50)
print("АНАЛИЗ ПОЛЬЗОВАТЕЛЕЙ (MEMBER vs CASUAL)")
print("="*50)

user_type_stats = df_2020_filtered.group_by('member_casual').agg([
    pl.len().alias('count'),
    pl.col('duration_minutes').mean().alias('avg_duration'),
    pl.col('distance_km').mean().alias('avg_distance'),
    pl.col('speed_kmh').mean().alias('avg_speed'),
    pl.col('start_hour').mode().first().alias('most_common_hour'),
    (pl.col('rideable_type') == 'electric_bike').mean().alias('electric_pct')
])

print("\nСравнение пользователей:")
print(user_type_stats)

# Детальный анализ по дням недели
print("\nИспользование по дням недели:")
weekday_stats = df_2020_filtered.group_by(['start_day_of_week', 'member_casual']).agg(
    pl.len().alias('count')
).sort('start_day_of_week')
print(weekday_stats)

# ============= ТОП-10 САМЫХ ПОПУЛЯРНЫХ СТАНЦИЙ =============
print("\n" + "="*50)
print("ТОП-10 САМЫХ ПОПУЛЯРНЫХ СТАНЦИЙ")
print("="*50)

# Стартовые станции
print("\nТоп-10 стартовых станций:")
top_start_stations = df_2020_filtered.filter(
    pl.col('start_station_name') != 'Unknown Station'
).group_by('start_station_name').agg(
    pl.len().alias('start_count'),
    (pl.col('member_casual') == 'casual').mean().alias('casual_pct')
).sort('start_count', descending=True).head(10)

print(top_start_stations)

# Конечные станции
print("\nТоп-10 конечных станций:")
top_end_stations = df_2020_filtered.filter(
    pl.col('end_station_name') != 'Unknown Station'
).group_by('end_station_name').agg(
    pl.len().alias('end_count')
).sort('end_count', descending=True).head(10)

print(top_end_stations)

# Популярные маршруты (только между известными станциями)
print("\nТоп-10 самых популярных маршрутов:")
popular_routes = df_2020_filtered.filter(
    (pl.col('has_station_info') == True) &
    (pl.col('start_station_name') != 'Unknown Station') &
    (pl.col('end_station_name') != 'Unknown Station')
).group_by(['start_station_name', 'end_station_name']).agg(
    pl.len().alias('ride_count'),
    pl.col('duration_minutes').mean().alias('avg_duration'),
    pl.col('distance_km').mean().alias('avg_distance')
).sort('ride_count', descending=True).head(10)

print(popular_routes)

# ============= ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ =============
print("\n" + "="*50)
print("ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ")
print("="*50)

# Распределение по времени суток
print("\nРаспределение поездок по часам (весь год):")
hourly_distribution = df_2020_filtered.group_by('start_hour').agg(
    pl.len().alias('count')
).sort('start_hour')
print(hourly_distribution)

# Самые длинные поездки
print("\nТоп-10 самых длинных поездок (по времени):")
longest_rides = df_2020_filtered.sort('duration_minutes', descending=True).head(10).select([
    'ride_id', 'duration_minutes', 'distance_km', 'start_station_name', 'end_station_name', 'member_casual'
])
print(longest_rides)

# Самые короткие поездки (но больше 1 минуты)
print("\nТоп-10 самых коротких поездок (более 1 минуты):")
shortest_rides = df_2020_filtered.filter(
    pl.col('duration_minutes') > 1
).sort('duration_minutes').head(10).select([
    'ride_id', 'duration_minutes', 'distance_km', 'start_station_name', 'end_station_name', 'member_casual'
])
print(shortest_rides)

# Анализ по дням недели для casual пользователей
print("\nЛюбимые дни casual пользователей:")
casual_weekday = df_2020_filtered.filter(
    pl.col('member_casual') == 'casual'
).group_by('start_day_of_week').agg(
    pl.len().alias('count')
).sort('count', descending=True)
print(casual_weekday)

# Анализ по дням недели для member пользователей
print("\nЛюбимые дни member пользователей:")
member_weekday = df_2020_filtered.filter(
    pl.col('member_casual') == 'member'
).group_by('start_day_of_week').agg(
    pl.len().alias('count')
).sort('count', descending=True)
print(member_weekday)

# Финальная статистика
print("\n" + "="*50)
print("ФИНАЛЬНАЯ СТАТИСТИКА")
print("="*50)
print(f"Исходный размер данных: 3,469,383 поездок")
print(f"После очистки: {df_2020_filtered.height:,} поездок")
print(f"Удалено всего: {3469383 - df_2020_filtered.height:,} поездок")
print(f"Сохранено: {df_2020_filtered.height/3469383*100:.1f}% данных")

print("\nКлючевые выводы:")
print("1. Данные успешно очищены от аномалий")
print("2. Созданы аналитические поля (длительность, расстояние, скорость)")
print("3. Удалены нереалистичные скорости (> 30 км/ч)")
print("4. Подготовлены данные для дальнейшего анализа и визуализации")

print("\nФайлы сохранены:")
print(f"  1. {CLEAN_PATH} - полностью очищенные данные")
print(f"  2. {FILTERED_PATH} - дополнительно отфильтрованные по скорости")
print(f"  3. {SAMPLE_PATH} - сэмпл данных для быстрого просмотра")
````

# Анализ очистки данных о велопрокате в Чикаго (2020)

## Обзор проекта

Мы проанализировали данные о 3.5 миллионах велопоездок в Чикаго за 2020 год. Исходные данные содержали несколько проблем, которые были устранены в процессе очистки.

## Проблемы в исходных данных и их решение

### Проблема 1: Аномально долгие поездки
**Проблема:** В данных присутствовали поездки длительностью более 12 часов (720+ минут), что физически невозможно для системы велопроката.

**Решение:** Установлен лимит в 720 минут (12 часов) - максимально допустимое время аренды.

```python
# Код фильтрации аномально долгих поездок
df_2020_cleaned = df_2020_raw.filter(
    pl.col('duration_minutes') <= 720  # Максимум 12 часов
)
```

**Результат:** Удалено 8,000-14,000 поездок в месяц (около 100,000 всего).

### Проблема 2: Пропущенные координаты окончания
**Проблема:** С июля по декабрь 2020 года встречались поездки с отсутствующими координатами конца маршрута (`end_lat`, `end_lng` = NaN).

**Решение:** Удалены поездки с пропущенными координатами окончания.

```python
# Код удаления поездок с пропущенными координатами
df_2020_cleaned = df_2020_cleaned.filter(
    pl.col('end_lat').is_not_null() & 
    pl.col('end_lng').is_not_null()
)
```

**Результат:** Удалено 111-938 поездок в месяц (июль-декабрь).

### Проблема 3: Неправильные типы велосипедов
**Проблема:** Встречался тип `docked_bike`, который идентичен `classic_bike`.

**Решение:** Стандартизация типов велосипедов.

```python
# Код стандартизации типов велосипедов
df_2020_cleaned = df_2020_cleaned.with_columns(
    pl.when(pl.col('rideable_type') == 'docked_bike')
      .then(pl.lit('classic_bike'))
      .otherwise(pl.col('rideable_type'))
      .alias('rideable_type')
)
```

### Проблема 4: Пропуски в названиях станций
**Проблема:** Часть поездок не имела информации о станциях начала/окончания.

**Решение:** Маркировка поездок без станций и создание флага.

```python
# Код обработки пропущенных станций
df_2020_cleaned = df_2020_cleaned.with_columns([
    pl.when(pl.col('start_station_name').is_null())
      .then(pl.lit('Unknown Station'))
      .otherwise(pl.col('start_station_name'))
      .alias('start_station_name'),
    
    pl.when(pl.col('end_station_name').is_null())
      .then(pl.lit('Unknown Station'))
      .otherwise(pl.col('end_station_name'))
      .alias('end_station_name'),
    
    ((pl.col('start_station_name') != 'Unknown Station') & 
     (pl.col('end_station_name') != 'Unknown Station')).alias('has_station_info')
])
```

### Проблема 5: Аномальные скорости
**Проблема:** Некоторые поездки имели нереалистично высокую скорость (до 358 км/ч).

**Решение:** Фильтрация поездок со скоростью > 30 км/ч.

```python
# Код фильтрации аномальных скоростей
df_2020_filtered = df_2020_cleaned.filter(
    (pl.col('speed_kmh') <= 30) |  # Разумный предел для велосипеда
    (pl.col('same_start_end') == True)  # Или если старт=финиш
)
```

**Результат:** Удалено 502 поездки (0.01%).

````python
import folium
import pandas as pd
from folium.plugins import HeatMap
from pathlib import Path

if "FILTERED_PATH" not in globals():
    PROJECT_ROOT = Path.cwd()
    if PROJECT_ROOT.name == "notebooks":
        PROJECT_ROOT = PROJECT_ROOT.parent
    PROCESSED_2020_DIR = PROJECT_ROOT / "data" / "processed" / "2020"
    FILTERED_PATH = PROCESSED_2020_DIR / "cleaned_2020_data_filtered.parquet"

df = pl.read_parquet(FILTERED_PATH)
df_sample = df.head(3469383)
df_pandas = df_sample.to_pandas()

chicago_map = folium.Map(
    location=[41.8781, -87.6298],  # Координаты Чикаго
    zoom_start=12,
    tiles='cartodbpositron'
)

heat_data_start = [[row['start_lat'], row['start_lng']] 
                   for index, row in df_pandas.iterrows() 
                   if not pd.isna(row['start_lat'])]

HeatMap(heat_data_start, 
        radius=10, 
        blur=15, 
        max_zoom=13,
        min_opacity=0.5).add_to(chicago_map)

top_stations = df.filter(pl.col('start_station_name') != 'Unknown Station') \
    .group_by('start_station_name', 'start_lat', 'start_lng') \
    .agg(pl.len().alias('count')) \
    .sort('count', descending=True) \
    .head(20)

top_stations_pd = top_stations.to_pandas()
for idx, row in top_stations_pd.iterrows():
    popup_text = f"""
    <b>{row['start_station_name']}</b><br>
    Количество поездок: {row['count']:,}<br>
    Координаты: {row['start_lat']:.4f}, {row['start_lng']:.4f}
    """
    
    folium.CircleMarker(
        location=[row['start_lat'], row['start_lng']],
        radius=min(20, row['count'] / 1000),  # Динамический размер
        popup=folium.Popup(popup_text, max_width=300),
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(chicago_map)
top_routes = df.filter(
    (pl.col('has_station_info') == True) &
    (pl.col('start_station_name') != 'Unknown Station') &
    (pl.col('end_station_name') != 'Unknown Station')
).group_by(['start_station_name', 'end_station_name', 
           'start_lat', 'start_lng', 'end_lat', 'end_lng']).agg(
    pl.len().alias('count')
).sort('count', descending=True).head(10)

top_routes_pd = top_routes.to_pandas()

for idx, row in top_routes_pd.iterrows():
    folium.PolyLine(
        locations=[[row['start_lat'], row['start_lng']], 
                  [row['end_lat'], row['end_lng']]],
        popup=f"Маршрут: {row['start_station_name']} → {row['end_station_name']}<br>"
              f"Количество поездок: {row['count']}",
        color='green',
        weight=min(5, row['count'] / 100),  # Толщина линии зависит от популярности
        opacity=0.7
    ).add_to(chicago_map)
chicago_map.save('chicago_bike_map.html')

print("Карта сохранена как 'chicago_bike_map.html'")
````
