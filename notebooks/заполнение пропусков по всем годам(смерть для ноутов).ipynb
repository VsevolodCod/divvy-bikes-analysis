```python
import polars as pl
from haversine import haversine
import numpy as np

def prepare_dataframe(df):
    """Подготавливает DataFrame для обработки, приводя типы данных к единому формату"""
    # Копируем DataFrame
    df = df.clone()
    
    # Список колонок станций, которые могут быть в разных годах
    station_cols = ["start_station_name", "start_station_id", 
                    "end_station_name", "end_station_id"]
    
    # Преобразуем "None" в None и приводим к строковому типу
    for col in station_cols:
        if col in df.columns:
            df = df.with_columns(
                pl.when(pl.col(col).is_null() | (pl.col(col) == "None"))
                .then(None)
                .otherwise(pl.col(col).cast(pl.String))
                .alias(col)
            )
    
    # Обеспечиваем наличие колонок с координатами
    if "start_lat" not in df.columns or "start_lng" not in df.columns:
        raise ValueError(f"DataFrame не содержит колонок с координатами")
    
    return df

def create_station_cache_from_data(df):
    """Создает кэш станций из DataFrame"""
    station_cache = []
    
    # Собираем стартовые станции
    if "start_station_name" in df.columns and "start_station_id" in df.columns:
        start_stations = df.filter(
            pl.col("start_station_name").is_not_null() &
            pl.col("start_station_id").is_not_null()
        ).select([
            "start_station_name", "start_station_id", "start_lat", "start_lng"
        ]).unique()
        
        for row in start_stations.iter_rows(named=True):
            station_cache.append({
                'name': row['start_station_name'],
                'id': row['start_station_id'],
                'lat': row['start_lat'],
                'lng': row['start_lng']
            })
    
    # Собираем конечные станции
    if "end_station_name" in df.columns and "end_station_id" in df.columns:
        end_stations = df.filter(
            pl.col("end_station_name").is_not_null() &
            pl.col("end_station_id").is_not_null()
        ).select([
            "end_station_name", "end_station_id", "end_lat", "end_lng"
        ]).unique()
        
        for row in end_stations.iter_rows(named=True):
            station_cache.append({
                'name': row['end_station_name'],
                'id': row['end_station_id'],
                'lat': row['end_lat'],
                'lng': row['end_lng']
            })
    
    # Удаляем дубликаты
    unique_stations = {}
    for station in station_cache:
        key = (station['name'], station['id'])
        if key not in unique_stations:
            unique_stations[key] = station
    
    return list(unique_stations.values())

def find_nearest_station_simple(lat, lng, stations, max_distance_km=0.1):
    """Находит ближайшую станцию к заданным координатам (простая версия)"""
    if lat is None or lng is None or not stations:
        return None, None, False
    
    min_distance = float('inf')
    nearest_station = None
    
    for station in stations:
        # Рассчитываем евклидово расстояние в градусах (быстро)
        dist = ((lat - station['lat']) ** 2 + (lng - station['lng']) ** 2) ** 0.5
        
        # ~0.001 градуса ≈ 111 метров
        if dist < min_distance and dist < max_distance_km / 111.0:
            min_distance = dist
            nearest_station = station
    
    if nearest_station:
        return nearest_station['name'], nearest_station['id'], True
    else:
        return "Non-Station Parking", "NON_STATION", False

def fill_missing_stations_vectorized(df, station_cache):
    """
    Заполняет пропуски в станциях с использованием векторизованных операций Polars
    """
    # Подготавливаем данные
    df = prepare_dataframe(df)
    
    # Проверяем, есть ли колонки станций
    has_start_station = "start_station_name" in df.columns
    has_end_station = "end_station_name" in df.columns
    
    if not has_start_station or not has_end_station:
        print(f"Предупреждение: DataFrame не содержит полных колонок станций")
        return df
    
    # Создаем UDF функции для поиска ближайших станций
    def fill_start_station_udf(row):
        lat, lng, name, id_ = row
        if name is not None and id_ is not None:
            return (name, id_, False)
        return find_nearest_station_simple(lat, lng, station_cache)
    
    def fill_end_station_udf(row):
        lat, lng, name, id_ = row
        if name is not None and id_ is not None:
            return (name, id_, False)
        return find_nearest_station_simple(lat, lng, station_cache)
    
    # Заполняем стартовые станции
    df = df.with_columns([
        pl.struct(["start_lat", "start_lng", "start_station_name", "start_station_id"])
        .map_elements(
            lambda x: fill_start_station_udf((x["start_lat"], x["start_lng"], 
                                             x["start_station_name"], x["start_station_id"])),
            return_dtype=pl.List(pl.Utf8)
        ).alias("_temp_start_fill")
    ])
    
    df = df.with_columns([
        pl.col("_temp_start_fill").list.get(0).alias("start_station_name_filled"),
        pl.col("_temp_start_fill").list.get(1).alias("start_station_id_filled"),
        pl.col("_temp_start_fill").list.get(2).cast(pl.Boolean).alias("start_station_was_filled")
    ]).drop("_temp_start_fill")
    
    # Заполняем конечные станции
    df = df.with_columns([
        pl.struct(["end_lat", "end_lng", "end_station_name", "end_station_id"])
        .map_elements(
            lambda x: fill_end_station_udf((x["end_lat"], x["end_lng"], 
                                           x["end_station_name"], x["end_station_id"])),
            return_dtype=pl.List(pl.Utf8)
        ).alias("_temp_end_fill")
    ])
    
    df = df.with_columns([
        pl.col("_temp_end_fill").list.get(0).alias("end_station_name_filled"),
        pl.col("_temp_end_fill").list.get(1).alias("end_station_id_filled"),
        pl.col("_temp_end_fill").list.get(2).cast(pl.Boolean).alias("end_station_was_filled")
    ]).drop("_temp_end_fill")
    
    # Заменяем оригинальные колонки заполненными значениями
    df = df.with_columns([
        pl.col("start_station_name_filled").alias("start_station_name"),
        pl.col("start_station_id_filled").alias("start_station_id"),
        pl.col("end_station_name_filled").alias("end_station_name"),
        pl.col("end_station_id_filled").alias("end_station_id")
    ]).drop(["start_station_name_filled", "start_station_id_filled",
             "end_station_name_filled", "end_station_id_filled"])
    
    return df

def create_master_station_cache(dataframes):
    """Создает общий кэш станций из всех годов данных"""
    master_cache = []
    
    for year, df in dataframes.items():
        print(f"Собираем станции из {year} года...")
        df_prepared = prepare_dataframe(df)
        year_cache = create_station_cache_from_data(df_prepared)
        master_cache.extend(year_cache)
    
    # Удаляем дубликаты
    unique_stations = {}
    for station in master_cache:
        key = (station['name'], station['id'])
        if key not in unique_stations:
            unique_stations[key] = station
        else:
            # Если уже есть, усредняем координаты для большей точности
            existing = unique_stations[key]
            existing['lat'] = (existing['lat'] + station['lat']) / 2
            existing['lng'] = (existing['lng'] + station['lng']) / 2
    
    print(f"Создан общий кэш из {len(unique_stations)} уникальных станций")
    return list(unique_stations.values())

def process_year_with_cache(df, year, station_cache, output_dir):
    """Обрабатывает данные одного года с использованием общего кэша станций"""
    print(f"\nОбработка данных за {year} год...")
    print(f"Размер данных: {df.height:,} строк")
    
    # Проверяем наличие пропусков
    if "start_station_name" in df.columns:
        missing_start = df.filter(
            pl.col("start_station_name").is_null() | 
            (pl.col("start_station_name") == "None")
        ).height
        print(f"Пропусков в start_station_name: {missing_start:,} ({missing_start/df.height*100:.1f}%)")
    
    if "end_station_name" in df.columns:
        missing_end = df.filter(
            pl.col("end_station_name").is_null() | 
            (pl.col("end_station_name") == "None")
        ).height
        print(f"Пропусков в end_station_name: {missing_end:,} ({missing_end/df.height*100:.1f}%)")
    
    # Заполняем пропуски
    df_filled = fill_missing_stations_vectorized(df, station_cache)
    
    # Проверяем результат
    if "start_station_name" in df_filled.columns:
        non_station_start = df_filled.filter(
            pl.col("start_station_name") == "Non-Station Parking"
        ).height
        print(f"Поездок без станции начала после заполнения: {non_station_start:,} ({non_station_start/df_filled.height*100:.1f}%)")
    
    if "end_station_name" in df_filled.columns:
        non_station_end = df_filled.filter(
            pl.col("end_station_name") == "Non-Station Parking"
        ).height
        print(f"Поездок без станции окончания после заполнения: {non_station_end:,} ({non_station_end/df_filled.height*100:.1f}%)")
    
    # Сохраняем результат
    output_path = f"{output_dir}/{year}/rides_{year}_filled.parquet"
    df_filled.write_parquet(output_path)
    print(f"Результат сохранен в: {output_path}")
    
    return df_filled

# Основной скрипт
def main():
    # Загружаем данные
    print("Загрузка данных...")
    data_paths = {
        2020: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2020/cleaned_2020_data_filtered.parquet",
        2021: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2021/cleaned_scooter_data_2021.parquet",
        2022: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2022/cleaned_scooter_data_2022.parquet",
        2023: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2023/rides_2023_cleaned.parquet",
        2024: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2024/rides_2024_cleaned.parquet",
        2025: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2025/rides_2025_cleaned.parquet"
    }
    
    dataframes = {}
    for year, path in data_paths.items():
        try:
            dataframes[year] = pl.read_parquet(path)
            print(f"Загружены данные {year} года: {dataframes[year].height:,} строк")
        except Exception as e:
            print(f"Ошибка загрузки данных {year} года: {e}")
    
    # Создаем общий кэш станций
    print("\nСоздание общего кэша станций...")
    master_station_cache = create_master_station_cache(dataframes)
    
    # Обрабатываем каждый год
    output_dir = "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed"
    processed_dfs = {}
    
    for year, df in dataframes.items():
        try:
            processed_dfs[year] = process_year_with_cache(
                df, year, master_station_cache, output_dir
            )
        except Exception as e:
            print(f"Ошибка обработки {year} года: {e}")
            import traceback
            traceback.print_exc()
    
    # Сводная статистика
    print("\n" + "="*60)
    print("СВОДНАЯ СТАТИСТИКА ПО ГОДАМ")
    print("="*60)
    
    summary_data = []
    for year, df in processed_dfs.items():
        if df is not None:
            total = df.height
            if "start_station_name" in df.columns:
                non_station_start = df.filter(
                    pl.col("start_station_name") == "Non-Station Parking"
                ).height
            else:
                non_station_start = 0
            
            if "end_station_name" in df.columns:
                non_station_end = df.filter(
                    pl.col("end_station_name") == "Non-Station Parking"
                ).height
            else:
                non_station_end = 0
            
            summary_data.append({
                "year": year,
                "total_rides": total,
                "non_station_start": non_station_start,
                "non_station_start_pct": non_station_start/total*100 if total > 0 else 0,
                "non_station_end": non_station_end,
                "non_station_end_pct": non_station_end/total*100 if total > 0 else 0
            })
    
    # Выводим таблицу с результатами
    summary_df = pl.DataFrame(summary_data)
    print(summary_df)
    
    # Сохраняем сводную статистику
    summary_df.write_csv(f"{output_dir}/filling_summary.csv")
    print(f"\nСводная статистика сохранена в: {output_dir}/filling_summary.csv")
    
    # Примеры поездок без станций
    print("\n" + "="*60)
    print("ПРИМЕРЫ ПОЕЗДОК БЕЗ СТАНЦИЙ (2022 год):")
    print("="*60)
    
    if 2022 in processed_dfs and processed_dfs[2022] is not None:
        sample_non_station = processed_dfs[2022].filter(
            (pl.col("start_station_name") == "Non-Station Parking") |
            (pl.col("end_station_name") == "Non-Station Parking")
        ).select([
            "ride_id", "started_at", "ended_at", 
            "start_station_name", "end_station_name",
            "start_lat", "start_lng", "end_lat", "end_lng"
        ]).head(5)
        
        print(sample_non_station)

# Альтернативный упрощенный метод для быстрой обработки
def simple_fill_missing_stations(df, station_cache):
    """Упрощенный метод заполнения пропусков - для каждого года отдельно"""
    df = prepare_dataframe(df)
    
    # Функция для заполнения одной станции
    def fill_station_info(lat, lng, orig_name, orig_id):
        if orig_name is not None and orig_id is not None and orig_name != "None":
            return orig_name, orig_id, False
        
        # Ищем ближайшую станцию
        if lat is None or lng is None or not station_cache:
            return "Non-Station Parking", "NON_STATION", False
        
        nearest_dist = float('inf')
        nearest_station = None
        
        for station in station_cache:
            # Быстрое приблизительное расстояние
            dist = abs(lat - station['lat']) + abs(lng - station['lng'])
            if dist < nearest_dist and dist < 0.01:  # ~1.1 км
                nearest_dist = dist
                nearest_station = station
        
        if nearest_station:
            return nearest_station['name'], nearest_station['id'], True
        else:
            return "Non-Station Parking", "NON_STATION", False
    
    # Применяем к стартовым станциям
    if "start_station_name" in df.columns:
        df = df.with_columns([
            pl.struct(["start_lat", "start_lng", "start_station_name", "start_station_id"])
            .map_elements(
                lambda x: fill_station_info(x["start_lat"], x["start_lng"], 
                                           x["start_station_name"], x["start_station_id"]),
                return_dtype=pl.List(pl.String)
            ).alias("_temp_start")
        ])
        
        df = df.with_columns([
            pl.col("_temp_start").list.get(0).alias("start_station_name"),
            pl.col("_temp_start").list.get(1).alias("start_station_id"),
            pl.col("_temp_start").list.get(2).cast(pl.Boolean).alias("station_filled_start")
        ]).drop("_temp_start")
    
    # Применяем к конечным станциям
    if "end_station_name" in df.columns:
        df = df.with_columns([
            pl.struct(["end_lat", "end_lng", "end_station_name", "end_station_id"])
            .map_elements(
                lambda x: fill_station_info(x["end_lat"], x["end_lng"], 
                                           x["end_station_name"], x["end_station_id"]),
                return_dtype=pl.List(pl.String)
            ).alias("_temp_end")
        ])
        
        df = df.with_columns([
            pl.col("_temp_end").list.get(0).alias("end_station_name"),
            pl.col("_temp_end").list.get(1).alias("end_station_id"),
            pl.col("_temp_end").list.get(2).cast(pl.Boolean).alias("station_filled_end")
        ]).drop("_temp_end")
    
    return df

if __name__ == "__main__":
    # Запуск основного скрипта
    main()
    
    # Альтернативно: быстрая обработка каждого года отдельно
    print("\n" + "="*60)
    print("АЛЬТЕРНАТИВНАЯ БЫСТРАЯ ОБРАБОТКА")
    print("="*60)
    
    # Обрабатываем каждый год отдельно с его собственным кэшем
    for year in [2020, 2021, 2022, 2023, 2024, 2025]:
        try:
            print(f"\nБыстрая обработка {year} года...")
            
            # Загружаем данные
            path = f"/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/{year}/"
            if year == 2020:
                df = pl.read_parquet(path + "cleaned_2020_data_filtered.parquet")
            elif year == 2021:
                df = pl.read_parquet(path + "cleaned_scooter_data_2021.parquet")
            elif year == 2022:
                df = pl.read_parquet(path + "cleaned_scooter_data_2022.parquet")
            else:
                df = pl.read_parquet(path + f"rides_{year}_cleaned.parquet")
            
            # Создаем кэш станций для этого года
            station_cache = create_station_cache_from_data(prepare_dataframe(df))
            print(f"  Создан кэш из {len(station_cache)} станций")
            
            # Заполняем пропуски
            df_filled = simple_fill_missing_stations(df, station_cache)
            
            # Сохраняем результат
            output_path = f"/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/{year}/rides_{year}_simple_filled.parquet"
            df_filled.write_parquet(output_path)
            print(f"  Результат сохранен в: {output_path}")
            
            # Статистика
            total = df_filled.height
            non_station = df_filled.filter(
                pl.col("start_station_name") == "Non-Station Parking"
            ).height
            print(f"  Всего поездок: {total:,}")
            print(f"  Поездок без станций: {non_station:,} ({non_station/total*100:.1f}%)")
            
        except Exception as e:
            print(f"Ошибка обработки {year} года: {e}")
```

```python
import polars as pl
import numpy as np
from datetime import datetime
import os

def prepare_station_columns(df):
    """Подготавливает колонки станций к обработке"""
    df = df.clone()
    
    # Определяем колонки станций, которые могут быть в данных
    station_cols = ["start_station_name", "start_station_id", 
                    "end_station_name", "end_station_id"]
    
    # Приводим колонки станций к строковому типу и заменяем "None" на None
    for col in station_cols:
        if col in df.columns:
            # Сначала конвертируем в строки
            df = df.with_columns(
                pl.col(col).cast(pl.String).alias(col)
            )
            # Затем заменяем "None" на None
            df = df.with_columns(
                pl.when(pl.col(col) == "None")
                .then(None)
                .otherwise(pl.col(col))
                .alias(col)
            )
    
    return df

def create_station_cache_from_df(df):
    """Создает кэш станций из DataFrame"""
    station_cache = []
    
    # Обрабатываем DataFrame в пакетах для экономии памяти
    batch_size = 100000
    total_rows = df.height
    
    for start in range(0, total_rows, batch_size):
        end = min(start + batch_size, total_rows)
        batch = df.slice(start, end - start)
        
        # Собираем стартовые станции
        if "start_station_name" in df.columns:
            start_stations = batch.filter(
                pl.col("start_station_name").is_not_null() &
                pl.col("start_station_id").is_not_null()
            ).select([
                "start_station_name", "start_station_id", "start_lat", "start_lng"
            ])
            
            for row in start_stations.iter_rows(named=True):
                if row['start_station_name'] and row['start_station_id']:
                    station_cache.append({
                        'name': str(row['start_station_name']),
                        'id': str(row['start_station_id']),
                        'lat': float(row['start_lat']),
                        'lng': float(row['start_lng'])
                    })
        
        # Собираем конечные станции
        if "end_station_name" in df.columns:
            end_stations = batch.filter(
                pl.col("end_station_name").is_not_null() &
                pl.col("end_station_id").is_not_null()
            ).select([
                "end_station_name", "end_station_id", "end_lat", "end_lng"
            ])
            
            for row in end_stations.iter_rows(named=True):
                if row['end_station_name'] and row['end_station_id']:
                    station_cache.append({
                        'name': str(row['end_station_name']),
                        'id': str(row['end_station_id']),
                        'lat': float(row['end_lat']),
                        'lng': float(row['end_lng'])
                    })
    
    # Удаляем дубликаты
    unique_stations = {}
    for station in station_cache:
        key = (station['name'], station['id'])
        if key not in unique_stations:
            unique_stations[key] = station
        else:
            # Усредняем координаты для одинаковых станций
            existing = unique_stations[key]
            existing['lat'] = (existing['lat'] + station['lat']) / 2
            existing['lng'] = (existing['lng'] + station['lng']) / 2
    
    return list(unique_stations.values())

def haversine_distance(lat1, lon1, lat2, lon2):
    """Вычисляет расстояние в метрах между двумя точками"""
    # Земной радиус в метрах
    R = 6371000
    
    # Конвертируем градусы в радианы
    lat1_rad = np.radians(lat1)
    lon1_rad = np.radians(lon1)
    lat2_rad = np.radians(lat2)
    lon2_rad = np.radians(lon2)
    
    # Разницы
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    
    # Формула гаверсинуса
    a = np.sin(dlat/2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    return R * c

def find_nearest_station_for_batch(coords_batch, station_cache, max_distance_m=100):
    """
    Находит ближайшие станции для пакета координат
    coords_batch: список кортежей (lat, lng)
    """
    results = []
    
    for lat, lng in coords_batch:
        if lat is None or lng is None or not station_cache:
            results.append(("Non-Station Parking", "NON_STATION", False))
            continue
        
        min_distance = float('inf')
        nearest_station = None
        
        for station in station_cache:
            distance = haversine_distance(lat, lng, station['lat'], station['lng'])
            if distance < min_distance and distance <= max_distance_m:
                min_distance = distance
                nearest_station = station
        
        if nearest_station:
            results.append((nearest_station['name'], nearest_station['id'], True))
        else:
            results.append(("Non-Station Parking", "NON_STATION", False))
    
    return results

def fill_stations_for_year(df, year, output_dir, use_master_cache=True, master_cache=None):
    """Заполняет пропущенные станции для одного года"""
    print(f"\n{'='*60}")
    print(f"Обработка {year} года")
    print(f"{'='*60}")
    
    # Подготавливаем данные
    df_clean = prepare_station_columns(df)
    
    # Проверяем наличие колонок станций
    has_start_station = "start_station_name" in df_clean.columns
    has_end_station = "end_station_name" in df_clean.columns
    
    if not has_start_station or not has_end_station:
        print(f"  Предупреждение: Нет полных колонок станций для {year} года")
        return df_clean
    
    # Считаем пропуски до обработки
    total_rows = df_clean.height
    
    missing_start_before = df_clean.filter(
        pl.col("start_station_name").is_null()
    ).height
    
    missing_end_before = df_clean.filter(
        pl.col("end_station_name").is_null()
    ).height
    
    print(f"  Всего поездок: {total_rows:,}")
    print(f"  Пропущено стартовых станций: {missing_start_before:,} ({missing_start_before/total_rows*100:.1f}%)")
    print(f"  Пропущено конечных станций: {missing_end_before:,} ({missing_end_before/total_rows*100:.1f}%)")
    
    # Создаем или используем кэш станций
    if use_master_cache and master_cache is not None:
        station_cache = master_cache
        print(f"  Используется общий кэш из {len(station_cache)} станций")
    else:
        print(f"  Создание кэша станций из данных {year} года...")
        station_cache = create_station_cache_from_df(df_clean)
        print(f"  Создан кэш из {len(station_cache)} станций")
    
    if not station_cache:
        print(f"  Предупреждение: Не удалось создать кэш станций")
        return df_clean
    
    # Заполняем стартовые станции
    print(f"  Заполнение стартовых станций...")
    
    # Получаем координаты для строк с пропущенными станциями
    rows_with_missing_start = df_clean.filter(
        pl.col("start_station_name").is_null()
    ).select(["start_lat", "start_lng"]).to_numpy()
    
    if len(rows_with_missing_start) > 0:
        # Находим ближайшие станции для пропущенных
        start_coords = [(float(row[0]), float(row[1])) for row in rows_with_missing_start]
        filled_start_stations = find_nearest_station_for_batch(start_coords, station_cache)
        
        # Создаем DataFrame с заполненными значениями
        filled_start_df = pl.DataFrame({
            "temp_start_name": [x[0] for x in filled_start_stations],
            "temp_start_id": [x[1] for x in filled_start_stations],
            "temp_start_filled": [x[2] for x in filled_start_stations]
        })
        
        # Получаем индексы строк с пропущенными станциями
        missing_start_indices = df_clean.filter(
            pl.col("start_station_name").is_null()
        ).select(pl.arange(0, pl.len()).alias("idx"))
        
        # Объединяем с оригинальным DataFrame
        df_with_indices = df_clean.with_columns(
            pl.arange(0, pl.len()).alias("original_index")
        )
        
        # Присоединяем заполненные значения
        df_filled = df_with_indices.join(
            missing_start_indices.with_columns(
                pl.Series("fill_idx", range(len(missing_start_indices)))
            ).hstack(filled_start_df),
            left_on="original_index",
            right_on="idx",
            how="left"
        )
        
        # Обновляем стартовые станции
        df_filled = df_filled.with_columns([
            pl.when(pl.col("temp_start_name").is_not_null())
            .then(pl.col("temp_start_name"))
            .otherwise(pl.col("start_station_name"))
            .alias("start_station_name"),
            
            pl.when(pl.col("temp_start_id").is_not_null())
            .then(pl.col("temp_start_id"))
            .otherwise(pl.col("start_station_id"))
            .alias("start_station_id"),
            
            pl.when(pl.col("temp_start_filled").is_not_null())
            .then(pl.col("temp_start_filled"))
            .otherwise(pl.lit(False))
            .alias("start_station_was_filled")
        ]).drop(["original_index", "idx", "fill_idx", 
                 "temp_start_name", "temp_start_id", "temp_start_filled"])
    else:
        df_filled = df_clean.with_columns(
            pl.lit(False).alias("start_station_was_filled")
        )
    
    # Заполняем конечные станции
    print(f"  Заполнение конечных станций...")
    
    rows_with_missing_end = df_filled.filter(
        pl.col("end_station_name").is_null()
    ).select(["end_lat", "end_lng"]).to_numpy()
    
    if len(rows_with_missing_end) > 0:
        # Находим ближайшие станции для пропущенных
        end_coords = [(float(row[0]), float(row[1])) for row in rows_with_missing_end]
        filled_end_stations = find_nearest_station_for_batch(end_coords, station_cache)
        
        # Создаем DataFrame с заполненными значениями
        filled_end_df = pl.DataFrame({
            "temp_end_name": [x[0] for x in filled_end_stations],
            "temp_end_id": [x[1] for x in filled_end_stations],
            "temp_end_filled": [x[2] for x in filled_end_stations]
        })
        
        # Получаем индексы строк с пропущенными станциями
        missing_end_indices = df_filled.filter(
            pl.col("end_station_name").is_null()
        ).select(pl.arange(0, pl.len()).alias("idx"))
        
        # Объединяем с DataFrame
        df_with_indices = df_filled.with_columns(
            pl.arange(0, pl.len()).alias("original_index")
        )
        
        # Присоединяем заполненные значения
        df_filled2 = df_with_indices.join(
            missing_end_indices.with_columns(
                pl.Series("fill_idx", range(len(missing_end_indices)))
            ).hstack(filled_end_df),
            left_on="original_index",
            right_on="idx",
            how="left"
        )
        
        # Обновляем конечные станции
        df_filled2 = df_filled2.with_columns([
            pl.when(pl.col("temp_end_name").is_not_null())
            .then(pl.col("temp_end_name"))
            .otherwise(pl.col("end_station_name"))
            .alias("end_station_name"),
            
            pl.when(pl.col("temp_end_id").is_not_null())
            .then(pl.col("temp_end_id"))
            .otherwise(pl.col("end_station_id"))
            .alias("end_station_id"),
            
            pl.when(pl.col("temp_end_filled").is_not_null())
            .then(pl.col("temp_end_filled"))
            .otherwise(pl.lit(False))
            .alias("end_station_was_filled")
        ]).drop(["original_index", "idx", "fill_idx", 
                 "temp_end_name", "temp_end_id", "temp_end_filled"])
    else:
        df_filled2 = df_filled.with_columns(
            pl.lit(False).alias("end_station_was_filled")
        )
    
    # Считаем результаты
    non_station_start = df_filled2.filter(
        pl.col("start_station_name") == "Non-Station Parking"
    ).height
    
    non_station_end = df_filled2.filter(
        pl.col("end_station_name") == "Non-Station Parking"
    ).height
    
    filled_start = df_filled2.filter(
        pl.col("start_station_was_filled") == True
    ).height
    
    filled_end = df_filled2.filter(
        pl.col("end_station_was_filled") == True
    ).height
    
    print(f"  Результаты:")
    print(f"  • Поездок без стартовой станции: {non_station_start:,} ({non_station_start/total_rows*100:.1f}%)")
    print(f"  • Поездок без конечной станции: {non_station_end:,} ({non_station_end/total_rows*100:.1f}%)")
    print(f"  • Стартовых станций заполнено: {filled_start:,} ({filled_start/total_rows*100:.1f}%)")
    print(f"  • Конечных станций заполнено: {filled_end:,} ({filled_end/total_rows*100:.1f}%)")
    
    # Сохраняем результат
    output_path = f"{output_dir}/{year}/rides_{year}_filled.parquet"
    df_filled2.write_parquet(output_path)
    print(f"  Результат сохранен в: {output_path}")
    
    return df_filled2

def create_master_station_cache_all_years(dataframes):
    """Создает общий кэш станций из всех годов"""
    print("Создание общего кэша станций из всех лет...")
    
    all_stations = []
    
    for year, df in dataframes.items():
        print(f"  Сбор станций из {year} года...")
        df_prepared = prepare_station_columns(df)
        year_stations = create_station_cache_from_df(df_prepared)
        all_stations.extend(year_stations)
        print(f"    Добавлено {len(year_stations)} станций")
    
    # Удаляем дубликаты
    unique_stations = {}
    for station in all_stations:
        key = (station['name'], station['id'])
        if key not in unique_stations:
            unique_stations[key] = station
        else:
            # Усредняем координаты
            existing = unique_stations[key]
            existing['lat'] = (existing['lat'] + station['lat']) / 2
            existing['lng'] = (existing['lng'] + station['lng']) / 2
    
    master_cache = list(unique_stations.values())
    print(f"Общий кэш создан: {len(master_cache)} уникальных станций")
    
    return master_cache

def main():
    """Основная функция"""
    print("=" * 60)
    print("ЗАПОЛНЕНИЕ ПРОПУЩЕННЫХ СТАНЦИЙ В ДАННЫХ DIVVY BIKES")
    print("=" * 60)
    
    # Загружаем данные
    print("\nЗагрузка данных...")
    
    data_paths = {
        2020: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2020/cleaned_2020_data_filtered.parquet",
        2021: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2021/cleaned_scooter_data_2021.parquet",
        2022: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2022/cleaned_scooter_data_2022.parquet",
        2023: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2023/rides_2023_cleaned.parquet",
        2024: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2024/rides_2024_cleaned.parquet",
        2025: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2025/rides_2025_cleaned.parquet"
    }
    
    dataframes = {}
    for year, path in data_paths.items():
        try:
            dataframes[year] = pl.read_parquet(path)
            print(f"  Загружены данные {year} года: {dataframes[year].height:,} строк")
        except Exception as e:
            print(f"  Ошибка загрузки {year} года: {e}")
    
    # Создаем общий кэш станций
    master_cache = create_master_station_cache_all_years(dataframes)
    
    # Обрабатываем каждый год
    output_dir = "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed"
    processed_dfs = {}
    
    # Создаем сводную таблицу результатов
    summary_data = []
    
    for year in [2020, 2021, 2022, 2023, 2024, 2025]:
        if year in dataframes:
            try:
                df_processed = fill_stations_for_year(
                    dataframes[year], 
                    year, 
                    output_dir,
                    use_master_cache=True,
                    master_cache=master_cache
                )
                processed_dfs[year] = df_processed
                
                # Собираем статистику для сводки
                total = df_processed.height
                non_station_start = df_processed.filter(
                    pl.col("start_station_name") == "Non-Station Parking"
                ).height
                non_station_end = df_processed.filter(
                    pl.col("end_station_name") == "Non-Station Parking"
                ).height
                
                summary_data.append({
                    "year": year,
                    "total_rides": total,
                    "non_station_start": non_station_start,
                    "non_station_start_pct": non_station_start/total*100,
                    "non_station_end": non_station_end,
                    "non_station_end_pct": non_station_end/total*100
                })
                
            except Exception as e:
                print(f"\nОшибка обработки {year} года: {e}")
                import traceback
                traceback.print_exc()
    
    # Выводим сводную таблицу
    if summary_data:
        print("\n" + "="*60)
        print("СВОДНАЯ СТАТИСТИКА")
        print("="*60)
        
        summary_df = pl.DataFrame(summary_data)
        print(summary_df)
        
        # Сохраняем сводную статистику
        summary_path = f"{output_dir}/station_filling_summary.csv"
        summary_df.write_csv(summary_path)
        print(f"\nСводная статистика сохранена в: {summary_path}")
        
        # Выводим примеры поездок без станций
        print("\n" + "="*60)
        print("ПРИМЕРЫ ПОЕЗДОК БЕЗ СТАНЦИЙ (2022 год):")
        print("="*60)
        
        if 2022 in processed_dfs:
            sample = processed_dfs[2022].filter(
                (pl.col("start_station_name") == "Non-Station Parking") |
                (pl.col("end_station_name") == "Non-Station Parking")
            ).select([
                "ride_id", "rideable_type", "started_at", "ended_at",
                "start_station_name", "end_station_name",
                "start_lat", "start_lng", "end_lat", "end_lng",
                "member_casual", "duration_minutes"
            ]).head(5)
            
            print(sample)
    else:
        print("\nНе удалось обработать ни одного года данных")
    
    print("\n" + "="*60)
    print("ОБРАБОТКА ЗАВЕРШЕНА")
    print("="*60)

# Упрощенная версия для быстрой обработки (без создания общего кэша)
def simple_fill_stations(df, year, output_dir):
    """Упрощенная версия заполнения станций"""
    print(f"\nПростая обработка {year} года...")
    
    # Подготавливаем данные
    df = prepare_station_columns(df)
    
    # Создаем кэш станций из этого года
    station_cache = create_station_cache_from_df(df)
    print(f"  Создан кэш из {len(station_cache)} станций")
    
    # Функция для заполнения одной станции
    def fill_station(row, station_type="start"):
        if station_type == "start":
            lat = row['start_lat']
            lng = row['start_lng']
            orig_name = row['start_station_name']
            orig_id = row['start_station_id']
        else:
            lat = row['end_lat']
            lng = row['end_lng']
            orig_name = row['end_station_name']
            orig_id = row['end_station_id']
        
        # Если уже есть значение, возвращаем его
        if orig_name is not None and orig_id is not None:
            return orig_name, orig_id, False
        
        # Ищем ближайшую станцию
        if lat is None or lng is None or not station_cache:
            return "Non-Station Parking", "NON_STATION", False
        
        min_distance = float('inf')
        nearest_station = None
        
        for station in station_cache:
            # Быстрое приблизительное расстояние в градусах
            distance = abs(lat - station['lat']) + abs(lng - station['lng'])
            if distance < min_distance and distance < 0.001:  # ~111 метров
                min_distance = distance
                nearest_station = station
        
        if nearest_station:
            return nearest_station['name'], nearest_station['id'], True
        else:
            return "Non-Station Parking", "NON_STATION", False
    
    # Обрабатываем небольшими пакетами
    batch_size = 100000
    total_rows = df.height
    result_rows = []
    
    for start in range(0, total_rows, batch_size):
        end = min(start + batch_size, total_rows)
        batch = df.slice(start, end - start)
        
        batch_rows = []
        for row in batch.iter_rows(named=True):
            # Заполняем стартовую станцию
            start_name, start_id, start_filled = fill_station(row, "start")
            # Заполняем конечную станцию
            end_name, end_id, end_filled = fill_station(row, "end")
            
            # Обновляем строку
            row['start_station_name'] = start_name
            row['start_station_id'] = start_id
            row['start_station_was_filled'] = start_filled
            row['end_station_name'] = end_name
            row['end_station_id'] = end_id
            row['end_station_was_filled'] = end_filled
            
            batch_rows.append(row)
        
        result_rows.extend(batch_rows)
        print(f"  Обработано {min(end, total_rows):,} из {total_rows:,} строк")
    
    # Создаем новый DataFrame
    result_df = pl.DataFrame(result_rows)
    
    # Сохраняем
    output_path = f"{output_dir}/{year}/rides_{year}_simple_filled.parquet"
    result_df.write_parquet(output_path)
    print(f"  Результат сохранен в: {output_path}")
    
    return result_df

if __name__ == "__main__":
    # Запускаем основную обработку
    main()
    
    # Альтернатива: простая обработка каждого года
    print("\n" + "="*60)
    print("АЛЬТЕРНАТИВНАЯ ПРОСТАЯ ОБРАБОТКА")
    print("="*60)
    
    output_dir = "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed"
    
    for year in [2020, 2021, 2022, 2023, 2024, 2025]:
        try:
            path = f"/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/{year}/"
            if year == 2020:
                df = pl.read_parquet(path + "cleaned_2020_data_filtered.parquet")
            elif year == 2021:
                df = pl.read_parquet(path + "cleaned_scooter_data_2021.parquet")
            elif year == 2022:
                df = pl.read_parquet(path + "cleaned_scooter_data_2022.parquet")
            else:
                df = pl.read_parquet(path + f"rides_{year}_cleaned.parquet")
            
            simple_fill_stations(df, year, output_dir)
            
        except Exception as e:
            print(f"Ошибка простой обработки {year} года: {e}")
```

```python
import polars as pl
import numpy as np
import time
from datetime import datetime

# ------------------------------------------------------------------
# БАЗОВЫЕ ФУНКЦИИ ДЛЯ ПОДГОТОВКИ ДАННЫХ И СОЗДАНИЯ КЭША СТАНЦИЙ
# ------------------------------------------------------------------

def prepare_station_columns(df):
    """Подготавливает колонки станций к обработке"""
    df = df.clone()
    
    # Определяем колонки станций, которые могут быть в данных
    station_cols = ["start_station_name", "start_station_id", 
                    "end_station_name", "end_station_id"]
    
    # Приводим колонки станций к строковому типу и заменяем "None" на None
    for col in station_cols:
        if col in df.columns:
            # Сначала конвертируем в строки
            df = df.with_columns(
                pl.col(col).cast(pl.String).alias(col)
            )
            # Затем заменяем "None" на None
            df = df.with_columns(
                pl.when(pl.col(col) == "None")
                .then(None)
                .otherwise(pl.col(col))
                .alias(col)
            )
    
    return df

def create_station_cache_from_df(df):
    """Создает кэш станций из DataFrame"""
    station_cache = []
    
    # Обрабатываем DataFrame в пакетах для экономии памяти
    batch_size = 100000
    total_rows = df.height
    
    for start in range(0, total_rows, batch_size):
        end = min(start + batch_size, total_rows)
        batch = df.slice(start, end - start)
        
        # Собираем стартовые станции
        if "start_station_name" in df.columns:
            start_stations = batch.filter(
                pl.col("start_station_name").is_not_null() &
                pl.col("start_station_id").is_not_null()
            ).select([
                "start_station_name", "start_station_id", "start_lat", "start_lng"
            ])
            
            for row in start_stations.iter_rows(named=True):
                if row['start_station_name'] and row['start_station_id']:
                    station_cache.append({
                        'name': str(row['start_station_name']),
                        'id': str(row['start_station_id']),
                        'lat': float(row['start_lat']),
                        'lng': float(row['start_lng'])
                    })
        
        # Собираем конечные станции
        if "end_station_name" in df.columns:
            end_stations = batch.filter(
                pl.col("end_station_name").is_not_null() &
                pl.col("end_station_id").is_not_null()
            ).select([
                "end_station_name", "end_station_id", "end_lat", "end_lng"
            ])
            
            for row in end_stations.iter_rows(named=True):
                if row['end_station_name'] and row['end_station_id']:
                    station_cache.append({
                        'name': str(row['end_station_name']),
                        'id': str(row['end_station_id']),
                        'lat': float(row['end_lat']),
                        'lng': float(row['end_lng'])
                    })
    
    # Удаляем дубликаты
    unique_stations = {}
    for station in station_cache:
        key = (station['name'], station['id'])
        if key not in unique_stations:
            unique_stations[key] = station
        else:
            # Усредняем координаты для одинаковых станций
            existing = unique_stations[key]
            existing['lat'] = (existing['lat'] + station['lat']) / 2
            existing['lng'] = (existing['lng'] + station['lng']) / 2
    
    return list(unique_stations.values())

# ------------------------------------------------------------------
# ОПТИМИЗИРОВАННЫЕ ФУНКЦИИ ДЛЯ ЗАПОЛНЕНИЯ ПРОПУСКОВ
# ------------------------------------------------------------------

def create_simple_spatial_index(station_cache, grid_size=0.001):
    """Создает простой пространственный индекс без scipy"""
    index = {}
    
    for station in station_cache:
        grid_lat = int(station['lat'] / grid_size)
        grid_lng = int(station['lng'] / grid_size)
        key = (grid_lat, grid_lng)
        
        if key not in index:
            index[key] = []
        
        index[key].append(station)
    
    return index

def find_in_spatial_index(lat, lng, spatial_index, grid_size=0.001):
    """Ищет в простом пространственном индексе"""
    if lat is None or lng is None:
        return "Non-Station Parking", "NON_STATION", False
    
    grid_lat = int(lat / grid_size)
    grid_lng = int(lng / grid_size)
    
    # Ищем в текущей и соседних ячейках
    candidates = []
    
    for dlat in [-1, 0, 1]:
        for dlng in [-1, 0, 1]:
            key = (grid_lat + dlat, grid_lng + dlng)
            if key in spatial_index:
                candidates.extend(spatial_index[key])
    
    if not candidates:
        return "Non-Station Parking", "NON_STATION", False
    
    # Находим ближайшую
    nearest = None
    min_dist = float('inf')
    
    for station in candidates:
        dist = ((lat - station['lat'])**2 + (lng - station['lng'])**2)**0.5
        
        # ~0.001 градуса ≈ 111 метров
        if dist < min_dist and dist < 0.0015:  # ~167 метров
            min_dist = dist
            nearest = station
    
    if nearest:
        return nearest['name'], nearest['id'], True
    else:
        return "Non-Station Parking", "NON_STATION", False

def ultra_fast_fill(df, station_cache):
    """Сверхбыстрое заполнение для 2022 года"""
    print(f"  Сверхбыстрое заполнение для больших данных...")
    start_time = time.time()
    
    # Создаем простой пространственный индекс
    spatial_index = create_simple_spatial_index(station_cache)
    
    # Подготавливаем DataFrame
    df_clean = df.clone()
    
    # Подготавливаем колонки станций
    df_clean = prepare_station_columns(df_clean)
    
    # Обрабатываем только строки с пропусками
    missing_start_mask = df_clean["start_station_name"].is_null()
    missing_end_mask = df_clean["end_station_name"].is_null()
    
    total_missing = missing_start_mask.sum() + missing_end_mask.sum()
    print(f"    Всего пропусков для заполнения: {total_missing:,}")
    
    if total_missing == 0:
        return df_clean.with_columns([
            pl.lit(False).alias("start_station_was_filled"),
            pl.lit(False).alias("end_station_was_filled")
        ])
    
    # Обрабатываем пакетами
    batch_size = 50000
    total_rows = df_clean.height
    
    # Создаем списки для результатов
    start_filled_names = [None] * total_rows
    start_filled_ids = [None] * total_rows
    start_filled_flags = [False] * total_rows
    
    end_filled_names = [None] * total_rows
    end_filled_ids = [None] * total_rows
    end_filled_flags = [False] * total_rows
    
    # Заполняем существующие значения
    for i in range(total_rows):
        if not missing_start_mask[i]:
            start_filled_names[i] = df_clean["start_station_name"][i]
            start_filled_ids[i] = df_clean["start_station_id"][i]
        
        if not missing_end_mask[i]:
            end_filled_names[i] = df_clean["end_station_name"][i]
            end_filled_ids[i] = df_clean["end_station_id"][i]
    
    # Заполняем пропущенные значения пакетами
    for start_idx in range(0, total_rows, batch_size):
        end_idx = min(start_idx + batch_size, total_rows)
        
        for i in range(start_idx, end_idx):
            if missing_start_mask[i]:
                lat = df_clean["start_lat"][i]
                lng = df_clean["start_lng"][i]
                name, id_, filled = find_in_spatial_index(lat, lng, spatial_index)
                start_filled_names[i] = name
                start_filled_ids[i] = id_
                start_filled_flags[i] = filled
            
            if missing_end_mask[i]:
                lat = df_clean["end_lat"][i]
                lng = df_clean["end_lng"][i]
                name, id_, filled = find_in_spatial_index(lat, lng, spatial_index)
                end_filled_names[i] = name
                end_filled_ids[i] = id_
                end_filled_flags[i] = filled
        
        if start_idx % 100000 == 0:
            print(f"    Обработано {end_idx:,} из {total_rows:,} строк")
    
    # Обновляем DataFrame
    df_filled = df_clean.with_columns([
        pl.Series("start_station_name", start_filled_names),
        pl.Series("start_station_id", start_filled_ids),
        pl.Series("start_station_was_filled", start_filled_flags),
        pl.Series("end_station_name", end_filled_names),
        pl.Series("end_station_id", end_filled_ids),
        pl.Series("end_station_was_filled", end_filled_flags)
    ])
    
    elapsed_time = time.time() - start_time
    print(f"    Заполнение завершено за {elapsed_time:.1f} секунд")
    
    return df_filled

# ------------------------------------------------------------------
# ОСНОВНАЯ ФУНКЦИЯ ДЛЯ ОБРАБОТКИ ВСЕХ ЛЕТ
# ------------------------------------------------------------------

def process_year_optimized(df, year, station_cache, output_dir):
    """
    Оптимизированная обработка года
    """
    print(f"\n{'='*60}")
    print(f"ОПТИМИЗИРОВАННАЯ ОБРАБОТКА {year} ГОДА")
    print(f"{'='*60}")
    
    total_rows = df.height
    
    # Подготавливаем данные
    df_prepared = prepare_station_columns(df)
    
    # Подсчитываем пропуски
    missing_start = df_prepared.filter(pl.col("start_station_name").is_null()).height
    missing_end = df_prepared.filter(pl.col("end_station_name").is_null()).height
    
    print(f"  Всего поездок: {total_rows:,}")
    print(f"  Пропущено стартовых станций: {missing_start:,} ({missing_start/total_rows*100:.1f}%)")
    print(f"  Пропущено конечных станций: {missing_end:,} ({missing_end/total_rows*100:.1f}%)")
    
    if missing_start == 0 and missing_end == 0:
        print(f"  Нет пропусков для заполнения")
        # Добавляем флаги заполнения
        df_result = df_prepared.with_columns([
            pl.lit(False).alias("start_station_was_filled"),
            pl.lit(False).alias("end_station_was_filled")
        ])
    else:
        # Используем сверхбыстрый метод
        df_result = ultra_fast_fill(df_prepared, station_cache)
    
    # Статистика результатов
    non_station_start = df_result.filter(
        pl.col("start_station_name") == "Non-Station Parking"
    ).height
    
    non_station_end = df_result.filter(
        pl.col("end_station_name") == "Non-Station Parking"
    ).height
    
    filled_start = df_result.filter(
        pl.col("start_station_was_filled") == True
    ).height
    
    filled_end = df_result.filter(
        pl.col("end_station_was_filled") == True
    ).height
    
    print(f"  Результаты:")
    print(f"  • Поездок без стартовой станции: {non_station_start:,} ({non_station_start/total_rows*100:.1f}%)")
    print(f"  • Поездок без конечной станции: {non_station_end:,} ({non_station_end/total_rows*100:.1f}%)")
    print(f"  • Стартовых станций заполнено: {filled_start:,} ({filled_start/total_rows*100:.1f}%)")
    print(f"  • Конечных станций заполнено: {filled_end:,} ({filled_end/total_rows*100:.1f}%)")
    
    # Сохраняем результат
    output_path = f"{output_dir}/{year}/rides_{year}_optimized_filled.parquet"
    df_result.write_parquet(output_path)
    print(f"  Результат сохранен в: {output_path}")
    
    return df_result

def main_optimized():
    """Основная оптимизированная функция"""
    print("=" * 60)
    print("ОПТИМИЗИРОВАННОЕ ЗАПОЛНЕНИЕ ПРОПУЩЕННЫХ СТАНЦИЙ")
    print("=" * 60)
    
    # Загружаем данные
    print("\nЗагрузка данных...")
    
    data_paths = {
        2020: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2020/cleaned_2020_data_filtered.parquet",
        2021: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2021/cleaned_scooter_data_2021.parquet",
        2022: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2022/cleaned_scooter_data_2022.parquet",
        2023: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2023/rides_2023_cleaned.parquet",
        2024: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2024/rides_2024_cleaned.parquet",
        2025: "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2025/rides_2025_cleaned.parquet"
    }
    
    dataframes = {}
    for year, path in data_paths.items():
        try:
            dataframes[year] = pl.read_parquet(path)
            print(f"  Загружены данные {year} года: {dataframes[year].height:,} строк")
        except Exception as e:
            print(f"  Ошибка загрузки {year} года: {e}")
    
    print("\nСоздание общего кэша станций...")
    master_cache = []
    
    for year, df in dataframes.items():
        print(f"  Сбор станций из {year} года...")
        df_prepared = prepare_station_columns(df)
        year_cache = create_station_cache_from_df(df_prepared)
        master_cache.extend(year_cache)
        print(f"    Добавлено {len(year_cache)} станций")
    
    # Удаляем дубликаты
    unique_stations = {}
    for station in master_cache:
        key = (station['name'], station['id'])
        if key not in unique_stations:
            unique_stations[key] = station
        else:
            existing = unique_stations[key]
            existing['lat'] = (existing['lat'] + station['lat']) / 2
            existing['lng'] = (existing['lng'] + station['lng']) / 2
    
    master_cache = list(unique_stations.values())
    print(f"\nОбщий кэш создан: {len(master_cache)} уникальных станций")
    
    # Обрабатываем каждый год
    output_dir = "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed"
    processed_dfs = {}
    
    # Создаем сводную таблицу результатов
    summary_data = []
    
    # Порядок обработки: сначала года с меньшим количеством пропусков
    processing_order = [2020, 2021, 2025, 2024, 2023, 2022]  # 2022 в конце
    
    for year in processing_order:
        if year in dataframes:
            try:
                df_processed = process_year_optimized(
                    dataframes[year], 
                    year, 
                    master_cache, 
                    output_dir
                )
                
                processed_dfs[year] = df_processed
                
                # Собираем статистику
                total = df_processed.height
                non_station_start = df_processed.filter(
                    pl.col("start_station_name") == "Non-Station Parking"
                ).height
                non_station_end = df_processed.filter(
                    pl.col("end_station_name") == "Non-Station Parking"
                ).height
                
                summary_data.append({
                    "year": year,
                    "total_rides": total,
                    "non_station_start": non_station_start,
                    "non_station_start_pct": non_station_start/total*100 if total > 0 else 0,
                    "non_station_end": non_station_end,
                    "non_station_end_pct": non_station_end/total*100 if total > 0 else 0
                })
                
            except Exception as e:
                print(f"\nОшибка обработки {year} года: {e}")
                import traceback
                traceback.print_exc()
    
    # Выводим сводную таблицу
    if summary_data:
        print("\n" + "="*60)
        print("СВОДНАЯ СТАТИСТИКА")
        print("="*60)
        
        summary_df = pl.DataFrame(summary_data)
        print(summary_df)
        
        # Сохраняем сводную статистику
        summary_path = f"{output_dir}/optimized_filling_summary.csv"
        summary_df.write_csv(summary_path)
        print(f"\nСводная статистика сохранена в: {summary_path}")
        
        # Выводим примеры поездок без станций
        print("\n" + "="*60)
        print("ПРИМЕРЫ ПОЕЗДОК БЕЗ СТАНЦИЙ (2022 год):")
        print("="*60)
        
        if 2022 in processed_dfs:
            sample = processed_dfs[2022].filter(
                (pl.col("start_station_name") == "Non-Station Parking") |
                (pl.col("end_station_name") == "Non-Station Parking")
            ).select([
                "ride_id", "rideable_type", "started_at", "ended_at",
                "start_station_name", "end_station_name",
                "start_lat", "start_lng", "end_lat", "end_lng",
                "member_casual", "duration_minutes"
            ]).head(5)
            
            print(sample)
    else:
        print("\nНе удалось обработать ни одного года данных")
    
    print("\n" + "="*60)
    print("ОПТИМИЗИРОВАННАЯ ОБРАБОТКА ЗАВЕРШЕНА")
    print("="*60)
    
    return processed_dfs, summary_df

# ------------------------------------------------------------------
# СПЕЦИАЛЬНАЯ ОБРАБОТКА ДЛЯ 2022 ГОДА
# ------------------------------------------------------------------

def process_2022_only():
    """Специальная обработка только для 2022 года"""
    print("=" * 60)
    print("СПЕЦИАЛЬНАЯ ОБРАБОТКА 2022 ГОДА")
    print("=" * 60)
    
    # Загружаем данные 2022 года
    path_2022 = "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2022/cleaned_scooter_data_2022.parquet"
    df_2022 = pl.read_parquet(path_2022)
    print(f"Загружены данные 2022 года: {df_2022.height:,} строк")
    
    # Загружаем другие годы для создания кэша
    print("\nСоздание кэша станций из всех годов...")
    
    # Список всех годов
    years = [2020, 2021, 2022, 2023, 2024, 2025]
    all_stations = []
    
    for year in years:
        try:
            if year == 2020:
                path = f"/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/{year}/cleaned_2020_data_filtered.parquet"
            elif year == 2021:
                path = f"/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/{year}/cleaned_scooter_data_{year}.parquet"
            elif year == 2022:
                path = path_2022
            else:
                path = f"/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/{year}/rides_{year}_cleaned.parquet"
            
            df = pl.read_parquet(path)
            df = prepare_station_columns(df)
            stations = create_station_cache_from_df(df)
            all_stations.extend(stations)
            print(f"  Добавлены станции из {year} года: {len(stations)}")
        except Exception as e:
            print(f"  Ошибка загрузки {year} года: {e}")
    
    # Удаляем дубликаты
    unique_stations = {}
    for station in all_stations:
        key = (station['name'], station['id'])
        if key not in unique_stations:
            unique_stations[key] = station
    
    station_cache = list(unique_stations.values())
    print(f"\nСоздан кэш из {len(station_cache)} уникальных станций")
    
    # Заполняем пропуски
    print("\nЗаполнение пропусков...")
    df_filled = ultra_fast_fill(df_2022, station_cache)
    
    # Сохраняем
    output_path = "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/2022/rides_2022_ultra_fast_filled.parquet"
    df_filled.write_parquet(output_path)
    print(f"\nРезультат сохранен в: {output_path}")
    
    return df_filled

# ------------------------------------------------------------------
# ЗАПУСК
# ------------------------------------------------------------------

if __name__ == "__main__":
    print("Выберите вариант обработки:")
    print("1. Обработка всех годов (рекомендуется)")
    print("2. Только 2022 год (самый медленный)")
    print("3. Только годы с пропусками (2022-2025)")
    
    choice = input("Введите номер (1-3): ").strip()
    
    if choice == "1":
        # Обработка всех годов
        processed_dfs, summary_df = main_optimized()
    elif choice == "2":
        # Только 2022 год
        df_2022_filled = process_2022_only()
    elif choice == "3":
        # Только годы с пропусками
        print("\nОбработка годов с пропусками (2022-2025)...")
        
        # Загружаем только нужные годы
        years_to_process = [2022, 2023, 2024, 2025]
        dataframes = {}
        
        for year in years_to_process:
            try:
                if year == 2022:
                    path = f"/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/{year}/cleaned_scooter_data_{year}.parquet"
                else:
                    path = f"/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed/{year}/rides_{year}_cleaned.parquet"
                
                dataframes[year] = pl.read_parquet(path)
                print(f"  Загружены данные {year} года: {dataframes[year].height:,} строк")
            except Exception as e:
                print(f"  Ошибка загрузки {year} года: {e}")
        
        # Создаем кэш
        print("\nСоздание кэша станций...")
        master_cache = []
        
        for year, df in dataframes.items():
            print(f"  Сбор станций из {year} года...")
            df_prepared = prepare_station_columns(df)
            year_cache = create_station_cache_from_df(df_prepared)
            master_cache.extend(year_cache)
            print(f"    Добавлено {len(year_cache)} станций")
        
        # Удаляем дубликаты
        unique_stations = {}
        for station in master_cache:
            key = (station['name'], station['id'])
            if key not in unique_stations:
                unique_stations[key] = station
        
        master_cache = list(unique_stations.values())
        print(f"\nКэш создан: {len(master_cache)} уникальных станций")
        
        # Обрабатываем каждый год
        output_dir = "/Users/vsevolod/Desktop/divvy-bikes-analysis/data/processed"
        
        for year in years_to_process:
            if year in dataframes:
                try:
                    process_year_optimized(
                        dataframes[year], 
                        year, 
                        master_cache, 
                        output_dir
                    )
                except Exception as e:
                    print(f"\nОшибка обработки {year} года: {e}")
    else:
        print("Неверный выбор. Запуск обработки всех годов по умолчанию.")
        processed_dfs, summary_df = main_optimized()
```
